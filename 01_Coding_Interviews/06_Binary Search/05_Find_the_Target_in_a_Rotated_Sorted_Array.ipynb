<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

A rotated sorted array is an array of numbers sorted in ascending order, in which a portion of the array is moved from the beginning to the end. For example, a possible rotation of [1, 2, 3, 4, 5] is [3, 4, 5, 1, 2], where the first two numbers are moved to the end. Given a rotated sorted array of unique numbers, return the index of a target value. If the target value is not present, return -1.

</div>

<h3 class="bbg-h3" style="color: var(--bbg-h3-color, var(--vscode-textLink-foreground, var(--jp-content-link-color, #2563eb))); font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 22px 0 10px;">Example:</h3>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

**Input:** `nums = [8, 9, 1, 2, 3, 4, 5, 6, 7], target = 1`

**Output:** `2`


</div>

<h2 class="bbg-h2" style="color: var(--bbg-title-color, var(--vscode-foreground, var(--jp-content-font-color1, CanvasText))); border-left: 5px solid var(--bbg-accent-green, var(--vscode-charts-green, var(--jp-success-color1, #16a34a))); font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Intuition</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

A naive solution is to iterate through the input array until we find the target number. This strategy takes **Linear Time**, and doesn’t take into account that the input is a rotated sorted array.

Given the array was sorted before it was rotated, we should figure out how **Binary Search** might be useful in finding the target. First, let’s define the search space for the **Binary Search**.

Since the target value could exist anywhere in the array, the search space should encompass the entire array. Now, let’s figure out how to narrow the search space, which is an interesting challenge considering the array isn’t perfectly sorted.

Let’s work through this by exploring an example:

It’s possible to encounter a situation where both subarrays are sorted. In this case, it doesn’t matter which subarray we use to check where the target is.

One final thing to note is that the array might not contain the target value at all. In this case, once the **Binary Search** terminates and narrows down to a single value, we need to check if this value is the target.

If not, it indicates the target is not present in the array, and we return -1.

<h4 class="bbg-h4" style="color: var(--bbg-h4-color, var(--vscode-terminal-ansiYellow, var(--jp-warn-color1, #b45309))); font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; letter-spacing: 0; margin: 18px 0 8px;">Python Implementation</h4>

```python
from typing import List

def find_the_target_in_a_rotated_sorted_array(nums: List[int], target: int) -> int:
    left, right = 0, len(nums) - 1
    while left < right:
        mid = (left + right) // 2
        if nums[mid] == target:
            return mid
        # If the left subarray [left : mid] is sorted, check if the target falls in
        # this range. If it does, search the left subarray. Otherwise, search the
        # right.
        elif nums[left] <= nums[mid]:
            if nums[left] <= target < nums[mid]:
                right = mid - 1
            else:
                left = mid + 1
        # If the right subarray [mid : right] is sorted, check if the target falls
        # in this range. If it does, search the right subarray. Otherwise, search the
        # left.
        else:
            if nums[mid] < target <= nums[right]:
                left = mid + 1
            else:
                right = mid - 1
    # If the target is found in the array, return it’s index. Otherwise, return -1.
    return left if nums and nums[left] == target else -1
```
