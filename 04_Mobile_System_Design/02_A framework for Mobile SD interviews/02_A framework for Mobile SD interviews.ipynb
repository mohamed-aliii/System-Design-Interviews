<p>Having a clear framework for Mobile System Design (MSD) interviews is more important than many realize. Without a clear structure, the interview can feel disorganized and challenging for both you and the interviewer. Unlike coding interviews with defined solutions, system design challenges invite multiple valid approaches, making a methodical strategy essential.</p>
<p>Below, we briefly outline a 5-step framework to give you a high-level understanding. Detailed explanations and practical applications of each step are explored through case studies in later chapters.</p></section>

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-2-1-2ETKJDTJ.svg" alt="Figure 1: High-level steps of the MSDI framework" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;" /></p>
<div style="border-bottom:2px solid rgba(233,30,99,0.35);padding-bottom:8px;margin:28px auto 14px;max-width:52rem;width:96%;box-sizing:border-box"><span style="font-size:1.3em;font-weight:600;color:#e91e63">Step 1: Understand the problem and establish design scope</span></div>

<p>MSD interviews typically start with broad questions such as &quot;Design a News Feed app&quot; or &quot;Design a Pagination library.&quot; While these questions may seem straightforward, understanding exactly what you need to design is a crucial part of the interview.</p>
<p>Your first task is to clarify what&#x27;s in and out of scope by asking targeted questions to define the boundaries of your design. The interviewer&#x27;s answers will guide your technology choices and help you make informed trade-offs. This initial scoping sets the foundation for your solution, making each interview unique.</p>
<p>What kind of questions should you ask? Here&#x27;s a list of different types of questions to help you get started:</p>
<ul><li>
What are we building? These help uncover both functional and non-functional requirements, as well as what is out of scope for the interview. For example, you could ask:

What features, screens, or critical user journeys are we designing?
Are we pre-fetching data? Are we including user authentication in the design?

</li><li>
For whom are we building? These help understand the scale, usage patterns, and how performant the solution should be. For example, you could ask:

How many daily active users (DAU) does the product have? What&#x27;s the expected growth?
Are we building a minimum viable product (MVP) or a final product?
What&#x27;s our target market? A specific region or worldwide? What&#x27;s the typical usage scenario? At home or on the train?

</li><li>
What are the system&#x27;s constraints? These clarify the system boundaries and technology choices. For example, you could ask:

Which mobile platforms are we targeting (e.g., iOS, Android, or both)?
Are there any existing systems or APIs the mobile app must integrate with?

</li></ul>
<p>While gathering requirements, you may find that some answers reveal information about other aspects of the system. Feel free to make reasonable assumptions based on these responses, just be sure to state them clearly so your interviewer can correct any misunderstandings.</p>
<p>The goal isn&#x27;t to ask every possible question. Instead, focus your questions on what&#x27;s most relevant for the specific problem at hand, using your judgment to determine which details are most important to clarify.</p>
<p><strong>📌 Remember</strong>: Your interviewer&#x27;s answers to these questions will guide your technical decisions and trade-offs throughout the design process. Make sure to reference back to these requirements when explaining your choices later in the interview.</p>
<div style="border-bottom:2px solid rgba(233,30,99,0.35);padding-bottom:8px;margin:28px auto 14px;max-width:52rem;width:96%;box-sizing:border-box"><span style="font-size:1.3em;font-weight:600;color:#e91e63">Step 2: API design</span></div>

<p>The API design step is where you establish the contract between our client apps and external dependencies. This step is crucial because it helps align your understanding with the interviewer&#x27;s expectations and creates a solid foundation for the rest of your design.</p>
<p>When designing an app that needs to communicate with a backend, start by defining the appropriate communication protocols for your specific use cases and the endpoints involved. On the other hand, if you&#x27;re designing a mobile library, focus on outlining its public APIs, including how host apps will initialize, configure, and interact with your library.</p>
<p>As you define the API, outline the data models that will power these interactions. Whether you&#x27;re working with JSON payloads, Swift structs, Kotlin data classes, or even protocol buffers for efficiency, these models need to capture the data&#x27;s structure and relationships clearly.</p>
<div style="border-bottom:2px solid rgba(233,30,99,0.35);padding-bottom:8px;margin:28px auto 14px;max-width:52rem;width:96%;box-sizing:border-box"><span style="font-size:1.3em;font-weight:600;color:#e91e63">Step 3: High-level client architecture</span></div>

<p>After finishing the API design, it&#x27;s time to create a high-level client architecture diagram that shows how these components work together.</p>
<p>This high-level diagram sets the stage for the design and shows exactly which components you need to fulfill our requirements. Most importantly, it helps you validate that you have a complete end-to-end flow that meets all the functionality our system needs. As you create the diagram, you should also identify interesting points that warrant deeper discussion in the deep dive step later.</p>
<div style="border-bottom:2px solid rgba(233,30,99,0.35);padding-bottom:8px;margin:28px auto 14px;max-width:52rem;width:96%;box-sizing:border-box"><span style="font-size:1.3em;font-weight:600;color:#e91e63">Step 4: Design deep dive</span></div>

<p>In this step, work with your interviewer to identify bottlenecks and explore specific components in greater detail. Generally, you should aim for two to three strong deep-dive discussions, though this isn&#x27;t a strict rule and can vary depending on the situation.</p>
<p>You should continue refining your client architecture during this step. Make sure to update your diagrams to reflect any new decisions or components you introduce.</p>
<p>Pay attention to your interviewer&#x27;s reactions and questions. If they express interest in multiple topics, manage your time carefully to give each topic the attention it deserves while staying within the interview&#x27;s time constraints.</p>
<div style="border-bottom:2px solid rgba(233,30,99,0.35);padding-bottom:8px;margin:28px auto 14px;max-width:52rem;width:96%;box-sizing:border-box"><span style="font-size:1.3em;font-weight:600;color:#e91e63">Step 5: Wrap-up (optional)</span></div>

<p>If time allows, the final phase of your interview provides an opportunity to demonstrate your ability to think critically about your design and consider future improvements. Here&#x27;s how to wrap up effectively:</p>
<ul><li><strong>Summarize key decisions</strong>. If your discussion involved multiple trade-offs or complex decisions, consider providing a brief recap. This helps solidify the key points of your design and ensures the interviewer understands your reasoning, particularly after a detailed technical discussion.</li><li><strong>Showcase critical thinking</strong>. When asked about potential improvements to your design, resist the urge to defend it as perfect. Instead, thoughtfully discuss areas where your solution could evolve. This demonstrates self-awareness and analytical thinking, qualities that interviewers value highly.</li><li><strong>Address edge cases</strong>. Take this opportunity to discuss any failure points or error cases you haven&#x27;t covered earlier. This shows that you think holistically about system reliability and user experience, not just the happy path.</li><li><strong>Consider future scale</strong>. Think about how your system would handle growth, whether that&#x27;s supporting more users or accommodating a larger development team. This demonstrates that you design with the future in mind and understand both technical and organizational scaling challenges.</li></ul>
<div style="border-bottom:2px solid rgba(233,30,99,0.35);padding-bottom:8px;margin:28px auto 14px;max-width:52rem;width:96%;box-sizing:border-box"><span style="font-size:1.3em;font-weight:600;color:#e91e63">Time allocation for each step</span></div>

<p>Time management is key in MSD interviews. With only 45-60 minutes to work with and a lot of ground to cover, having a solid plan helps you stay on track. Here&#x27;s how to break down a typical 45-minute interview:</p>
<p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/image-2-1-X2ITWQUT.svg" alt="Diagram" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;" /></p>
<ol><li><strong>Understand the problem and establish design scope (5-10 minutes)</strong>: Take time to fully understand the problem and define its scope.</li><li><strong>API design (5-10 minutes)</strong>: Map out how your app will communicate with backend services and define the core data models.</li><li><strong>High-level client architecture (10-15 minutes)</strong>: Outline the main components of your system and how they work together.</li><li><strong>Design deep dive (15-20 minutes)</strong>: Focus on the most important technical challenges and implementation details.</li><li><strong>Wrap-up (0-5 minutes)</strong>: Summarize key decisions and discuss potential improvements or trade-offs.</li></ol>
<p>Remember that these are guidelines rather than strict rules. <strong>Be ready to adjust based on the specific problem and your interviewer&#x27;s interests</strong>. The key is maintaining enough flexibility to explore important areas while ensuring you complete the full design within the time limit.</p></section>
