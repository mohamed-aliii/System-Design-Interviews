<p>Whenever you watch a video on YouTube, it usually starts streaming immediately and you do not wait until the whole video is downloaded. Downloading means the whole video is copied to your device, while streaming means your device continuously receives video streams from remote source videos. When you watch streaming videos, your client loads a little bit of data at a time so you can watch videos immediately and continuously.</p>
<p>Before we discuss video streaming flow, let us look at an important concept: streaming protocol. This is a standardized way to control data transfer for video streaming. Popular streaming protocols are:</p>
<ul>
<li>MPEG–DASH. MPEG stands for “Moving Picture Experts Group” and DASH stands for "Dynamic Adaptive Streaming over HTTP".</li>
<li>Apple HLS. HLS stands for “HTTP Live Streaming”.</li>
<li>Microsoft Smooth Streaming.</li>
<li>Adobe HTTP Dynamic Streaming (HDS).</li>
</ul>
<p>You do not need to fully understand or even remember those streaming protocol names as they are low-level details that require specific domain knowledge. The important thing here is to understand that different streaming protocols support different video encodings and playback players. When we design a video streaming service, we have to choose the right streaming protocol to support our use cases. To learn more about streaming protocols, here is an excellent article [7].</p>
<p>Videos are streamed from CDN directly. The edge server closest to you will deliver the video. Thus, there is very little latency. <section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(255,23,68,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #ff1744;">Video transcoding</span>
</div>

<p>When you record a video, the device (usually a phone or camera) gives the video file a certain format. If you want the video to be played smoothly on other devices, the video must be encoded into compatible bitrates and formats. Bitrate is the rate at which bits are processed over time. A higher bitrate generally means higher video quality. High bitrate streams need more processing power and fast internet speed.</p>
<p>Video transcoding is important for the following reasons:</p>
<ul>
<li>Raw video consumes large amounts of storage space. An hour-long high definition video recorded at 60 frames per second can take up a few hundred GB of space.</li>
<li>Many devices and browsers only support certain types of video formats. Thus, it is important to encode a video to different formats for compatibility reasons.</li>
<li>To ensure users watch high-quality videos while maintaining smooth playback, it is a good idea to deliver higher resolution video to users who have high network bandwidth and lower resolution video to users who have low bandwidth.</li>
<li>Network conditions can change, especially on mobile devices. To ensure a video is played continuously, switching video quality automatically or manually based on network conditions is essential for smooth user experience.</li>
</ul>
<p>Many types of encoding formats are available; however, most of them contain two parts:</p>
<ul>
<li>Container: This is like a basket that contains the video file, audio, and metadata. You can tell the container format by the file extension, such as .avi, .mov, or .mp4.</li>
<li>Codecs: These are compression and decompression algorithms aim to reduce the video size while preserving the video quality. The most used video codecs are H.264, VP9, and HEVC.</li>
</ul>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(255,23,68,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #ff1744;">Directed acyclic graph (DAG) model</span>
</div>

<p>Transcoding a video is computationally expensive and time-consuming. Besides, different content creators may have different video processing requirements. For instance, some content creators require watermarks on top of their videos, some provide thumbnail images themselves, and some upload high definition videos, whereas others do not.</p>
<p>To support different video processing pipelines and maintain high parallelism, it is important to add some level of abstraction and let client programmers define what tasks to execute. For example, Facebook’s streaming video engine uses a directed acyclic graph (DAG) programming model, which defines tasks in stages so they can be executed sequentially or parallelly [8]. In our design, we adopt a similar DAG model to achieve flexibility and parallelism. Figure 8 represents a DAG for video transcoding.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-14-16-D65ZJKSR.svg" alt="Figure 8" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>In Figure 8, the original video is split into video, audio, and metadata. Here are some of the tasks that can be applied on a video file:</p>
<ul>
<li>Inspection: Make sure videos have good quality and are not malformed.</li>
<li>Video encodings: Videos are converted to support different resolutions, codec, bitrates, etc. Figure 9 shows an example of video encoded files.</li>
<li>Thumbnail. Thumbnails can either be uploaded by a user or automatically generated by the system.</li>
<li>Watermark: An image overlay on top of your video contains identifying information about your video.</li>
</ul>
</section>

  <span style="font-size: 1.3em; font-weight: 600; color: #ff1744;">Video transcoding architecture</span>

<p>The proposed video transcoding architecture that leverages the cloud services, is shown in Figure 10.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-14-18-24OH4P2W.svg" alt="Figure 10" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>The architecture has six main components: preprocessor, DAG scheduler, resource manager, task workers, temporary storage, and encoded video as the output. <section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(255,23,68,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Preprocessor</span>
</div>

<p>The preprocessor has 4 responsibilities:</p>
<ol>
<li>Video splitting. Video stream is split or further split into smaller Group of Pictures (GOP) alignment. GOP is a group/chunk of frames arranged in a specific order. Each chunk is an independently playable unit, usually a few seconds in length.</li>
<li>Some old mobile devices or browsers might not support video splitting. Preprocessor split videos by GOP alignment for old clients.</li>
<li>DAG generation. The processor generates DAG based on configuration files client programmers write. Figure 12 is a simplified DAG representation which has 2 nodes and 1 edge:</li>
</ol>
<p>This DAG representation is generated from the two configuration files below (Figure 13):</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-14-21-AZ4TQFF2.svg" alt="Figure 13" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><ol>
<li>Cache data. The preprocessor is a cache for segmented videos. For better reliability, the preprocessor stores GOPs and metadata in temporary storage. If video encoding fails, the system could use persisted data for retry operations.</li>
</ol>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(255,23,68,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">DAG scheduler</span>
</div>

<p>The DAG scheduler splits a DAG graph into stages of tasks and puts them in the task queue in the resource manager. Figure 15 shows an example of how the DAG scheduler works.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-14-24-LIPZPHHR.svg" alt="Figure 15" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>As shown in Figure 15, the original video is split into three stages: Stage 1: video, audio, and metadata. The video file is further split into two tasks in stage 2: video encoding and thumbnail. The audio file requires audio encoding as part of the stage 2 tasks.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(255,23,68,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Resource manager</span>
</div>

<p>The resource manager is responsible for managing the efficiency of resource allocation. It contains 3 queues and a task scheduler as shown in Figure 17.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-14-26-DQ2B7KIJ.svg" alt="Figure 17" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><ul>
<li>Task queue: It is a priority queue that contains tasks to be executed.</li>
<li>Worker queue: It is a priority queue that contains worker utilization info.</li>
<li>Running queue: It contains info about the currently running tasks and workers running the tasks.</li>
</ul>
<p>•Task scheduler: It picks the optimal task/worker, and instructs the chosen task worker to execute the job.</p>
<p>The resource manager works as follows:</p>
<ul>
<li>The task scheduler gets the highest priority task from the task queue.</li>
<li>The task scheduler gets the optimal task worker to run the task from the worker queue.</li>
<li>The task scheduler instructs the chosen task worker to run the task.</li>
<li>The task scheduler binds the task/worker info and puts it in the running queue.</li>
<li>The task scheduler removes the job from the running queue once the job is done.</li>
</ul>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">

  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Task workers</span>
</div>

<p>Task workers run the tasks which are defined in the DAG. Different task workers may run different tasks as shown in Figure 19.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(255,23,68,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Temporary storage</span>
</div>

<p>Multiple storage systems are used here. The choice of storage system depends on factors like data type, data size, access frequency, data life span, etc. For instance, metadata is frequently accessed by workers, and the data size is usually small. Thus, caching metadata in memory is a good idea. For video or audio data, we put them in blob storage. Data in temporary storage is freed up once the corresponding video processing is complete.</p>
</section>

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(255,23,68,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Speed optimization: parallelize video uploading</span>
</div>

<p>Uploading a video as a whole unit is inefficient. We can split a video into smaller chunks by GOP alignment as shown in Figure 22.</p>
<p>This allows fast resumable uploads when the previous upload failed. The job of splitting a video file by GOP can be implemented by the client to improve the upload speed as shown in Figure 23.</p>

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(255,23,68,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Speed optimization: place upload centers close to users</span>
</div>

<p>Another way to improve the upload speed is by setting up multiple upload centers across the globe (Figure 24). People in the United States can upload videos to the North America upload center, and people in China can upload videos to the Asian upload center. To achieve this, we use CDN as upload centers.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(255,23,68,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Speed optimization: parallelism everywhere</span>
</div>

<p>Achieving low latency requires serious efforts. Another optimization is to build a loosely coupled system and enable high parallelism.</p>
<p>Our design needs some modifications to achieve high parallelism. Let us zoom in to the flow of how a video is transferred from original storage to the CDN. The flow is shown in Figure 25, revealing that the output depends on the input of the previous step. This dependency makes parallelism difficult.</p>
<p>To make the system more loosely coupled, we introduced message queues as shown in Figure 26. Let us use an example to explain how message queues make the system more loosely coupled.</p>
<ul>
<li>Before the message queue is introduced, the encoding module must wait for the output of the download module.</li>
<li>After the message queue is introduced, the encoding module does not need to wait for the output of the download module anymore. If there are events in the message queue, the encoding module can execute those jobs in parallel.</li>
</ul>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(255,23,68,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Safety optimization: pre-signed upload URL</span>
</div>

<p>Safety is one of the most important aspects of any product. To ensure only authorized users upload videos to the right location, we introduce pre-signed URLs as shown in Figure 27.</p>
<p>The upload flow is updated as follows:</p>
<ol>
<li>The client makes a HTTP request to API servers to fetch the pre-signed URL, which gives the access permission to the object identified in the URL. The term pre-signed URL is used by uploading files to Amazon S3. Other cloud service providers might use a different name. For instance, Microsoft Azure blob storage supports the same feature, but call it “Shared Access Signature” [10].</li>
<li>API servers respond with a pre-signed URL.</li>
<li>Once the client receives the response, it uploads the video using the pre-signed URL.</li>
</ol>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(255,23,68,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Safety optimization: protect your videos</span>
</div>

<p>Many content makers are reluctant to post videos online because they fear their original videos will be stolen. To protect copyrighted videos, we can adopt one of the following three safety options:</p>
<ul>
<li>Digital rights management (DRM) systems: Three major DRM systems are Apple FairPlay, Google Widevine, and Microsoft PlayReady.</li>
<li>AES encryption: You can encrypt a video and configure an authorization policy. The encrypted video will be decrypted upon playback. This ensures that only authorized users can watch an encrypted video.</li>
<li>Visual watermarking: This is an image overlay on top of your video that contains identifying information for your video. It can be your company logo or company name.</li>
</ul>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(255,23,68,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Cost-saving optimization</span>
</div>

<p>CDN is a crucial component of our system. It ensures fast video delivery on a global scale. However, from the back of the envelope calculation, we know CDN is expensive, especially when the data size is large. How can we reduce the cost?</p>
<p>Previous research shows that YouTube video streams follow long-tail distribution [11] [12]. It means a few popular videos are accessed frequently but many others have few or no viewers. Based on this observation, we implement a few optimizations:</p>
<ol>
<li>Only serve the most popular videos from CDN and other videos from our high capacity storage video servers (Figure 28).</li>
<li>For less popular content, we may not need to store many encoded video versions. Short videos can be encoded on-demand.</li>
<li>Some videos are popular only in certain regions. There is no need to distribute these videos to other regions.</li>
<li>Build your own CDN like Netflix and partner with Internet Service Providers (ISPs). Building your CDN is a giant project; however, this could make sense for large streaming companies. An ISP can be Comcast, AT&T, Verizon, or other internet providers. ISPs are located all around the world and are close to users. By partnering with ISPs, you can improve the viewing experience and reduce the bandwidth charges.</li>
</ol>
<p>All those optimizations are based on content popularity, user access pattern, video size, etc. It is important to analyze historical viewing patterns before doing any optimization. Here are some of the interesting articles on this topic: [12] [13].</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(255,23,68,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #ff1744;">Error handling</span>
</div>

<p>For a large-scale system, system errors are unavoidable. To build a highly fault-tolerant system, we must handle errors gracefully and recover from them fast. Two types of errors exist:</p>
<ul>
<li>Recoverable error. For recoverable errors such as video segment fails to transcode, the general idea is to retry the operation a few times. If the task continues to fail and the system believes it is not recoverable, it returns a proper error code to the client.</li>
<li>Non-recoverable error. For non-recoverable errors such as malformed video format, the system stops the running tasks associated with the video and returns the proper error code to the client.</li>
</ul>
<p>Typical errors for each system component are covered by the following playbook:</p>
<ul>
<li>Upload error: retry a few times.</li>
<li>Split video error: if older versions of clients cannot split videos by GOP alignment, the entire video is passed to the server. The job of splitting videos is done on the server-side.</li>
<li>Transcoding error: retry.</li>
<li>Preprocessor error: regenerate DAG diagram.</li>
<li>DAG scheduler error: reschedule a task.</li>
<li>Resource manager queue down: use a replica.</li>
<li>Task worker down: retry the task on a new worker.</li>
<li>API server down: API servers are stateless so requests will be directed to a different API server.</li>
<li>Metadata cache server down: data is replicated multiple times. If one node goes down, you can still access other nodes to fetch data. We can bring up a new cache server to replace the dead one.</li>
<li>Metadata DB server down:</li>
<li>Master is down. If the master is down, promote one of the slaves to act as the new master.</li>
<li>Slave is down. If a slave goes down, you can use another slave for reads and bring up another database server to replace the dead one.</li>
</ul>
</section>

<p>In this chapter, we presented the architecture design for video streaming services like YouTube. If there is extra time at the end of the interview, here are a few additional points:</p>
<ul>
<li>Scale the API tier: Because API servers are stateless, it is easy to scale API tier horizontally.</li>
<li>Scale the database: You can talk about database replication and sharding.</li>
<li>Live streaming: It refers to the process of how a video is recorded and broadcasted in real time. Although our system is not designed specifically for live streaming, live streaming and non-live streaming have some similarities: both require uploading, encoding, and streaming. The notable differences are:</li>
<li>Live streaming has a higher latency requirement, so it might need a different streaming protocol.</li>
<li>Live streaming has a lower requirement for parallelism because small chunks of data are already processed in real-time.</li>
<li>Live streaming requires different sets of error handling. Any error handling that takes too much time is not acceptable.</li>
<li>Video takedowns: Videos that violate copyrights, pornography, or other illegal acts shall be removed. Some can be discovered by the system during the upload process, while others might be discovered through user flagging.</li>
</ul>
