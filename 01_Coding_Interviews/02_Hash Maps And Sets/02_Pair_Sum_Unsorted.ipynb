<h2 class="bbg-h2" style="color: #e5edf7; border-left: 5px solid #39ff88; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Problem</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

Given an array of integers, return the indexes of any two numbers that add up to a target. The order of the indexes in the result doesn't matter. If no pair is found, return an empty array.

</div>

<h3 class="bbg-h3" style="color: #38bdf8; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 22px 0 10px;">Example:</h3>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

**Input:** `nums = [-1, 3, 4, 2], target = 3`

**Output:** `[0, 2]`

**Explanation:**

- nums[0] + nums[2] = -1 + 4 = 3



</div>

<h3 class="bbg-h3" style="color: #38bdf8; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 22px 0 10px;">Constraints:</h3>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

- The same index cannot be used twice in the result.

</div>

<h2 class="bbg-h2" style="color: #e5edf7; border-left: 5px solid #39ff88; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Intuition</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

A **Brute Force** approach is to iterate through every possible pair in the array to see if their sum is equal to the target. This is the same as the **Brute Force** solution described in the Pair Sum - Sorted problem, which has a **Time Complexity** of $\mathcal{O}(n^2)$, where n is the length of the array.

We could also sort the array and then perform the two-pointer algorithm used in Pair Sum - Sorted, which would take $\mathcal{O}(n\log (n))$ time due to sorting. Let’s see if we can find an even faster solution.

Complement We're asked to find a pair ( x, y) such that x + y == target. In this equation, there are two unknowns: x and y.

An important observation is that if we know one of these numbers, we can easily calculate what the other number should be. For each number x in nums, we need to find another number y such that x + y = target, or in other words, y = target - x.

We can call this number the complement of x. Keep in mind, we need to return the indexes of the pair of numbers, not the pair itself.

So, we’ll need a way to find a number's complement as well as its index. One way we could do this is to loop through the array to find each number’s complement and corresponding index.

But this takes $\mathcal{O}(n^2)$ time since we’d need to do a linear traversal to search for each number’s complement. Instead, we’d like an efficient way to determine the index of any number in the array without needing to search the array.

Is there a data structure that can help with this? **Hash Map** A **Hash Map** works great because we can store and look up values in $\mathcal{O}(1)$ time. Each number and its index can be stored in the **Hash Map** as key-value pairs: This allows us to retrieve the index of any number’s complement efficiently.

Notice that duplicate numbers don’t need to be considered here since only one valid pair needs to be found. The most intuitive way to incorporate a **Hash Map** is to:

- In the first pass, populate the **Hash Map** with each number and its corresponding index.

- In the second pass, scan through the array to check if each number's complement exists in the **Hash Map**. If it does, we can return the indexes of that number and its complement.

Below is the code snippet for this two-pass approach:

</div>

<h4 class="bbg-h4" style="color: #f59e0b; font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; letter-spacing: 0; margin: 18px 0 8px;">Python Snippet</h4>

```python
from typing import List

def pair_sum_unsorted_two_pass(nums: List[int], target: int) -> List[int]:
    num_map = {}
    # First pass: Populate the hash map with each number and its index.
    for i, num in enumerate(nums):
        num_map[num] = i
    # Second pass: Check for each number's complement in the hash map.
    for i, num in enumerate(nums):
        complement = target - num
        if complement in num_map and num_map[complement] != i:
            return [i, num_map[complement]]
    return []
```

<h2 class="bbg-h2" style="color: #e5edf7; border-left: 5px solid #39ff88; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Implementation</h2>

<h4 class="bbg-h4" style="color: #f59e0b; font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; letter-spacing: 0; margin: 18px 0 8px;">Python Implementation</h4>

```python
from typing import List

def pair_sum_unsorted(nums: List[int], target: int) -> List[int]:
    hashmap = {}
    for i, x in enumerate(nums):
        if target - x in hashmap:
            return [hashmap[target - x], i]
        hashmap[x] = i
    return []
```

<h2 class="bbg-h2" style="color: #e5edf7; border-left: 5px solid #39ff88; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Complexity Analysis</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

**Time Complexity**: The **Time Complexity** of pair_sum_unsorted is $\mathcal{O}(n)$ because we iterate through each element in the nums array once and perform constant-time **Hash Map** operations during each iteration. **Space Complexity**: The **Space Complexity** is $\mathcal{O}(n)$ since the **Hash Map** can grow up to n in size.

</div>
