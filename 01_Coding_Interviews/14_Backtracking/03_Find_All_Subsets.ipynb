Return all possible subsets of a given set of unique integers. Each subset can be ordered in any way, and the subsets can be returned in any order.

</div>

<h3 class="bbg-h3" style="color: #38bdf8; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 22px 0 10px;">Example:</h3>

<h2 class="bbg-h2" style="color: #e5edf7; border-left: 5px solid #39ff88; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Intuition</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

For example, from the array [4, 5, 6], the subset [4, 6] is created by including 4, excluding 5, and including 6.

As shown, the final level of the **Tree** (i.e., when i == n, where n denotes the length of the input array) contains all the subsets of the input array. We can add each of these subsets to our output. To get to these subsets, we need to traverse the **Tree**, and **Backtracking** is great for this.

<h4 class="bbg-h4" style="color: #f59e0b; font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; letter-spacing: 0; margin: 18px 0 8px;">Python Implementation</h4>

```python
from typing import List

def find_all_subsets(nums: List[int]) -> List[List[int]]:
    res = []
    backtrack(0, [], nums, res)
    return res

def backtrack(i: int, curr_subset: List[int], nums: List[int], res: List[List[int]]) -> None:
    # Base case: if all elements have been considered, add the current subset to the
    # output.
    if i == len(nums):
        res.append(curr_subset[:])
        return
    # Include the current element and recursively explore all paths that branch from
    # this subset.
    curr_subset.append(nums[i])
    backtrack(i + 1, curr_subset, nums, res)
    # Exclude the current element and recursively explore all paths that branch from
    # this subset.
    curr_subset.pop()
    backtrack(i + 1, curr_subset, nums, res)
```

<h2 class="bbg-h2" style="color: #e5edf7; border-left: 5px solid #39ff88; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Complexity Analysis</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

**Time Complexity**: The **Time Complexity** of find_all_subsets is $\mathcal{O}(n^2\cdot n)$. This is because the state space **Tree** has a depth of n and a branching factor of 2 since there are two decisions we make at each state.

For each of the 2n subsets created, we make a copy of them and add the copy to the output, which takes $\mathcal{O}(n)$ time. This results in a total **Time Complexity** of $\mathcal{O}(n^2\cdot n)$. **Space Complexity**: The **Space Complexity** is $\mathcal{O}(n)$ because the maximum depth of the **Recursion** **Tree** is n.

The algorithm also maintains the curr_subset data structure, which also contributes $\mathcal{O}(n)$ space. Note, the res array does not contribute to **Space Complexity**.
