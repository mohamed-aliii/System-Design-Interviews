<p>To achieve horizontal scaling, it is important to distribute requests/data efficiently and evenly across servers. Consistent hashing is a commonly used technique to achieve this goal. But first, let us take an in-depth look at the problem.</p>
</section>


<div style="border-left: 5px solid #d500f9; padding: 12px 0 12px 20px; margin: 32px auto 18px; max-width: 52rem; width: 96%; box-sizing: border-box; background: linear-gradient(90deg, rgba(213,0,249,0.08) 0%, transparent 60%); border-radius: 4px;">
  <span style="font-size: 1.6em; font-weight: 700; color: #e5edf7; line-height: 1.25;">The rehashing problem</span>
</div>



<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<p>If you have <em>n</em> cache servers, a common way to balance the load is to use the following hash method:</p>
<p><em>serverIndex = hash(key) % N</em>, where <em>N</em> is the size of the server pool.</p>
<p>Let us use an example to illustrate how it works. As shown in Table 1, we have 4 servers and 8 string keys with their hashes.</p>
<table><thead><tr><th><strong>key</strong></th><th><strong>hash</strong></th><th><strong>hash % 4</strong></th></tr></thead><tbody><tr><td>key0</td><td>18358617</td><td>1</td></tr><tr><td>key1</td><td>26143584</td><td>0</td></tr><tr><td>key2</td><td>18131146</td><td>2</td></tr><tr><td>key3</td><td>35863496</td><td>0</td></tr><tr><td>key4</td><td>34085809</td><td>1</td></tr><tr><td>key5</td><td>27581703</td><td>3</td></tr><tr><td>key6</td><td>38164978</td><td>2</td></tr><tr><td>key7</td><td>22530351</td><td>3</td></tr></tbody></table>
<p>Table 1</p>
<p>To fetch the server where a key is stored, we perform the modular operation <em>f(key) % 4</em>. For instance, <em>hash(key0) % 4 = 1</em> means a client must contact server 1 to fetch the cached data. Figure 1 shows the distribution of keys based on Table 1.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-1-JYB5LSFH.svg" alt="Figure 1" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>This approach works well when the size of the server pool is fixed, and the data distribution is even. However, problems arise when new servers are added, or existing servers are removed. For example, if server 1 goes offline, the size of the server pool becomes 3. Using the same hash function, we get the same hash value for a key. But applying modular operation gives us different server indexes because the number of servers is reduced by 1. We get the results as shown in Table 2 by applying <em>hash % 3</em>:</p>

<p>Table 2</p>
<p>Figure 2 shows the new distribution of keys based on Table 2.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-2-25DSOGUM.svg" alt="Figure 2" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>As shown in Figure 2, most keys are redistributed, not just the ones originally stored in the offline server (server 1). This means that when server 1 goes offline, most cache clients will connect to the wrong servers to fetch data. This causes a storm of cache misses. Consistent hashing is an effective technique to mitigate this problem.</p>
</section>


<div style="border-left: 5px solid #d500f9; padding: 12px 0 12px 20px; margin: 32px auto 18px; max-width: 52rem; width: 96%; box-sizing: border-box; background: linear-gradient(90deg, rgba(213,0,249,0.08) 0%, transparent 60%); border-radius: 4px;">
  <span style="font-size: 1.6em; font-weight: 700; color: #e5edf7; line-height: 1.25;">Consistent hashing</span>
</div>



<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<p>Quoted from Wikipedia: "Consistent hashing is a special kind of hashing such that when a hash table is re-sized and consistent hashing is used, only k/n keys need to be remapped on average, where k is the number of keys, and n is the number of slots. In contrast, in most traditional hash tables, a change in the number of array slots causes nearly all keys to be remapped [1]”.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(213,0,249,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #d500f9;">Hash space and hash ring</span>
</div>

<p>Now we understand the definition of consistent hashing, let us find out how it works. Assume SHA-1 is used as the hash function f, and the output range of the hash function is: <em>x0, x1, x2, x3, …, xn</em>. In cryptography, SHA-1’s hash space goes from 0 to 2^160 - 1. That means <em>x0</em> corresponds to 0, <em>xn</em> corresponds to 2^160 – 1, and all the other hash values in the middle fall between 0 and 2^160 - 1. Figure 3 shows the hash space.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-3-REYWWCBF.svg" alt="Figure 3" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>By collecting both ends, we get a hash ring as shown in Figure 4:</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-4-IYSYQWO7.svg" alt="Figure 4" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p>
</section>

<div style="border-bottom: 2px solid rgba(213,0,249,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #d500f9;">Hash servers</span>
</div>

<p>Using the same hash function f, we map servers based on server IP or name onto the ring. Figure 5 shows that 4 servers are mapped on the hash ring.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-5-4LOAYCHE.svg" alt="Figure 5" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(213,0,249,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #d500f9;">Hash keys</span>
</div>

<p>One thing worth mentioning is that hash function used here is different from the one in “the rehashing problem,” and there is no modular operation. As shown in Figure 6, 4 cache keys (key0, key1, key2, and key3) are hashed onto the hash ring</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-6-SDSBYUI6.svg" alt="Figure 6" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(213,0,249,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #d500f9;">Server lookup</span>
</div>

<p>To determine which server a key is stored on, we go clockwise from the key position on the ring until a server is found. Figure 7 explains this process. Going clockwise, <em>key0</em> is stored on <em>server 0</em>; <em>key1</em> is stored on <em>server 1</em>; <em>key2</em> is stored on <em>server 2</em> and <em>key3</em> is stored on <em>server 3</em>.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-7-OIUSEHJP.svg" alt="Figure 7" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(213,0,249,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #d500f9;">Add a server</span>
</div>

<p>Using the logic described above, adding a new server will only require redistribution of a fraction of keys.</p>
<p>In Figure 8, after a new <em>server 4</em> is added, only <em>key0</em> needs to be redistributed. <em>k1, k2,</em> and <em>k3</em> remain on the same servers. Let us take a close look at the logic. Before <em>server 4</em> is added, <em>key0</em> is stored on <em>server 0</em>. Now, <em>key0</em> will be stored on <em>server 4</em> because <em>server 4</em> is the first server it encounters by going clockwise from <em>key0</em>’s position on the ring. The other keys are not redistributed based on consistent hashing algorithm.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-8-OBPLVFJF.svg" alt="Figure 8" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(213,0,249,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #d500f9;">Remove a server</span>
</div>

<p>When a server is removed, only a small fraction of keys require redistribution with consistent hashing. In Figure 9, when <em>server 1</em> is removed, only <em>key1</em> must be remapped to <em>server 2</em>. The rest of the keys are unaffected.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-9-NK63F2OI.svg" alt="Figure 9" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(213,0,249,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #d500f9;">Two issues in the basic approach</span>
</div>

<p>The consistent hashing algorithm was introduced by Karger et al. at MIT [1]. The basic steps are:</p>
<ul>
<li>Map servers and keys on to the ring using a uniformly distributed hash function.</li>
<li>To find out which server a key is mapped to, go clockwise from the key position until the first server on the ring is found.</li>
</ul>
<p>Two problems are identified with this approach. First, it is impossible to keep the same size of partitions on the ring for all servers considering a server can be added or removed. A partition is the hash space between adjacent servers. It is possible that the size of the partitions on the ring assigned to each server is very small or fairly large. In Figure 10, if <em>s1</em> is removed, <em>s2’s</em> partition (highlighted with the bidirectional arrows) is twice as large as <em>s0</em> and <em>s3’s</em> partition.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-10-FCDAIIM2.svg" alt="Figure 10" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>Second, it is possible to have a non-uniform key distribution on the ring. For instance, if servers are mapped to positions listed in Figure 11, most of the keys are stored on <em>server 2</em>. However, <em>server 1</em> and <em>server 3</em> have no data.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-11-BZXRZPW7.svg" alt="Figure 11" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>A technique called virtual nodes or replicas is used to solve these problems.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(213,0,249,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #d500f9;">Virtual nodes</span>
</div>

<p>A virtual node refers to the real node, and each server is represented by multiple virtual nodes on the ring. In Figure 12, both <em>server 0</em> and <em>server 1</em> have 3 virtual nodes. The 3 is arbitrarily chosen; and in real-world systems, the number of virtual nodes is much larger. Instead of using <em>s0</em>, we have <em>s0_0, s0_1</em>, and <em>s0_2</em> to represent <em>server 0</em> on the ring. Similarly, <em>s1_0, s1_1</em>, and <em>s1_2</em> represent server 1 on the ring. With virtual nodes, each server is responsible for multiple partitions. Partitions (edges) with label <em>s0</em> are managed by server 0. On the other hand, partitions with label <em>s1</em> are managed by <em>server 1</em>.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-12-JS3PL6KQ.svg" alt="Figure 12" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>To find which server a key is stored on, we go clockwise from the key’s location and find the first virtual node encountered on the ring. In Figure 13, to find out which server <em>k0</em> is stored on, we go clockwise from <em>k0</em>’s location and find virtual node <em>s1_1</em>, which refers to <em>server 1</em>.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-13-EH76K4PV.svg" alt="Figure 13" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>As the number of virtual nodes increases, the distribution of keys becomes more balanced. This is because the standard deviation gets smaller with more virtual nodes, leading to balanced data distribution. Standard deviation measures how data are spread out. The outcome of an experiment carried out by online research [2] shows that with one or two hundred virtual nodes, the standard deviation is between 5% (200 virtual nodes) and 10% (100 virtual nodes) of the mean. The standard deviation will be smaller when we increase the number of virtual nodes. However, more spaces are needed to store data about virtual nodes. This is a tradeoff, and we can tune the number of virtual nodes to fit our system requirements.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(213,0,249,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #d500f9;">Find affected keys</span>
</div>

<p>When a server is added or removed, a fraction of data needs to be redistributed. How can we find the affected range to redistribute the keys?</p>
<p>In Figure 14, <em>server 4</em> is added onto the ring. The affected range starts from <em>s4</em> (newly added node) and moves anticlockwise around the ring until a server is found (<em>s3</em>). Thus, keys located between <em>s3</em> and <em>s4</em> need to be redistributed to <em>s4</em>.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-14-HZNKA3J3.svg" alt="Figure 14" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>When a server (<em>s1</em>) is removed as shown in Figure 15, the affected range starts from <em>s1</em> (removed node) and moves anticlockwise around the ring until a server is found (<em>s0</em>). Thus, keys located between <em>s0</em> and <em>s1</em> must be redistributed to <em>s2</em>.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-15-PNWYOYZO.svg" alt="Figure 15" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p>
</section>

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<p>In this chapter, we had an in-depth discussion about consistent hashing, including why it is needed and how it works. The benefits of consistent hashing include:</p>
<ul>
<li>Minimized keys are redistributed when servers are added or removed.</li>
<li>It is easy to scale horizontally because data are more evenly distributed.</li>
<li>Mitigate hotspot key problem. Excessive access to a specific shard could cause server overload. Imagine data for Katy Perry, Justin Bieber, and Lady Gaga all end up on the same shard. Consistent hashing helps to mitigate the problem by distributing the data more evenly.</li>
</ul>
<p>Consistent hashing is widely used in real-world systems, including some notable ones:</p>
<ul>
<li>Partitioning component of Amazon’s Dynamo database [3]</li>
<li>Data partitioning across the cluster in Apache Cassandra [4]</li>
<li>Discord chat application [5]</li>
<li>Akamai content delivery network [6]</li>
<li>Maglev network load balancer [7]</li>
</ul>
