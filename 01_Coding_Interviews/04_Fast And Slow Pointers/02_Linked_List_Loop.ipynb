<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

A straightforward approach is to iterate through the **Linked List** while keeping track of the nodes that were already visited in a hash set. Encountering a previously-visited node during the traversal indicates the presence of a cycle. Below is the code snippet for this approach:

</div>

<h4 class="bbg-h4" style="color: var(--bbg-h4-color, var(--vscode-terminal-ansiYellow, var(--jp-warn-color1, #b45309))); font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; letter-spacing: 0; margin: 18px 0 8px;">Naive Python</h4>

```python
from ds import ListNode

def linked_list_loop_naive(head: ListNode) -> bool:
    visited = set()
    curr = head
    while curr:
        # Cycle detected if the current node has already been visited.
        if curr in visited:
            return True
        visited.add(curr)
        curr = curr.next
    return False
```

This solution takes $\mathcal{O}(n)$ time, where n is the number of nodes in the **Linked List**, since each node is visited once. However, this comes at the cost of $\mathcal{O}(n)$ extra space due to the hash set. Is there a way to achieve a linear **Time Complexity** while using constant space? Imagine a race track represented as a circular **Linked List** (i.e., a **Linked List** with a perfect cycle) where two runners start at the same node.

If both runners move at the same speed, they will always be together at each node. However, consider what happens when one runner (the slow runner) moves one step at a time, while the other runner (the fast runner) moves two steps at a time.

In this scenario, the fast runner will overtake the slow runner at some point since the track is cyclic. But how can we use this information to detect a cycle?

In a **Linked List**, detecting whether the fast runner has overtaken the slow runner is difficult due to the lack of positional indicators (like indexes in an array). A better way to find a cycle is to see if the fast runner reunites with the slow runner by both landing on the same node at some point.

This would be a clear sign the **Linked List** has a cycle. The question now is, will the fast runner reunite with the slow runner, or is there a chance for the fast runner to consistently bypass the slow runner without ever converging on the same node?

To answer this, let’s start by looking at some examples. Perfect cycle First, let’s check whether the two runners, represented as a slow pointer and a fast pointer, will reunite in a **Linked List** that forms a perfect cycle.

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

This algorithm is formally known as ‘Floyd's Cycle Detection’ algorithm [1].

</div>

<h4 class="bbg-h4" style="color: var(--bbg-h4-color, var(--vscode-terminal-ansiYellow, var(--jp-warn-color1, #b45309))); font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; letter-spacing: 0; margin: 18px 0 8px;">Python Implementation</h4>

```python
from ds import ListNode

def linked_list_loop(head: ListNode) -> bool:
    slow = fast = head
    # Check both 'fast' and 'fast.next' to avoid null pointer exceptions when we
    # perform 'fast.next' and 'fast.next.next'.
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if fast == slow:
            return True
    return False
```

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

**Time Complexity**: The **Time Complexity** of linked_list_loop is $\mathcal{O}(n)$ because the fast pointer will meet the slow pointer in a linear number of steps, as described. **Space Complexity**: The **Space Complexity** is $\mathcal{O}(1)$.

</div>
