<ul>
<li>Multi-master replication</li>
<li>Universally unique identifier (UUID)</li>
<li>Ticket server</li>
<li>Twitter snowflake approach</li>
</ul>
<p>Let us look at each of them, how they work, and the pros/cons of each option.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(255,61,0,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #ff3d00;">Multi-master replication</span>
</div>

<p>As shown in Figure 2, the first approach is multi-master replication.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-7-2-4ZOIVNTR.svg" alt="Figure 2" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>This approach uses the databases’ <em>auto_increment</em> feature. Instead of increasing the next ID by 1, we increase it by <em>k,</em> where <em>k</em> is the number of database servers in use. As illustrated in Figure 2, next ID to be generated is equal to the previous ID in the same server plus 2. This solves some scalability issues because IDs can scale with the number of database servers. However, this strategy has some major drawbacks:</p>
<ul>
<li>Hard to scale with multiple data centers</li>
<li>IDs do not go up with time across multiple servers.</li>
<li>It does not scale well when a server is added or removed.</li>
</ul>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(255,61,0,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #ff3d00;">UUID</span>
</div>

<p>A UUID is another easy way to obtain unique IDs. UUID is a 128-bit number used to identify information in computer systems. UUID has a very low probability of getting collusion. Quoted from Wikipedia, “after generating 1 billion UUIDs every second for approximately 100 years would the probability of creating a single duplicate reach 50%” [1].</p>
<p>Here is an example of UUID: <em>09c93e62-50b4-468d-bf8a-c07e1040bfb2</em>. UUIDs can be generated independently without coordination between servers. Figure 3 presents the UUIDs design.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-7-3-Y3ZNPIH3.svg" alt="Figure 3" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>In this design, each web server contains an ID generator, and a web server is responsible for generating IDs independently.</p>
<p>Pros:</p>
<ul>
<li>Generating UUID is simple. No coordination between servers is needed so there will not be any synchronization issues.</li>
<li>The system is easy to scale because each web server is responsible for generating IDs they consume. ID generator can easily scale with web servers.</li>
</ul>
<p>Cons:</p>
<ul>
<li>IDs are 128 bits long, but our requirement is 64 bits.</li>
<li>IDs do not go up with time.</li>
<li>IDs could be non-numeric.</li>
</ul>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(255,61,0,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #ff3d00;">Ticket Server</span>
</div>

<p>Ticket servers are another interesting way to generate unique IDs. Flicker developed ticket servers to generate distributed primary keys [2]. It is worth mentioning how the system works.</p>
<p>The idea is to use a centralized <em>auto_increment</em> feature in a single database server (Ticket Server). To learn more about this, refer to flicker’s engineering blog article [2].</p>
<p>Pros:</p>
<ul>
<li>Numeric IDs.</li>
<li>It is easy to implement, and it works for small to medium-scale applications.</li>
</ul>
<p>Cons:</p>
<ul>
<li>Single point of failure. Single ticket server means if the ticket server goes down, all systems that depend on it will face issues. To avoid a single point of failure, we can set up multiple ticket servers. However, this will introduce new challenges such as data synchronization.</li>
</ul>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(255,61,0,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #ff3d00;">Twitter snowflake approach</span>
</div>

<p>Approaches mentioned above give us some ideas about how different ID generation systems work. However, none of them meet our specific requirements; thus, we need another approach. Twitter’s unique ID generation system called “snowflake” [3] is inspiring and can satisfy our requirements.</p>
<p>Divide and conquer is our friend. Instead of generating an ID directly, we divide an ID into different sections. Figure 5 shows the layout of a 64-bit ID.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-7-5-R4B5RGNK.svg" alt="Figure 5" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>Each section is explained below.</p>
<ul>
<li>Sign bit: 1 bit. It will always be 0. This is reserved for future uses. It can potentially be used to distinguish between signed and unsigned numbers.</li>
<li>Timestamp: 41 bits. Milliseconds since the epoch or custom epoch. We use Twitter snowflake default epoch 1288834974657, equivalent to Nov 04, 2010, 01:42:54 UTC.</li>
<li>Datacenter ID: 5 bits, which gives us 2 ^ 5 = 32 datacenters.</li>
<li>Machine ID: 5 bits, which gives us 2 ^ 5 = 32 machines per datacenter.</li>
<li>Sequence number: 12 bits. For every ID generated on that machine/process, the sequence number is incremented by 1. The number is reset to 0 every millisecond.</li>
</ul>
</section>


<div style="border-left: 5px solid #ff3d00; padding: 12px 0 12px 20px; margin: 32px auto 18px; max-width: 52rem; width: 96%; box-sizing: border-box; background: linear-gradient(90deg, rgba(255,61,0,0.08) 0%, transparent 60%); border-radius: 4px;">
  <span style="font-size: 1.6em; font-weight: 700; color: #e5edf7; line-height: 1.25;">Step 3 - Design deep dive</span>
</div>



<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<p>In the high-level design, we discussed various options to design a unique ID generator in distributed systems. We settle on an approach that is based on the Twitter snowflake ID generator. Let us dive deep into the design. To refresh our memory, the design diagram is relisted below.</p>
<p>Datacenter IDs and machine IDs are chosen at the startup time, generally fixed once the system is up running. Any changes in datacenter IDs and machine IDs require careful review since an accidental change in those values can lead to ID conflicts. Timestamp and sequence numbers are generated when the ID generator is running.</p>
<p>Timestamp</p>
<p>The most important 41 bits make up the timestamp section. As timestamps grow with time, IDs are sortable by time. Figure 7 shows an example of how binary representation is converted to UTC. You can also convert UTC back to binary representation using a similar method.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-7-7-PNNDQIF5.svg" alt="Figure 7" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>The maximum timestamp that can be represented in 41 bits is</p>
<p><em>2 ^ 41 - 1 = 2199023255551</em> milliseconds (ms), which gives us: ~ 69 years = <em>2199023255551 ms / 1000 / 365 days / 24 hours/ 3600 seconds</em>. This means the ID generator will work for 69 years and having a custom epoch time close to today’s date delays the overflow time. After 69 years, we will need a new epoch time or adopt other techniques to migrate IDs.</p>
<p>Sequence number</p>
<p>Sequence number is 12 bits, which give us 2 ^ 12 = 4096 combinations. This field is 0 unless more than one ID is generated in a millisecond on the same server. In theory, a machine can support a maximum of 4096 new IDs per millisecond.</p>
</section>


<div style="border-left: 5px solid #ff3d00; padding: 12px 0 12px 20px; margin: 32px auto 18px; max-width: 52rem; width: 96%; box-sizing: border-box; background: linear-gradient(90deg, rgba(255,61,0,0.08) 0%, transparent 60%); border-radius: 4px;">
  <span style="font-size: 1.6em; font-weight: 700; color: #e5edf7; line-height: 1.25;">Step 4 - Wrap up</span>
</div>



<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<p>In this chapter, we discussed different approaches to design a unique ID generator: multi-master replication, UUID, ticket server, and Twitter snowflake-like unique ID generator. We settle on snowflake as it supports all our use cases and is scalable in a distributed environment.</p>
<p>If there is extra time at the end of the interview, here are a few additional talking points:</p>
<ul>
<li>Clock synchronization. In our design, we assume ID generation servers have the same clock. This assumption might not be true when a server is running on multiple cores. The same challenge exists in multi-machine scenarios. Solutions to clock synchronization are out of the scope of this course; however, it is important to understand the problem exists. Network Time Protocol is the most popular solution to this problem. For interested readers, refer to the reference material [4].</li>
<li>Section length tuning. For example, fewer sequence numbers but more timestamp bits are effective for low concurrency and long-term applications.</li>
<li>High availability. Since an ID generator is a mission-critical system, it must be highly available.</li>
</ul>
<p>Congratulations on getting this far! Now give yourself a pat on the back. Good job!</p>
