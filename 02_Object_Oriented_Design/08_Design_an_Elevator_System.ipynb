In this chapter, we will explore the object-oriented design of an Elevator System. Compared to some of the other popular interview problems, this one places a stronger emphasis on modeling behavior, rather than on modeling data. Our approach will focus on designing key components such as how to represent real-world elevators, the elevator’s state, incoming hallway call requests, and the algorithm that determines the elevator’s movement.

<div style="border-bottom: 2px solid rgba(72,149,239,0.3); padding-bottom: 8px; margin: 28px 0 14px 0;">
  <span style="font-size: 1.3em; font-weight: 600; color: #4895ef;">Requirements</span>
</div>


Functional requirements:


* The system manages multiple elevator cars, all of which serve the same set of floors.

* On each floor, there are “up” and “down” buttons that users press to call an elevator car.

* Each elevator car should display its current floor and state (moving up, down, or idle).

* Each elevator car has an internal control panel with buttons for every floor.

* If a user presses a floor button in a direction opposite to the elevator’s current movement, the request should be ignored.



Non-functional requirements:


* The dispatching algorithm should be configurable, allowing the system to easily switch between different optimization strategies.



<div style="border-bottom: 2px solid rgba(72,149,239,0.3); padding-bottom: 8px; margin: 28px 0 14px 0;">
  <span style="font-size: 1.3em; font-weight: 600; color: #4895ef;">Understanding elevator control panels</span>
</div>


Elevator systems include two types of buttons:


* **Hallway Buttons** (Outside): Located on each floor — an “up” button and a “down” button.

* **Floor Buttons** (Inside): Located on the control panel inside the elevator car, one for each floor.

* **Elevator Dispatch**: Handles hallway calls by assigning the most appropriate elevator using the dispatching strategy.

* **Elevator Car**: A single unit that transports passengers. Each car operates independently with its own control panel.

Since the use cases are clearer than the underlying data model, we’ll start with the system’s behaviors and use them to guide the definition of classes.


<div style="border-bottom: 2px solid rgba(72,149,239,0.3); padding-bottom: 8px; margin: 28px 0 14px 0;">
  <span style="font-size: 1.3em; font-weight: 600; color: #4895ef;">ElevatorSystem</span>
</div>


The ElevatorSystem class serves as the central controller, providing an API for controlling all the elevator cars.


Three core responsibilities: **Get status**, **Request elevator** (hallway button), **Select destination** (floor button inside car).


![Figure 3: ElevatorSystem class](assets/image-8-3-YVLY4HDU.svg)


<div style="border-bottom: 2px solid rgba(72,149,239,0.3); padding-bottom: 8px; margin: 28px 0 14px 0;">
  <span style="font-size: 1.3em; font-weight: 600; color: #4895ef;">ElevatorCar</span>
</div>


The ElevatorCar class models the behaviors of an elevator car. It maintains a queue of target floors and delegates state management to the ElevatorStatus class.


![Figure 4: ElevatorCar and ElevatorStatus](assets/image-8-4-PHMTFAIX.svg)


**Design choice**: We separated the elevator car’s state into a dedicated ElevatorStatus class to promote reusability and clarity.


<div style="border-bottom: 2px solid rgba(72,149,239,0.3); padding-bottom: 8px; margin: 28px 0 14px 0;">
  <span style="font-size: 1.3em; font-weight: 600; color: #4895ef;">Direction</span>
</div>


The Direction enum provides a type-safe way to represent an elevator’s movement direction: **UP**, **DOWN**, or **IDLE**.


![Figure 5: Direction enum](assets/image-8-5-ORT5QMKG.svg)


<div style="border-bottom: 2px solid rgba(72,149,239,0.3); padding-bottom: 8px; margin: 28px 0 14px 0;">
  <span style="font-size: 1.3em; font-weight: 600; color: #4895ef;">Elevator Dispatch and Dispatching Strategy</span>
</div>


The ElevatorDispatch class manages user requests and selects the appropriate elevator car using the Strategy Pattern.


The general dispatching process follows three steps:


* **Review the Request**: Evaluate the incoming request and the status of all elevator cars.

* **Select the Best Car**: Based on the DispatchingStrategy, determine the most suited elevator.

* **Update Next Stops**: Add the requested floor to the selected car’s list of upcoming stops.



![Figure 6: ElevatorDispatch and DispatchingStrategy](assets/image-8-6-BODQCSS4.svg)


<div style="margin: 22px 0 10px 0; border-left: 3px solid #4895ef; padding-left: 10px;">
  <span style="font-size: 1.15em; font-weight: 600; font-style: italic; color: #4895ef;">Common dispatching strategies</span>
</div>


* **First Come, First Serve (FCFS)**: Assigns the request to the next available elevator.

* **Shortest Seek Time First (SSTF)**: Assigns the request to the closest elevator moving in the right direction.

* **Dynamic Strategies**: Configurable based on traffic patterns.

<div style="border-left: 4px solid #4895ef; border-radius: 4px; overflow: hidden; margin: 15px 0;">

```java
public class ElevatorSystem {
    private final List<ElevatorCar> elevators;
    private final ElevatorDispatch dispatchController;

    public ElevatorSystem(List<ElevatorCar> elevators, DispatchingStrategy strategy) {
        this.elevators = elevators;
        this.dispatchController = new ElevatorDispatch(strategy);
    }

    public List<ElevatorStatus> getAllElevatorStatuses() {
        List<ElevatorStatus> statuses = new ArrayList<>();
        for (ElevatorCar elevator : elevators) {
            statuses.add(elevator.getStatus());
        }
        return statuses;
    }

    public void requestElevator(int currentFloor, Direction direction) {
        dispatchController.dispatchElevatorCar(currentFloor, direction, elevators);
    }

    public void selectFloor(ElevatorCar car, int destinationFloor) {
        car.addFloorRequest(destinationFloor);
    }
}
```

</div>


<div style="border-bottom: 2px solid rgba(72,149,239,0.3); padding-bottom: 8px; margin: 28px 0 14px 0;">
  <span style="font-size: 1.3em; font-weight: 600; color: #4895ef;">ElevatorCar</span>
</div>


<div style="border-left: 4px solid #4895ef; border-radius: 4px; overflow: hidden; margin: 15px 0;">

```java
public class ElevatorCar {
    private ElevatorStatus status;
    private final Queue<Integer> targetFloors;

    public ElevatorCar(int startingFloor) {
        this.status = new ElevatorStatus(startingFloor, Direction.IDLE);
        this.targetFloors = new LinkedList<>();
    }

    public ElevatorStatus getStatus() { return status; }

    public void addFloorRequest(int floor) {
        if (!targetFloors.contains(floor)) {
            targetFloors.offer(floor);
            updateDirection(floor);
        }
    }

    public boolean isIdle() { return targetFloors.isEmpty(); }

    private void updateDirection(int targetFloor) {
        if (status.getCurrentFloor() < targetFloor) {
            status = new ElevatorStatus(status.getCurrentFloor(), Direction.UP);
        } else if (status.getCurrentFloor() > targetFloor) {
            status = new ElevatorStatus(status.getCurrentFloor(), Direction.DOWN);
        }
    }
}
```

</div>


**Implementation choice**: We used a Queue (FIFO) to ensure floor requests are processed in the order they are received. A PriorityQueue could optimize travel time but might confuse passengers expecting stops in request order.


<div style="border-bottom: 2px solid rgba(72,149,239,0.3); padding-bottom: 8px; margin: 28px 0 14px 0;">
  <span style="font-size: 1.3em; font-weight: 600; color: #4895ef;">ElevatorDispatch</span>
</div>


<div style="border-left: 4px solid #4895ef; border-radius: 4px; overflow: hidden; margin: 15px 0;">

```java
public class ElevatorDispatch {
    private final DispatchingStrategy strategy;

    public ElevatorDispatch(DispatchingStrategy strategy) {
        this.strategy = strategy;
    }

    public void dispatchElevatorCar(int floor, Direction direction,
                                     List<ElevatorCar> elevators) {
        ElevatorCar selectedElevator = strategy.selectElevator(elevators, floor, direction);
        if (selectedElevator != null) {
            selectedElevator.addFloorRequest(floor);
        }
    }
}
```

</div>


<div style="border-bottom: 2px solid rgba(72,149,239,0.3); padding-bottom: 8px; margin: 28px 0 14px 0;">
  <span style="font-size: 1.3em; font-weight: 600; color: #4895ef;">Dispatching Strategies</span>

<div style="border-left: 4px solid #4895ef; border-radius: 4px; overflow: hidden; margin: 15px 0;">

```java
public class FirstComeFirstServeStrategy implements DispatchingStrategy {
    @Override
    public ElevatorCar selectElevator(List<ElevatorCar> elevators,
                                       int floor, Direction direction) {
        for (ElevatorCar elevator : elevators) {
            if (elevator.isIdle() || elevator.getCurrentDirection() == direction) {
                return elevator;
            }
        }
        return elevators.get((int) (Math.random() * elevators.size()));
    }
}
```

</div>


<div style="border-left: 4px solid #4895ef; border-radius: 4px; overflow: hidden; margin: 15px 0;">

```java
public class ShortestSeekTimeFirstStrategy implements DispatchingStrategy {
    @Override
    public ElevatorCar selectElevator(List<ElevatorCar> elevators,
                                       int floor, Direction direction) {
        ElevatorCar bestElevator = null;
        int shortestDistance = Integer.MAX_VALUE;
        for (ElevatorCar elevator : elevators) {
            int distance = Math.abs(elevator.getCurrentFloor() - floor);
            if ((elevator.isIdle() || elevator.getCurrentDirection() == direction)
                    && distance < shortestDistance) {
                bestElevator = elevator;
                shortestDistance = distance;
            }
        }
        return bestElevator;
    }
}
```

  <span style="font-size: 1.3em; font-weight: 600; color: #4895ef;">Event-driven elevator request handling</span>
</div>


To decouple hallway buttons from the dispatch controller, we introduce an event-driven approach using the **Observer Pattern**.


**Observable Subject**: Hallway buttons act as subjects. When pressed, they trigger an event that notifies the dispatch controller.


**Observer**: The dispatch controller listens to button-press events and responds by allocating an appropriate elevator car.


![Figure 8: Observer pattern for hallway calls](assets/image-8-8-B2XTQDAE.svg)


<div style="margin: 22px 0 10px 0; border-left: 3px solid #4895ef; padding-left: 10px;">
  <span style="font-size: 1.15em; font-weight: 600; font-style: italic; color: #4895ef;">Code changes</span>
</div>


<div style="border-left: 4px solid #4895ef; border-radius: 4px; overflow: hidden; margin: 15px 0;">

```java
// Observable Subject: HallwayButtonPanel
public class HallwayButtonPanel {
    private final int floor;
    private final List<ElevatorObserver> observers;

    public HallwayButtonPanel(int floor) {
        this.floor = floor;
        this.observers = new ArrayList<>();
    }

    public void pressButton(Direction direction) {
        notifyObservers(direction);
    }

    public void addObserver(ElevatorObserver observer) {
        observers.add(observer);
    }

    private void notifyObservers(Direction direction) {
        for (ElevatorObserver observer : observers) {
            observer.update(floor, direction);
        }
    }
}

// Observer Interface
public interface ElevatorObserver {
    void update(int floor, Direction direction);
}

// Observer Implementation
public class ElevatorDispatchController implements ElevatorObserver {
    @Override
    public void update(int floor, Direction direction) {
        // Logic to handle the floor request
    }
}
```

<div style="border-bottom: 2px solid rgba(72,149,239,0.3); padding-bottom: 8px; margin: 28px 0 14px 0;">
  <span style="font-size: 1.3em; font-weight: 600; color: #4895ef;">Handling elevators serving different floor sets</span>
</div>


To support elevators that only stop at specific floors, we add an `accessibleFloors` attribute to ElevatorCar and update the dispatching logic:


<div style="border-left: 4px solid #4895ef; border-radius: 4px; overflow: hidden; margin: 15px 0;">

```java
// In ElevatorCar: add accessible floors
private final Set<Integer> accessibleFloors;

public void addFloorRequest(int floor) {
    if (accessibleFloors.contains(floor) && !targetFloors.contains(floor)) {
        targetFloors.offer(floor);
        updateDirection(floor);
    }
}
```

</div>


<div style="border-left: 4px solid #4895ef; border-radius: 4px; overflow: hidden; margin: 15px 0;">

```java
// Updated ShortestSeekTimeFirstStrategy
public class ShortestSeekTimeFirstStrategy implements DispatchingStrategy {
    @Override
    public ElevatorCar selectElevator(List<ElevatorCar> elevators,
                                       int floor, Direction direction) {
        ElevatorCar bestElevator = null;
        int shortestDistance = Integer.MAX_VALUE;
        for (ElevatorCar elevator : elevators) {
            int distance = Math.abs(elevator.getCurrentFloor() - floor);
            if ((elevator.isIdle() || elevator.getCurrentDirection() == direction)
                    && elevator.getAccessibleFloors().contains(floor)
                    && distance < shortestDistance) {
                bestElevator = elevator;
                shortestDistance = distance;
            }
        }
        return bestElevator;
    }
}
```

The Observer is a behavioral pattern that lets you define a subscription mechanism, allowing multiple objects to receive notifications automatically whenever the observed object changes state.


<div style="margin: 22px 0 10px 0; border-left: 3px solid #4895ef; padding-left: 10px;">
  <span style="font-size: 1.15em; font-weight: 600; font-style: italic; color: #4895ef;">Problem</span>
</div>


Imagine developing a news application that delivers real-time updates. Directly linking the publisher to each user creates tight coupling.


<div style="margin: 22px 0 10px 0; border-left: 3px solid #4895ef; padding-left: 10px;">
  <span style="font-size: 1.15em; font-weight: 600; font-style: italic; color: #4895ef;">Solution</span>
</div>


The Observer pattern establishes a one-to-many relationship between the publisher (subject) and subscribers (observers). When state changes, all subscribed observers are notified automatically.


![Figure 9: Observer design pattern](assets/image-8-9-TUYG6ULH.svg)


<div style="margin: 22px 0 10px 0; border-left: 3px solid #4895ef; padding-left: 10px;">
  <span style="font-size: 1.15em; font-weight: 600; font-style: italic; color: #4895ef;">When to use</span>
</div>


* Changes in one object require notifying other objects, especially when the set of objects isn’t known in advance.

* When objects need to observe other objects for specific conditions or limited time.
