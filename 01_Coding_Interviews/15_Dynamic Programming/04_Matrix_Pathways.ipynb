<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

<h2 class="bbg-h2" style="color: var(--bbg-title-color, var(--vscode-foreground, var(--jp-content-font-color1, CanvasText))); border-left: 5px solid var(--bbg-accent-green, var(--vscode-charts-green, var(--jp-success-color1, #16a34a))); font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Intuition</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

At each cell, we can either move right or move down. No matter which direction we choose at any point, it will always move us closer to the destination.

This means we just need to keep moving either right or down until we can no longer do so, at which point we’ve reached the bottom-right corner. Let's think about this problem backward.

Assume we have already reached the bottom-right corner. How did we get here?

We know for certain we came from either the cell directly above, or the cell directly to the left of the current position.

</div>

<figure class="bbg-figure" style="background: #ffffff; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 14px auto 20px; padding: 14px; box-sizing: border-box; width: 96%;">
  <img src="assets/image-15-03-1-TAOW5PIC.svg" alt="Diagram 2" style="display: block; height: auto; margin: 0 auto; max-width: 100%; width: 100%;">
  <figcaption class="bbg-caption" style="color: #475569; font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; font-size: 12px; margin-top: 8px; text-align: center;">Diagram 2</figcaption>
</figure>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

This is equally true for any cell on the matrix, which means a generalization can be made: the number of paths to any cell is equal to the sum of the number of paths to the cell above it and the cell to its left.

</div>

<figure class="bbg-figure" style="background: #ffffff; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 14px auto 20px; padding: 14px; box-sizing: border-box; width: 96%;">
  <img src="assets/image-15-03-2-MFWZHEZM.svg" alt="Diagram 3" style="display: block; height: auto; margin: 0 auto; max-width: 100%; width: 100%;">
  <figcaption class="bbg-caption" style="color: #475569; font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; font-size: 12px; margin-top: 8px; text-align: center;">Diagram 3</figcaption>
</figure>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

This demonstrates the existence of subproblems, and that this problem has an **Optimal** substructure, where we need to solve two subproblems in order to solve the main problem. This makes this problem well-suited for DP.

So, let’s translate the above recurrence relation to a DP formula: dp[r][c] = dp[r - 1][c] + dp[r][c - 1], where dp[r][c] represents the total number of paths that lead to cell ( r, c). Before populating the DP table, we need to know our base cases.

Base cases We know dp[0][0] should be 1 because there’s only one path leading to cell (0, 0). What else do we know for certain?

Since we can only move right or down, once we leave a row or column, we can never return to it because we can't move left or up. This means, for any cell in row 0 or column 0, there’s only one path to those cells:

</div>

<figure class="bbg-figure" style="background: #ffffff; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 14px auto 20px; padding: 14px; box-sizing: border-box; width: 96%;">

</figure>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

Therefore, we can set all cells in row 0 and column 0 to 1 as the base cases. Problem-solving tip: another way to identify row 0 and column 0 as base cases is by examining the DP formula.

Since we need the values from row r - 1 and column c - 1 to populate dp[r][c], all values at r = 0 and c = 0 must be pre-populated before using the formula to avoid index out-of-bound errors. Populating the DP table Once the base cases are set, we can populate the remaining DP table, starting from cell (1, 1), using our DP formula ( dp[r][c] = dp[r-1][c] + dp[r][c-1]):

</div>

<figure class="bbg-figure" style="background: #ffffff; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 14px auto 20px; padding: 14px; box-sizing: border-box; width: 96%;">

After we fill in the DP table, we can return dp[m - 1][n - 1], which contains the number of paths to the bottom-right corner.

</div>

<h2 class="bbg-h2" style="color: var(--bbg-title-color, var(--vscode-foreground, var(--jp-content-font-color1, CanvasText))); border-left: 5px solid var(--bbg-accent-green, var(--vscode-charts-green, var(--jp-success-color1, #16a34a))); font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Try it yourself</h2>

<h2 class="bbg-h2" style="color: var(--bbg-title-color, var(--vscode-foreground, var(--jp-content-font-color1, CanvasText))); border-left: 5px solid var(--bbg-accent-green, var(--vscode-charts-green, var(--jp-success-color1, #16a34a))); font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Implementation</h2>

<h4 class="bbg-h4" style="color: var(--bbg-h4-color, var(--vscode-terminal-ansiYellow, var(--jp-warn-color1, #b45309))); font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; letter-spacing: 0; margin: 18px 0 8px;">Python Implementation</h4>

```python
def matrix_pathways(m: int, n: int) -> int:
    # Base cases: Set all cells in row 0 and column 0 to 1. We can do this by
    # initializing all cells in the DP table to 1.
    dp = [[1] * n for _ in range(m)]
    # Fill in the rest of the DP table.
    for r in range(1, m):
        for c in range(1, n):
            # Paths to current cell = paths from above + paths from left.
            dp[r][c] = dp[r - 1][c] + dp[r][c - 1]
    return dp[m - 1][n - 1]
```

<h2 class="bbg-h2" style="color: var(--bbg-title-color, var(--vscode-foreground, var(--jp-content-font-color1, CanvasText))); border-left: 5px solid var(--bbg-accent-green, var(--vscode-charts-green, var(--jp-success-color1, #16a34a))); font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Complexity Analysis</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

**Time Complexity**: The **Time Complexity** of matrix_pathways is $\mathcal{O}(m\cdot n)$ because each cell in the DP table is populated once. **Space Complexity**: The **Space Complexity** is $\mathcal{O}(m\cdot n)$ due to the DP table, which contains m$\cdot$ n elements.

</div>

<h2 class="bbg-h2" style="color: var(--bbg-title-color, var(--vscode-foreground, var(--jp-content-font-color1, CanvasText))); border-left: 5px solid var(--bbg-accent-green, var(--vscode-charts-green, var(--jp-success-color1, #16a34a))); font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Optimization</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

We can optimize our solution by understanding that, for each cell in the DP table, we only need to access the cells directly above it and to its left.

</div>

<figure class="bbg-figure" style="background: #ffffff; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 14px auto 20px; padding: 14px; box-sizing: border-box; width: 96%;">
  <img src="assets/image-15-03-5-LJYBRPZ2.svg" alt="Diagram 6" style="display: block; height: auto; margin: 0 auto; max-width: 100%; width: 100%;">
  <figcaption class="bbg-caption" style="color: #475569; font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; font-size: 12px; margin-top: 8px; text-align: center;">Diagram 6</figcaption>
</figure>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

- To get the cell above it ( dp[r-1][c]), we only need access to the previous row.

- To get the cell to its left ( dp[r][c-1]), we just need to look at the cell to the left of the current cell, which is in the same row we're currently populating.

Therefore, we only need to maintain two rows:

</div>

<figure class="bbg-figure" style="background: #ffffff; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 14px auto 20px; padding: 14px; box-sizing: border-box; width: 96%;">
  <img src="assets/image-15-03-6-EIJMVIJT.svg" alt="Diagram 7" style="display: block; height: auto; margin: 0 auto; max-width: 100%; width: 100%;">
  <figcaption class="bbg-caption" style="color: #475569; font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; font-size: 12px; margin-top: 8px; text-align: center;">Diagram 7</figcaption>
</figure>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

- prev_row: the previous row.

- curr_row: the current row being populated.

This effectively reduces the **Space Complexity** to $\mathcal{O}(n)$ because we now only need to maintain two arrays of size n. After populating the DP values for the current row, we’ll need to make sure to update prev_row with the values from curr_row to prepare for the next iteration since the next row’s previous row is the current row. Below is the optimized code:

</div>

<h4 class="bbg-h4" style="color: var(--bbg-h4-color, var(--vscode-terminal-ansiYellow, var(--jp-warn-color1, #b45309))); font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; letter-spacing: 0; margin: 18px 0 8px;">Python Snippet</h4>

```python
def matrix_pathways_optimized(m: int, n: int) -> int:
    # Initialize 'prev_row' as the DP values of row 0, which are all 1s.
    prev_row = [1] * n
    # Iterate through the matrix starting from row 1.
    for r in range(1, m):
        # Set the first cell of 'curr_row' to 1. This is done by
        # setting the entire row to 1.
        curr_row = [1] * n
        for c in range(1, n):
            # The number of unique paths to the current cell is the sum
            # of the paths from the cell above it ('prev_row[c]') and
            # the cell to the left ('curr_row[c - 1]').
            curr_row[c] = prev_row[c] + curr_row[c - 1]
        # Update 'prev_row' with 'curr_row' values for the next
        # iteration.
        prev_row = curr_row
    # The last element in 'prev_row' stores the result for the
    # bottom-right cell.
    return prev_row[n - 1]
```
