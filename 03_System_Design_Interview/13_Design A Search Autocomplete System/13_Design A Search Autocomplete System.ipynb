<ul>
<li>Fast response time: As a user types a search query, autocomplete suggestions must show up fast enough. An article about Facebook’s autocomplete system [1] reveals that the system needs to return results within 100 milliseconds. Otherwise it will cause stuttering.</li>
<li>Relevant: Autocomplete suggestions should be relevant to the search term.</li>
<li>Sorted: Results returned by the system must be sorted by popularity or other ranking models.</li>
<li>Scalable: The system can handle high traffic volume.</li>
<li>Highly available: The system should remain available and accessible when part of the system is offline, slows down, or experiences unexpected network errors.</li>
</ul>

<li>20 bytes of data per query string:</li>
<li>Assume we use ASCII character encoding. 1 character = 1 byte</li>
<li>Assume a query contains 4 words, and each word contains 5 characters on average.</li>
<li>That is 4 x 5 = 20 bytes per query.</li>
<li>For every character entered into the search box, a client sends a request to the backend for autocomplete suggestions. On average, 20 requests are sent for each search query. For example, the following 6 requests are sent to the backend by the time you finish typing “dinner”.  search?q=d  search?q=di  search?q=din  search?q=dinn  search?q=dinne  search?q=dinner</li>
<li>~24,000 query per second (QPS) = 10,000,000 users <em> 10 queries / day </em> 20 characters / 24 hours / 3600 seconds.</li>
<li>Peak QPS = QPS * 2 = ~48,000</li>
<li>Assume 20% of the daily queries are new. 10 million <em> 10 queries / day </em> 20 byte per query * 20% = 0.4 GB. This means 0.4GB of new data is added to storage daily.</li>
</ul>
</section>

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(198,255,0,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #c6ff00;">Trie data structure</span>
</div>
<p>Relational databases are used for storage in the high-level design. However, fetching the top 5 search queries from a relational database is inefficient. The data structure trie (prefix tree) is used to overcome the problem. As trie data structure is crucial for the system, we will dedicate significant time to design a customized trie. Please note that some of the ideas are from articles [2] and [3].</p>
<p>Understanding the basic trie data structure is essential for this interview question. However, this is more of a data structure question than a system design question. Besides, many online materials explain this concept. In this chapter, we will only discuss an overview of the trie data structure and focus on how to optimize the basic trie to improve response time.</p>
<p>Trie (pronounced “try”) is a tree-like data structure that can compactly store strings. The name comes from the word re<strong>trie</strong>val, which indicates it is designed for string retrieval operations. The main idea of trie consists of the following:</p>
<ul>
<li>A trie is a tree-like data structure.</li>
<li>The root represents an empty string.</li>
<li>Each node stores a character and has 26 children, one for each possible character. To save space, we do not draw empty links.</li>
<li>Each tree node represents a single word or a prefix string.</li>
</ul>
<p>Figure 5 shows a trie with search queries “tree”, “try”, “true”, “toy”, “wish”, “win”. Search queries are highlighted with a thicker border.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-13-7-KXYAA5K3.svg" alt="Figure 5" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>Basic trie data structure stores characters in nodes. To support sorting by frequency, frequency info needs to be included in nodes. Assume we have the following frequency table.</p>

<p>After adding frequency info to nodes, updated trie data structure is shown in Figure 6.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-13-8-4HDDZCTY.svg" alt="Figure 6" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>How does autocomplete work with trie? Before diving into the algorithm, let us define some terms.</p>
<ul>
<li>p: length of a prefix</li>
<li>n: total number of nodes in a trie</li>
<li>c: number of children of a given node</li>
</ul>
<p>Steps to get top <em>k</em> most searched queries are listed below:</p>
<ol>
<li>Find the prefix. Time complexity: <em>O(p)</em>.</li>
<li>Traverse the subtree from the prefix node to get all valid children. A child is valid if it can form a valid query string. Time complexity: <em>O(c)</em></li>
<li>Sort the children and get top <em>k</em>. Time complexity: <em>O(clogc)</em></li>
</ol>
<p>Let us use an example as shown in Figure 7 to explain the algorithm. Assume <em>k</em> equals to 2 and a user types “tr” in the search box. The algorithm works as follows:</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-13-9-O7NXHK4L.svg" alt="Figure 7" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><ul>
<li>Step 1: Find the prefix node “tr”.</li>
<li>Step 2: Traverse the subtree to get all valid children. In this case, nodes [tree: 10], [true: 35], [try: 29] are valid.</li>
<li>Step 3: Sort the children and get top 2. [true: 35] and [try: 29] are the top 2 queries with prefix “tr”.</li>
</ul>
<p>The time complexity of this algorithm is the sum of time spent on each step mentioned above: <strong>O(p) + O(c) + O(clogc)</strong></p>
<p>The above algorithm is straightforward. However, it is too slow because we need to traverse the entire trie to get top <em>k</em> results in the worst-case scenario. Below are two optimizations:</p>
<ol>
<li>Limit the max length of a prefix</li>
<li>Cache top search queries at each node</li>
</ol>
<p>Let us look at these optimizations one by one.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(198,255,0,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Limit the max length of a prefix</span>
</div>
<p>Users rarely type a long search query into the search box. Thus, it is safe to say <em>p</em> is a small integer number, say 50. If we limit the length of a prefix, the time complexity for “Find the prefix” can be reduced from <em>O(p)</em> to <em>O(small constant),</em> aka <em>O(1)</em>.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(198,255,0,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Cache top search queries at each node</span>
</div>
<p>To avoid traversing the whole trie, we store top <em>k</em> most frequently used queries at each node. Since 5 to 10 autocomplete suggestions are enough for users, <em>k</em> is a relatively small number. In our specific case, only the top 5 search queries are cached.</p>
<p>By caching top search queries at every node, we significantly reduce the time complexity to retrieve the top 5 queries. However, this design requires a lot of space to store top queries at every node. Trading space for time is well worth it as fast response time is very important.</p>
<p>Figure 8 shows the updated trie data structure. Top 5 queries are stored on each node. For example, the node with prefix “be” stores the following: [best: 35, bet: 29, bee: 20, be: 15, beer: 10].</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-13-10-JX4JXRHK.svg" alt="Figure 8" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>Let us revisit the time complexity of the algorithm after applying those two optimizations:</p>
<ol>
<li>Find the prefix node. Time complexity: <em>O(1)</em></li>
<li>Return top <em>k</em>. Since top <em>k</em> queries are cached, the time complexity for this step is <em>O(1)</em>.</li>
</ol>
<p>As the time complexity for each of the steps is reduced to <em>O(1)</em>, our algorithm takes only <em>O(1)</em> to fetch top <em>k</em> queries.</p>
</section>

<ul>
<li>Users may enter billions of queries per day. Updating the trie on every query significantly slows down the query service.</li>
<li>Top suggestions may not change much once the trie is built. Thus, it is unnecessary to update the trie frequently.</li>
</ul>
<p>To design a scalable data gathering service, we examine where data comes from and how data is used. Real-time applications like Twitter require up to date autocomplete suggestions. However, autocomplete suggestions for many Google keywords might not change much on a daily basis.</p>
<p>Despite the differences in use cases, the underlying foundation for data gathering service remains the same because data used to build the trie is usually from analytics or logging services.</p>

<p>In the high-level design, query service calls the database directly to fetch the top 5 results. Figure 11 shows the improved design as previous design is inefficient.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-13-15-6H4WZOA6.svg" alt="Figure 11" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><ol>

<li>The load balancer routes the request to API servers.</li>
<li>API servers get trie data from Trie Cache and construct autocomplete suggestions for the client.</li>
<li>In case the data is not in Trie Cache, we replenish data back to the cache. This way, all subsequent requests for the same prefix are returned from the cache. A cache miss can happen when a cache server is out of memory or offline.</li>
</ol>
<p>Query service requires lightning-fast speed. We propose the following optimizations:</p>
<ul>
<li>AJAX request. For web applications, browsers usually send AJAX requests to fetch autocomplete results. The main benefit of AJAX is that sending/receiving a request/response does not refresh the whole web page.</li>
<li>Browser caching. For many applications, autocomplete search suggestions may not change much within a short time. Thus, autocomplete suggestions can be saved in browser cache to allow subsequent requests to get results from the cache directly. Google search engine uses the same cache mechanism. Figure 12 shows the response header when you type “system design interview” on the Google search engine. As you can see, Google caches the results in the browser for 1 hour. Please note: “private” in cache-control means results are intended for a single user and must not be cached by a shared cache. “max-age=3600” means the cache is valid for 3600 seconds, aka, an hour.</li>
<li>Data sampling: For a large-scale system, logging every search query requires a lot of processing power and storage. Data sampling is important. For instance, only 1 out of every <em>N</em> requests is logged by the system.</li>
</ul>
</section>

</div>
<p>Trie is a core component of the autocomplete system. Let us look at how trie operations (create, update, and delete) work.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(198,255,0,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Create</span>
</div>
<p>Trie is created by workers using aggregated data. The source of data is from Analytics Log/DB.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(198,255,0,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Update</span>
</div>
<p>There are two ways to update the trie.</p>
<p>Option 1: Update the trie weekly. Once a new trie is created, the new trie replaces the old one.</p>
<p>Option 2: Update individual trie node directly. We try to avoid this operation because it is slow. However, if the size of the trie is small, it is an acceptable solution. When we update a trie node, its ancestors all the way up to the root must be updated because ancestors store top queries of children. Figure 13 shows an example of how the update operation works. On the left side, the search query “beer” has the original value 10. On the right side, it is updated to 30. As you can see, the node and its ancestors have the “beer” value updated to 30.</p>
</section>

  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Delete</span>
</div>
<p>We have to remove hateful, violent, sexually explicit, or dangerous autocomplete suggestions. We add a filter layer (Figure 14) in front of the Trie Cache to filter out unwanted suggestions. Having a filter layer gives us the flexibility of removing results based on different filter rules. Unwanted suggestions are removed physically from the database asynchronically so the correct data set will be used to build trie in the next update cycle.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(198,255,0,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #c6ff00;">Scale the storage</span>
</div>
<p>Now that we have developed a system to bring autocomplete queries to users, it is time to solve the scalability issue when the trie grows too large to fit in one server.</p>
<p>Since English is the only supported language, a naive way to shard is based on the first character. Here are some examples.</p>
<ul>
<li>If we need two servers for storage, we can store queries starting with ‘a’ to ‘m’ on the first server, and ‘n’ to ‘z’ on the second server.</li>
<li>If we need three servers, we can split queries into ‘a’ to ‘i’, ‘j’ to ‘r’ and ‘s’ to ‘z’.</li>
</ul>
<p>Following this logic, we can split queries up to 26 servers because there are 26 alphabetic characters in English. Let us define sharding based on the first character as first level sharding. To store data beyond 26 servers, we can shard on the second or even at the third level. For example, data queries that start with ‘<em>a</em>’ can be split into 4 servers: ‘<em>aa-ag</em>’, ‘<em>ah-an</em>’, ‘<em>ao-au</em>’, and ‘<em>av-az’</em>.</p>
<p>At the first glance this approach seems reasonable, until you realize that there are a lot more words that start with the letter ‘<em>c</em>’ than ‘<em>x</em>’. This creates uneven distribution.</p>
<p>To mitigate the data imbalance problem, we analyze historical data distribution pattern and apply smarter sharding logic as shown in Figure 15. The shard map manager maintains a lookup database for identifying where rows should be stored. For example, if there are a similar number of historical queries for ‘<em>s</em>’ and for ‘<em>u</em>’, ‘<em>v</em>’, ‘<em>w</em>’, ‘<em>x</em>’, ‘<em>y</em>’ and ‘<em>z</em>’ combined, we can maintain two shards: one for ‘s’ and one for ‘u’ to ‘z’.</p>
</section>

<p>After you finish the deep dive, your interviewer might ask you some follow up questions.</p>
<p>Interviewer: How do you extend your design to support multiple languages?</p>
<p>To support other non-English queries, we store Unicode characters in trie nodes. If you are not familiar with Unicode, here is the definition: “an encoding standard covers all the characters for all the writing systems of the world, modern and ancient” [5].</p>
<p>Interviewer: What if top search queries in one country are different from others?</p>
<p>In this case, we might build different tries for different countries. To improve the response time, we can store tries in CDNs.</p>
<p>Interviewer: How can we support the trending (real-time) search queries?</p>
<p>Assuming a news event breaks out, a search query suddenly becomes popular. Our original design will not work because:</p>
<ul>
<li>Offline workers are not scheduled to update the trie yet because this is scheduled to run on weekly basis.</li>
<li>Even if it is scheduled, it takes too long to build the trie.</li>
</ul>
<p>Building a real-time search autocomplete is complicated and is beyond the scope of this course so we will only give a few ideas:</p>
<ul>
<li>Reduce the working data set by sharding.</li>
<li>Change the ranking model and assign more weight to recent search queries.</li>
<li>Data may come as streams, so we do not have access to all the data at once. Streaming data means data is generated continuously. Stream processing requires a different set of systems: Apache Hadoop MapReduce [6], Apache Spark Streaming [7], Apache Storm [8], Apache Kafka [9], etc. Because all those topics require specific domain knowledge, we are not going into detail here.</li>
</ul>
<p>Congratulations on getting this far! Now give yourself a pat on the back. Good job!</p>
