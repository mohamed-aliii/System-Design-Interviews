<ul>
<li>5000 hotels and 1 million rooms in total.</li>
<li>Assume 70% of the rooms are occupied and the average stay duration is 3 days.</li>
<li>Estimated daily reservations: (1 million * 0.7) / 3 = 233,333 (rounding up to ~240,000)</li>
<li>Reservations per second = 240,000 / 10^5 seconds in a day = ~3. As we can see, the average reservation transaction per second (TPS) is not high.</li>
</ul>
<p>Next, let’s do a rough calculation of the QPS of all pages in the system. There are three steps in a typical customer flow:</p>
<ul>
<li>View hotel/room detail page. Users browse this page (query).</li>
<li>View the booking page. Users can confirm the booking details, such as dates, number of guests, payment information before booking (query).</li>
<li>Reserve a room. Users click on the “book” button to book the room and the room is reserved (transaction).</li>
</ul>
<p>Let’s assume around 10% of users reach the next step and 90% of users drop off the flow before reaching the final step. We can also assume that no prefetching feature (prefetching the content before the user reaches the next step) is implemented. Figure 1 shows a rough estimation of what the QPS looks like for different steps. We know the final reservation TPS is 3 so we can work backwards along the funnel. The QPS of the order confirmation page is 30 and the QPS for the detail page is 300.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-1-qps-distribution-4LPSPNX4.svg" alt="Figure 1" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p></section>

<p>Before we decide which database to use, let’s take a close look at the data access patterns. For the hotel reservation system, we need to support the following queries:</p>
<p>Query 1: View detailed information about a hotel.</p>
<p>Query 2: Find available types of rooms given a date range.</p>
<p>Query 3: Record a reservation.</p>
<p>Query 4: Look up a reservation or past history of reservations.</p>
<p>From the back-of-the-envelope estimation, we know the scale of the system is not large but we need to prepare for traffic surges during big events. With these requirements in mind, we choose a relational database because:</p>
<ul>
<li>A relational database works well with read-heavy and write less frequently workflows. This is because the number of users who visit the hotel website/apps is a few orders of magnitude higher than those who actually make reservations. NoSQL databases are generally optimized for writes and the relational database works well enough for read-heavy workflow.</li>
<li>A relational database provides ACID (atomicity, consistency, isolation, durability) guarantees. ACID properties are important for a reservation system. Without those properties, it’s not easy to prevent problems such as negative balance, double charge, double reservations, etc. ACID properties make application code a lot simpler and make the whole system easier to reason about. A relational database usually provides these guarantees.</li>
<li>A relational database can easily model the data. The structure of the business data is very clear and the relationship between different entities (hotel, room, room_type, etc) is stable. This kind of data model is easily modeled by a relational database.</li>
</ul>
<p>Now that we have chosen the relational database as our data store, let’s explore the schema design. Figure 2 shows a straightforward schema design and it is the most natural way for many candidates to model the hotel reservation system.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-3-reservation-status-J57N4ZZV.svg" alt="Figure 2" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>Most attributes are self-explanatory and we will only explain the <em>status</em> field in the reservation table. The <em>status</em> field can be in one of these states: pending, paid, refunded, canceled, rejected. The state machine is shown in Figure 3.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-4-high-level-design-4TSKHYJP.svg" alt="Figure 3" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>This schema design has a major issue. This data model works for companies like Airbnb as <em>room_id</em> (might be called <em>listing_id</em>) is given when users make reservations. However, this isn’t the case for hotels. A user actually reserves <strong>a type of room</strong> in a given hotel instead of a specific room. For instance, a room type can be a standard room, king-size room, queen-size room with two queen beds, etc. Room numbers are given when the guest checks in and not at the time of the reservation. We need to update our data model to reflect this new requirement. See “Improved data model” in the ‘Deep Dive’ section for more details.</p>

<p>We use the microservice architecture for this hotel reservation system. Over the past few years, microservice architecture has gained great popularity. Companies that use microservice include Amazon, Netflix, Uber, Airbnb, Twitter, etc. If you want to learn more about the benefits of a microservice architecture, you can check out some good resources [1] [2].</p>
<p>Our design is modeled with the microservice architecture and the high-level design diagram is shown in Figure 4.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-connections-between-services-7TQS6ISN.svg" alt="Figure 4" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>We will briefly go over each component of the system from top to bottom.</p>
<ul>
<li>User: a user books a hotel room on their mobile phone or computer.</li>
<li>Admin (hotel staff): authorized hotel staff perform administrative operations such as refunding a customer, canceling a reservation, updating room information, etc.</li>
<li>CDN (content delivery network): for better load time, CDN is used to cache all static assets, including JavaScript bundles, images, videos, HTML, etc.</li>
<li>Public API Gateway: this is a fully managed service that supports rate limiting, authentication, etc. The API gateway is configured to direct requests to specific services based on the endpoints. For example, requests to load the hotel homepage are directed to the hotel service and requests to book a hotel room are routed to the reservation service.</li>
<li>Internal APIs: those APIs are only available to authorized hotel staff. They are accessible through internal software or websites. They are usually further protected by a VPN (virtual private network).</li>
<li>Hotel Service: this provides detailed information on hotels and rooms. Hotel and room data are generally static, so can be easily cached.</li>
<li>Rate Service: this provides room rates for different future dates. An interesting fact about the hotel industry is that the price of a room depends on how full the hotel is expected to be on a given day.</li>
<li>Reservation Service: receives reservation requests and reserves the hotel rooms. This service also tracks room inventory as rooms are reserved or reservations are canceled.</li>
<li>Payment Service: executes payment from a customer and updates the reservation status to “paid” once a payment transaction succeeds, or “rejected” if the transaction fails.</li>
<li>Hotel Management Service: only available to authorized hotel staff. Hotel staff are eligible to use the following features: view the record of an upcoming reservation, reserve a room for a customer, cancel a reservation, etc.</li>
</ul>
<p>For clarity, Figure 4 omits many arrows of interactions between microservices. For example, as shown in Figure 5, there should be an arrow between Reservation service and Rate service. Reservation service queries Rate service for room rates. This is used to compute the total room charge for a reservation. Another example is that there should be many arrows connecting the Hotel Management Service with most of the other services. When an admin makes changes via Hotel Management Service, the requests are forwarded to the actual service owning the data, to handle the changes.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-7-two-reservations-are-made-ZWW4GV4K.svg" alt="Figure 5" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>For production systems, inter-service communication often employs a modern and high-performance remote procedure call (RPC) framework like gPRC. There are many benefits to using such frameworks. To learn more about gPRC in particular, check out [3].</p>

<li>Transaction 1 checks if there are enough rooms by checking if (total_reserved + rooms_to_book) <= total_inventory. Since there is 1 room left, it also returns true.</li>

<li>Then transaction 2 reserves the room. The isolation property in ACID means database transactions must complete their tasks independently from other transactions. This results in the system allowing both users to book a room, even though there is only 1 room left.</li>
<li>Transaction 1 successfully commits the change.</li>
<li>Transaction 2 successfully commits the change.</li>
</ul>

  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Option 1: Pessimistic locking</span>
</div>

<p>Pessimistic locking [6], also called pessimistic concurrency control, prevents simultaneous updates by placing a lock on a record as soon as one user starts to update it. Other users who attempt to update the record have to wait until the first user has released the lock (committed the changes).</p>
<p>For MySQL, the “SELECT ... FOR UPDATE'' statement works by locking the rows returned by a selection query. Let’s assume a transaction is started by “transaction 1”. Other transactions have to wait for transaction 1 to finish before beginning another transaction. A detailed explanation is shown in Figure 12.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-15-database-sharding-GER3MC5O.svg" alt="Figure 12" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>In Figure 12, the “SELECT ... FOR UPDATE” statement of transaction 2 waits for transaction 1 to finish because transaction 1 locks the rows. After transaction 1 finishes, <em>total_reserved</em> becomes 100, which means there is no room for user 2 to book.</p>
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(0,184,212,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Pros:</span>
</div>

<ul>
<li>Prevents applications from updating data that is being – or has been – changed.</li>
<li>It is easy to implement and it avoids conflict by serializing updates. Pessimistic locking is useful when data contention is heavy.</li>
</ul>
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(0,184,212,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Cons:</span>
</div>

<ul>
<li>Deadlocks may occur when multiple resources are locked. Writing deadlock-free application code could be challenging.</li>
<li>This approach is not scalable. If a transaction is locked for too long, other transactions cannot access the resource. This has a significant impact on database performance, especially when transactions are long-lived or involve a lot of entities.</li>
</ul>
<p>Due to these limitations, we do not recommend pessimistic locking for the reservation system.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(0,184,212,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Option 2: Optimistic locking</span>
</div>

<p>Optimistic locking [7], also referred to as optimistic concurrency control, allows multiple concurrent users to attempt to update the same resource.</p>
<p>There are two common ways to implement optimistic locking: version number and timestamp. Version number is generally considered to be a better option because the server clock can be inaccurate over time. We explain how optimistic locking works with version number.</p>
<p>Figure 13 shows a successful case and a failure case.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-17-monolithic-vs-microservice-5V7CXVAI.svg" alt="Figure 13" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><ul>
<li>A new column called “version” is added to the database table.</li>
<li>Before a user modifies a database row, the application reads the version number of the row.</li>
<li>When the user updates the row, the application increases the version number by 1 and writes it back to the database.</li>
<li>A database validation check is put in place; the next version number should exceed the current version number by 1. The transaction aborts if the validation fails and the user tries again from step 2.</li>
</ul>
<p>Optimistic locking is usually faster than pessimistic locking because we do not lock the database. However, the performance of optimistic locking drops dramatically when concurrency is high.</p>
<p>To understand why, consider the case when many clients try to reserve a hotel room at the same time. Because there is no limit on how many clients can read the available room count, all of them read back the same available room count and the current version number. When different clients make reservations and write back the results to the database, only one of them will succeed, and the rest of the clients receive a version check failure message. These clients have to retry. In the subsequent round of retries, there is only one successful client again, and the rest have to retry. Although the end result is correct, repeated retries cause a very unpleasant user experience.</p>
<p>Pros:</p>
<ul>
<li>It prevents applications from editing stale data.</li>
<li>We don’t need to lock the database resource. There's actually no locking from the database point of view. It's entirely up to the application to handle the logic with the version number.</li>
<li>Optimistic locking is generally used when the data contention is low. When conflicts are rare, transactions can complete without the expense of managing locks.</li>
</ul>
<p>Cons:</p>
<ul>
<li>Performance is poor when data contention is heavy.</li>
</ul>
<p>Optimistic locking is a good option for a hotel reservation system since the QPS for reservations is usually not high.</p>
</section>

  <span style="font-size: 1.3em; font-weight: 600; color: #00b8d4;">Scalability</span>
</div>

<p>Usually, the load of the hotel reservation system is not high. However, the interviewer might have a follow-up question: “what if the hotel reservation system is used not just for a hotel chain, but for a popular travel site such as booking.com or expedia.com?” In this case, the QPS could be 1,000 times higher.</p>
<p>When the system load is high, we need to understand what might become the bottleneck. All our services are stateless, so they can be easily expanded by adding more servers. However, the database contains all the states and cannot be scaled up by simply adding more databases. Let’s explore how to scale the database.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(0,184,212,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Database sharding</span>

<p>One way to scale the database is to apply database sharding. The idea is to split the data into multiple databases so that each of them only contains a portion of data.</p>
<p>When we shard a database, we need to consider how to distribute the data. As we can see from the data model section, most queries need to filter by <em>hotel_id</em>. So a natural conclusion is we shard data by <em>hotel_id</em>. In Figure 15, the load is spread among 16 shards. Assume the QPS is 30,000. After database sharding, each shard handles 30,000 / 16 = 1875 QPS, which is within a single MySQL server’s load capacity.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-19-microservice-architecture-7FGGR4AA.svg" alt="Figure 15" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p>

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">

</div>

<p>The hotel inventory data has an interesting characteristic; only current and future hotel inventory data are meaningful because customers can only book rooms in the near future.</p>
<p>So for the storage choice, ideally we want to have a time-to-live (TTL) mechanism to expire old data automatically. Historical data can be queried on a different database. Redis is a good choice because TTL and Least Recently Used (LRU) cache eviction policy help us make optimal use of memory.</p>
<p>If the loading speed and database scalability become issues (for instance, we are designing at booking.com or expedia.com’s scale), we can add a cache layer on top of the database and move the check room inventory and reserve room logic to the cache layer, as shown in Figure 16. In this design, only a small percentage of the requests hit the inventory database as most ineligible requests are blocked by the inventory cache. One thing worth mentioning is that even when there is enough inventory shown in Redis, we still need to recheck the inventory at the database side as a precaution. The database is the source of truth for the inventory data.</p><p>Let’s first go over each component in this system.</p>
<p><strong>Reservation service</strong>: supports the following inventory management APIs:</p>
<ul>
<li>Query the number of available rooms for a given hotel ID, room type, and date range.</li>
<li>Reserve a room by executing total_reserved + 1.</li>
<li>Update inventory when a user cancels a reservation.</li>
</ul>
<p><strong>Inventory cache</strong>: all inventory management query operations are moved to the inventory cache (Redis) and we need to pre-populate inventory data to the cache. The cache is a key-value store with the following structure:</p>
<p>For a hotel reservation system, the volume of read operations (check room inventory) is an order of magnitude higher than write operations. Most of the read operations are answered by the cache.</p>
<p><strong>Inventory DB</strong>: stores inventory data as the source of truth.</p>
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(0,184,212,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">New challenges posed by the cache</span>
</div>

<p>Adding a cache layer significantly increases the system scalability and throughput, but it also introduces a new challenge: how to maintain data consistency between the database and the cache.</p>
<p>When a user books a room, two operations are executed in the happy path:</p>
<ul>
<li>Query room inventory to find out if there are enough rooms left. The query runs on the Inventory cache.</li>
<li>Update inventory data. The inventory DB is updated first. The change is then propagated to the cache asynchronously. This asynchronous cache update could be invoked by the application code, which updates the inventory cache after data is saved to the database. It could also be propagated using change data capture (CDC) [8]. CDC is a mechanism that reads data changes from the database and applies the changes to another data system. One common solution is Debezium [9]. It uses a source connector to read changes from a database and applies them to cache solutions such as Redis [10].</li>
</ul>
<p>Because the inventory data is updated on the database first, there is a possibility that the cache does not reflect the latest inventory data. For example, the cache may report there is still an empty room when the database says there is no room left, or vice versa.</p>
<p>If you think carefully, you find that the inconsistency between inventory cache and database actually does not matter, as long as the database does the final inventory validation check.</p>
<p>Let’s take a look at an example. Let’s say the cache states there is still an empty room, but the database says there is not. In this case, when the user queries the room inventory, they find there is still room available, so they try to reserve it. When the request reaches the inventory database, the database does the validation and finds that there is no room left. In this case, the client receives an error, indicating someone else just booked the last room before them. When a user refreshes the website, they probably see there is no room left because the database has synchronized inventory data to the cache, before they click the refresh button.</p>
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(0,184,212,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Pros</span>
</div>

<ul>
<li>Reduced database load. Since read queries are answered by the cache layer, database load is significantly reduced.</li>
<li>High performance. Read queries are very fast because results are fetched from memory.</li>
</ul>
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(0,184,212,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Cons</span>
</div>

<ul>
<li>Maintaining data consistency between the database and cache is hard. We need to think carefully about how this inconsistency affects user experience.</li>
</ul>

<div style="border-bottom: 2px solid rgba(0,184,212,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #00b8d4;">Data consistency among services</span>
</div>

<p>In a traditional monolithic architecture [11], a shared relational database is used to ensure data consistency. In our microservice design, we chose a hybrid approach by having Reservation Service handle both reservation and inventory APIs, so that the inventory and reservation database tables are stored in the same relational database. As explained in the “Concurrency Issues” section, this arrangement allows us to leverage the ACID properties of the relational database to elegantly handle many concurrency issues that arise during the reservation flow.</p>
<p>However, if your interviewer is a microservice purist, they might challenge this hybrid approach. In their mind, for a microservice architecture, each microservice has its own databases, as shown on the right in Figure 17.</p><p>This pure design introduces many data consistency issues. Since this is the first time we cover microservices, let’s explain how and why it happens. To make it easier to understand, only two services are used in this discussion. In the real world, there could be hundreds of microservices within a company. In a monolithic architecture, as shown in Figure 18, different operations can be wrapped within a single transaction to ensure ACID properties.</p><p>However, in a microservice architecture, each service has its own database. One logically atomic operation can span multiple services. This means we cannot use a single transaction to ensure data consistency. As shown in Figure 19, if the update operation fails in the reservation database, we need to roll back the reserved room count in the inventory database. Generally, there is only one happy path, but many failure cases that could cause data inconsistency.</p>
<p>To address the data inconsistency, here is a high-level summary of industry-proven techniques. If you want to read the details, please refer to the reference materials.</p>
<ul>
<li>Two-phase commit (2PC) [12]. 2PC is a database protocol used to guarantee atomic transaction commit across multiple nodes, i.e., either all nodes succeeded or all nodes failed. Because 2PC is a blocking protocol, a single node failure blocks the progress until the node has recovered. It’s not performant.</li>
<li>Saga. A saga is a sequence of local transactions. Each transaction updates and publishes a message to trigger the next transaction step. If a step fails, the saga executes compensating transactions to undo the changes that were made by preceding transactions [13]. 2PC works as a single commit to perform ACID transactions while Saga consists of multiple steps and relies on eventual consistency.</li>
</ul>
<p>It is worth noting that addressing data inconsistency between microservices requires some complicated mechanisms that greatly increase the complexity of the overall design. It is up to you as an architect to decide if the added complexity is worth it. For this problem, we decided that it was not worth it and so went with the more pragmatic approach of storing reservation and inventory data under the same relational database.</p>
