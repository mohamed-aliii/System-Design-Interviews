But what if we have more than two linked lists? Combining two linked lists involves comparing two nodes at each iteration, but combining k linked lists would require k comparisons per iteration.

The reason we need to make so many comparisons is that we don't know which node has the smallest value at any point in the iteration, requiring us to search for it. Wouldn't it be nice to have an efficient way to access the smallest-valued node at any given point?

A min-**Heap** is perfect for this. We can essentially do the same thing as in our initial approach, but instead of using pointers to determine the smallest node, we use a min-**Heap**.

<h2 class="bbg-h2" style="color: #e5edf7; border-left: 5px solid #39ff88; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Implementation</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

Note that in the implementation below, we modify the ListNode class globally to simplify the solution. It's important to confirm with your interviewer that global variables are acceptable.

</div>

<h4 class="bbg-h4" style="color: #f59e0b; font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; letter-spacing: 0; margin: 18px 0 8px;">Python Implementation</h4>

```python
from typing import List
from ds import ListNode
import heapq

def combine_sorted_linked_lists(lists: List[ListNode]) -> ListNode:
    # Define a custom comparator for 'ListNode', enabling the min-heap to prioritize
    # nodes with smaller values.
    ListNode.__lt__ = lambda self, other: self.val < other.val
    heap = []
    # Push the head of each linked list into the heap.
    for head in lists:
        if head:
            heapq.heappush(heap, head)
    # Set a dummy node to point to the head of the output linked list.
    dummy = ListNode(-1)
    # Create a pointer to iterate through the combined linked list as we add nodes to
    # it.
    curr = dummy
    while heap:
        # Pop the node with the smallest value from the heap and add it to the output
        # linked list.
        smallest_node = heapq.heappop(heap)
        curr.next = smallest_node
        curr = curr.next
        # Push the popped node's subsequent node to the heap.
        if smallest_node.next:
            heapq.heappush(heap, smallest_node.next)
    return dummy.next
```

<h2 class="bbg-h2" style="color: #e5edf7; border-left: 5px solid #39ff88; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Complexity Analysis</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

**Time Complexity**: The **Time Complexity** of combine_sorted_linked_lists is $\mathcal{O}(n\log(k))$, where n denotes the total number of nodes across the linked lists. Here’s why:

- It takes $\mathcal{O}(k\log(k))$ time to create the **Heap** initially because we insert k nodes into the **Heap**.

- Then, for all n nodes, we perform a push and pop operation on the **Heap**, each taking $\mathcal{O}(\log(k))$ time.

This results in a total **Time Complexity** of $\mathcal{O}(k\log(k))$+n$\cdot$ $\mathcal{O}(\log(k))$=$\mathcal{O}(n\log(k))$. **Space Complexity**: The **Space Complexity** is $\mathcal{O}(k)$ because the **Heap** stores up to one node from each of the k linked lists at any time.
