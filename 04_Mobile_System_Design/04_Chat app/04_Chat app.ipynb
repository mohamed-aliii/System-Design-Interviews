<p>In this chapter, we&#x27;ll tackle the design of a chat app. Chat apps are so common that designing one has become a very common question in mobile system design interviews. Figure 1 showcases some of the most popular chat applications currently in the market.</p>
<p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-4-1-GAXQKZQC.svg" alt="Figure 1: Most popular chat apps" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;" /></p>
<p>Chat apps have evolved to meet a wide range of communication needs. They serve as platforms to host one-on-one conversations, help friend groups stay connected, allow families to share life updates and photos, and even enable gamers to strategize during online matches. With such a diversity of potential functions, it&#x27;s crucial to define clearly the specific requirements for the chat app we&#x27;re designing.</p></section>

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<p>When tackling the design of a chat app, it&#x27;s essential to first get on the same page as your interviewer. Here&#x27;s how that conversation might play out:</p>
<p><strong>Interviewer:</strong> Welcome to today&#x27;s session! We&#x27;ll be designing the mobile client app architecture and backend communication for a chat application.</p>
<p>This back-and-forth helps us build a comprehensive picture of our design requirements. Let&#x27;s take a moment to summarize what we&#x27;ve established so far.</p>
<p><strong>💡 Pro tip!</strong></p><p>When addressing a familiar interview question, take the lead by proposing both functional and non-functional requirements upfront. This proactive approach not only showcases your leadership skills but also helps your interview stand out.</p>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">Requirements</span></div>

<p>Based on our discussions, we&#x27;re designing a chat system with the following <strong>functional requirements</strong>:</p>
<ul><li>Users can view their recent conversations, organized by the timestamp of the latest message.</li><li>Users can chat in real time, including both sending and receiving text messages.</li><li>During conversations, users can monitor their contacts&#x27; online status, typing indicators, message delivery status, and read receipts.</li><li>Users can browse their chat history without an internet connection. The conversation history loads efficiently through pagination.</li><li>The app supports push notifications to alert users of new messages.</li></ul>
<p>For our <strong>non-functional requirements</strong>, we need to build a system that ensures:</p>
<ul><li>Scalability: Our system must handle 30 million daily active users worldwide and deliver reliable performance across various network conditions.</li><li>Data integrity: The app should maintain consistency during offline use, especially when storing messages that will be delivered once connectivity returns.</li><li>Durability: Client-side data must remain reliable since it serves as the source of truth for messages.</li></ul>
<p><strong>💡 Pro tip!</strong></p><p>When asking about Daily Active Users (DAU), you&#x27;re doing more than just getting a number. The scale and geographic distribution of users directly influence your technical decisions. High DAU numbers often signal a need to focus more on topics such as performance optimization, resource efficiency, scalable architecture patterns, and the right network protocol choice.</p><p>Always connect scale requirements to your technical choices during the interview. This demonstrates thoughtful system design and shows you understand how business requirements shape architecture decisions.</p>
<p>To keep our design focused, we&#x27;ll consider the following features <strong>out of scope</strong>:</p>
<ul><li>User authentication and registration.</li><li>Group chats and media file sharing.</li><li>Message encryption.</li></ul>
<p>As we move forward, it&#x27;s important to remember that the scale of our system will significantly influence our design decisions. We&#x27;ll need to consider carefully how each choice affects our ability to efficiently manage a large user base and high message volume.</p>

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<p>Now that we&#x27;ve established our requirements, let&#x27;s dive into the API design. This step is crucial for understanding how the client and backend will communicate. In this section, we&#x27;ll look at:</p>
<ul><li>Client and backend expectations.</li><li>Server-initiated connections.</li><li>Network protocols in the chat app.</li><li>Endpoints and data models.</li></ul>
<div style="border-bottom:2px solid rgba(76,175,80,0.35);padding-bottom:8px;margin:28px auto 14px;max-width:52rem;width:96%;box-sizing:border-box"><span style="font-size:1.3em;font-weight:600;color:#4caf50">Client and backend expectations</span></div>

<p>In a chat system, clients don&#x27;t communicate directly with each other. Instead, they connect to a backend that acts as a mediator. When a client sends a message, it goes to the backend. The backend then identifies the intended recipient and forwards the message. If the recipient isn&#x27;t online, the backend holds on to the message until they connect.</p>
<p>Figure 3 illustrates the interactions between clients and the chat backend service, showing how messages flow through the system.</p>
<p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-4-3-PSKY6X5N.svg" alt="Figure 3: Client and server interactions in a chat system" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;" /></p>
<p>Let&#x27;s examine how our client app and backend need to communicate. There are two distinct types of interactions we need to handle:</p>
<ol><li>When clients initiate contact with the backend, such as sending messages or syncing pending ones, HTTP works well for these scenarios since these client-initiated requests follow a simple request-response pattern.</li><li>When the backend needs to reach clients for real-time features, such as message delivery or online status updates, the recipient&#x27;s app needs to know about these events immediately, but HTTP only allows clients to make requests, not the other way around.</li></ol>
<p>While push notifications can help inform users about new messages or events when the app isn&#x27;t active, they aren&#x27;t enough for an active chat session. Real-time features like message delivery, read receipts, and online status updates require something more immediate and reliable than push notifications, which can be slow and inconsistent. Given how sensitive chat apps are to latency, we need a more robust solution that can deliver a seamless, real-time experience.</p>
<div style="border-bottom:2px solid rgba(76,175,80,0.35);padding-bottom:8px;margin:28px auto 14px;max-width:52rem;width:96%;box-sizing:border-box"><span style="font-size:1.3em;font-weight:600;color:#4caf50">Server-initiated connections</span></div>

<p>Over time, developers have created several techniques to support server-initiated connections. These include polling, long polling, Server-Sent Events (SSE), WebSockets, and other solutions built on top of HTTP/2 such as gRPC.</p>
<p>These approaches frequently come up in mobile system design interviews, so it&#x27;s important to understand them. Let&#x27;s explore the key methods that have shaped how mobile apps communicate with servers.</p>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">Polling</span></div>

<p>Polling (or short polling) is a basic technique whereby the client repeatedly asks the server, &quot;Do you have any new messages for me?&quot;. As shown in Figure 4, the process works like this: The client asks for new data; the server responds with or without new information; the client waits briefly, then asks again.</p>
<p>While easy to implement, polling can be wasteful. It often leads to many unnecessary requests, especially when there&#x27;s no new data to report. This constant back-and-forth can burden both the client and the server, consuming extra network resources and processing power.</p>
<p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-4-4-LRNDNYPJ.svg" alt="Figure 4: Polling strategy" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;" /></p>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">Long polling</span></div>

<p>Long polling emerged as an improvement by addressing some of the inefficiencies of traditional polling. Let&#x27;s take a closer look at how it works in Figure 5.</p>
<p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-4-5-3HKUKFVX.svg" alt="Figure 5: Long polling strategy" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;" /></p>
<p>With HTTP long polling, the client asks the server for updates and waits. Unlike short polling, the server holds the connection open until there&#x27;s new data or a timeout occurs. Once the client gets a response, it immediately asks again.</p>
<p>While better than short polling, long polling has some issues too:</p>
<ul><li>It&#x27;s still inefficient for infrequent chatters, as most requests will just time out.</li><li>In multi-server setups, the server with the new message might not be the one connected to the recipient.</li><li>Servers can&#x27;t easily tell if a client has disconnected.</li></ul>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">Server-Sent Events</span></div>

<p>Server-Sent Events (SSE) offer a way for servers to push updates to clients over a single HTTP connection. This approach, illustrated in Figure 6, differs from traditional request–response models.</p>
<p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-4-6-3EHEF7MJ.svg" alt="Figure 6: Server-Sent Events strategy" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;" /></p>
<p>At its core, SSE maintains an open HTTP connection between the client and server. To prevent the connection from timing out due to inactivity, the server may periodically send a colon-prefixed comment line. This simple technique keeps the connection alive, even when there&#x27;s no actual data to transmit.</p>
<p><strong>📝 Note!</strong> This is how SSE works under the hood:</p><ol><li>
The client sends an HTTP GET request to the SSE server endpoint, setting the Accept header to text/event-stream.
</li><li>
The server replies with a 200 OK status and sets the Content-Type to text/event-stream. It then keeps the connection open, ready to push updates.
</li><li>
The server sends events in a simple text format, continuously updating the client.
</li><li>
If the connection drops, the client typically reconnects automatically. To avoid missing events, it can include a Last-Event-ID header in its reconnection request, allowing the server to send only the events that occurred since the last received event.
</li></ol>
<p>SSE is far more efficient than polling methods. By establishing a single, persistent connection and pushing updates only when necessary, SSE significantly reduces network traffic and server load. This makes it a more resource-friendly option for real-time communication.</p>
<p>In the context of our chat app, we could leverage SSE for server-initiated events while using standard HTTP requests for client-initiated actions. This hybrid approach would allow us to receive messages and other events through the SSE connection, while sending user messages via separate HTTP requests.</p>
<p>Although SSE presents a viable solution, there&#x27;s another alternative to consider: WebSockets.</p>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">WebSocket protocol</span></div>

<p>The WebSocket (WS) protocol has one of the most common solutions for real-time, bidirectional communication between clients and servers. Figure 7 shows how it works.</p>
<p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-4-7-4NWI6UF2.svg" alt="Figure 7: WebSocket protocol in action" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;" /></p>
<p>WebSocket connections begin their life as standard HTTP connections that are then &quot;upgraded&quot; to WebSocket connections through a well-defined protocol. This approach allows WebSockets to operate seamlessly even in the presence of firewalls, as they utilize the same ports (80 and 443) as HTTP and HTTPS traffic.</p>
<p>Once established, this two-way channel enables both client and server to freely exchange messages. WebSockets don&#x27;t impose restrictions on data formats; they accommodate everything from plain text to binary data. This flexibility has contributed to their widespread adoption.</p>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">Messaging protocols used in real-world apps</span></div>

<p>WebSockets are built on top of TCP (Transmission Control Protocol) and provide a reliable connection for data transfer. TCP establishes and maintains a connection throughout the session, ensuring data packets are delivered in order and without errors. It handles error-checking, acknowledges received packets, and resends lost ones.</p>
<p>In contrast, UDP (User Datagram Protocol) doesn&#x27;t establish a connection first before sending data. It sends each packet independently, which means it can&#x27;t guarantee delivery, order, or error-free transmission. While this approach lacks TCP&#x27;s reliability, it has less overhead, making it faster for certain applications.</p>
<p>Over the years, developers have created various real-time messaging protocols designed for specific use cases. These protocols are typically built on either TCP or UDP, depending on their needs. Some notable examples include:</p>
<ul><li>WebSocket: This full-duplex protocol enables communication over a single TCP connection.

Slack uses WebSockets for real-time communication [2] [3], and Uber used it as well in the past [4] [5].

</li><li>XMPP (Extensible Messaging and Presence Protocol): An XML-based protocol supporting real-time messaging and presence information.

WhatsApp historically used XMPP as the backbone of its messaging system [6], and Kik Messenger also used it [7] .

</li><li>MQTT (Message Queuing Telemetry Transport): A lightweight publish/subscribe messaging protocol that operates over TCP.

Facebook Messenger&#x27;s initial implementation used MQTT [8].

</li><li>WebRTC (Web Real-Time Communication): This protocol primarily uses UDP (with TCP fallback) to support peer-to-peer audio, video, and data streams.

Discord [9] uses WebRTC for voice and video functionalities.

</li><li>IRC (Internet Relay Chat): Operating on TCP, IRC is used mainly for group communication in channels, but it also supports private messaging.

Mozilla used IRC as the primary tool for community communication [10].

</li></ul>
<div style="border-bottom:2px solid rgba(76,175,80,0.35);padding-bottom:8px;margin:28px auto 14px;max-width:52rem;width:96%;box-sizing:border-box"><span style="font-size:1.3em;font-weight:600;color:#4caf50">Network protocols in the chat app</span></div>

<p>In summary, Table 1 compares the network protocols that can be used for server-initiated connections to meet the system&#x27;s real-time requirements.</p>
<table><thead><tr><th style="text-align: left;">Option</th><th style="text-align: left;">Advantages</th><th style="text-align: left;">Disadvantages</th></tr></thead><tbody><tr><td style="text-align: left;">Short polling</td><td style="text-align: left;">Simple implementation. <br/>Works with standard HTTP infrastructure and wide compatibility across platforms.</td><td style="text-align: left;">Inefficient with excessive requests. <br/>High bandwidth usage occurs when many requests return no new data</td></tr><tr><td style="text-align: left;">Long polling</td><td style="text-align: left;">Near real-time updates with fewer requests than short polling. <br/>Good for non-critical updates.</td><td style="text-align: left;">Complex timeout handling. <br/>Doesn't scale well with many clients as it's resource-intensive on servers.</td></tr><tr><td style="text-align: left;">Server-Sent Events (SSE)</td><td style="text-align: left;">Single open HTTP connection for server-to-client messages. <br/>Built-in reconnection support and works well over HTTP/2.</td><td style="text-align: left;">Unidirectional (server to client only) so it requires a separate mechanism for client-to-server communication. <br/>Resource-intensive with many connections.</td></tr><tr><td style="text-align: left;">WebSocket</td><td style="text-align: left;">True bidirectional communication. <br/>Low-latency message delivery and reduced overhead without repeated request/response cycles.</td><td style="text-align: left;">More complex to implement as it requires more infrastructure support (e.g., load balancers). <br/>Scaling persistent connections is challenging.</td></tr></tbody></table>
<p>Table 1: Trade-offs for choosing the network protocol for real-time features</p>
<p>After evaluating the different options, we chose a hybrid approach for the chat app:</p>
<ol><li><strong>HTTP with REST APIs for client-initiated operations</strong> such as sending messages.</li><li><strong>WebSocket for receiving real-time messages</strong> from the server.</li></ol>
<p>While implementing two protocols adds some complexity, this creates a more maintainable system as the app grows. By using WebSocket only for real-time features rather than all communication, including sending messages, we avoid common scaling issues that typically emerge in production environments.</p>
<p><strong>⚠️ Careful!</strong></p><p>Scaling WebSocket services in production requires careful consideration. Some key challenges include:</p><ul><li>
Managing persistent connections consumes significant server resources.
</li><li>
Load balancing WebSocket connections is more complex than standard HTTP traffic.
</li><li>
Connection recovery and state management need thorough handling.
</li><li>
Monitoring and debugging WebSocket services require specialized tooling.
</li></ul><p>By limiting WebSocket usage to just real-time features rather than all communication, we can better manage these scaling challenges.</p>
<p><strong>We chose WebSockets over SSE for our chat application</strong> primarily because they offer better performance in chat scenarios. WebSockets provide lower latency and reduced header overhead, which is crucial when sending frequent small messages such as typing indicators and read receipts. While SSE could handle one-way server updates, WebSockets have more consistent support across mobile platforms and come with mature tooling for managing connections. This includes built-in capabilities for reconnection strategies and keep-alive mechanisms.</p>
<p>This hybrid approach, which combines HTTP and WebSockets, has been widely adopted in the industry. For example, Slack uses HTTP for most write operations through their Web API [11], while leveraging WebSocket for real-time events through their Real-Time Messaging (RTM) API [12]. Discord follows a similar pattern [13], demonstrating how well this architecture works at scale.</p>
<p>The client establishes the WebSocket connection for real-time updates when the app starts, and it remains active until either the client or server terminates it. <p>Building on the foundation laid in previous chapters, we choose a <strong>layered architecture</strong> that follows Unidirectional Data Flow (UDF), reactive programming, and other key principles, such as separation of concerns, to ensure our app is robust, performant, and scalable.</p>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">Data layer</span></div>

<p>The Message is the central type of application data in the chat app. To handle this effectively, we&#x27;ve incorporated various data layer components, with the <strong>Messages Repository</strong> at the core, managing all read and write operations. <p>The Messages Repository works with both local and remote data sources. While client-initiated requests flow through traditional HTTP endpoints via the <strong>Messages Local DataSource</strong> component, the dedicated <strong>Real time updates Remote DataSource</strong> component maintains a persistent WebSocket connection to receive real-time updates from the backend. This dual approach ensures that the system stays responsive to both user actions and server events.</p>
<p>The <strong>Push Notifications Client component</strong> serves as a listener for notifications from the external provider.</p>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">UI layer</span></div>

<p>The UI layer is built around the app&#x27;s main screens: the <strong>Chat List and Conversation</strong>. To optimize performance, we have centralized image loading into a single <strong>Image Loader</strong> component, which fetches data from the content delivery network (CDN).</p>
<p>The <strong>Message List</strong> is a key component. It displays all messages of a conversation in chronological order, including those that failed to send or are still pending. Failed messages are indicated on the UI, and a resend option is provided for each message.</p>
<div style="border-bottom:2px solid rgba(76,175,80,0.35);padding-bottom:8px;margin:28px auto 14px;max-width:52rem;width:96%;box-sizing:border-box"><span style="font-size:1.3em;font-weight:600;color:#4caf50">Data storage</span></div>

<p>Let&#x27;s explore the best way to store messages in the chat app. While mobile platforms offer simple storage options such as SharedPreferences on Android and UserDefaults on iOS, these key-value stores aren&#x27;t the right fit for our needs. They work well for small pieces of data such as user preferences, but they weren&#x27;t designed to handle growing message histories or the complex relationships between messages, conversations, and users. As message volume increases, these simple storage solutions would slow down and make it difficult to search through conversations effectively.</p>
<p>A <strong>relational database</strong> is a good choice for our chat app. Messages are naturally structured data: each message belongs to a conversation, has an author, and follows a clear format.</p>
<p><strong>📝 Note:</strong> For a deeper dive into comparing different storage alternatives, refer to the Data Storage section in Chapter 10: Mobile System Design Building Blocks.</p>
<p>Relational databases excel at managing these relationships while providing three crucial benefits:</p>
<ol><li>They handle complex queries efficiently. When users search their message history or filter conversations, the database can quickly find and return the right messages, even as the history grows large.</li><li>They ensure data consistency and provide ACID properties (Atomicity, Consistency, Isolation, and Durability). When multiple messages are being sent and received simultaneously, the database maintains data integrity and prevents conflicts. This reliability is essential for a chat application.</li><li>They support future growth. As we add features such as message reactions or thread replies, we can extend our database schema without disrupting existing messages. This flexibility lets our app evolve while keeping all message history intact.</li></ol>
<p>These capabilities make relational databases ideal for handling the complex, growing, and interconnected nature of chat message data. While they may require more initial setup than simpler solutions, they provide the robust foundation we need for a reliable chat application.</p>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">Eviction policy</span></div>

<p>Managing local storage is crucial for our chat app since all conversations are stored on the device. While we could automatically manage storage through eviction policies or time-based deletion, putting users in control often provides a better experience. Popular messaging apps take different approaches: WhatsApp alerts users when storage runs low and guides them through freeing up space [14], while Slack automatically removes messages older than 90 days or a year for their free tier[15].</p>
<p>Similarly, we could implement one of the following options:</p>
<ul><li>Implement eviction policies that remove messages and files based on specific criteria such as message date or attachment size.</li><li>Set up automatic deletion after a defined time period.</li><li>Let users manage their own storage by showing them usage stats and cleanup options.</li></ul>
<p><strong>💡 Pro tip!</strong></p><p>When designing mobile systems, certain decisions, such as how to handle message storage under memory constraints, often involve product teams. While you can propose different approaches and analyze their trade-offs during the interview, it&#x27;s wise to first check with you interviewer on whether these product-level decisions are within scope before diving into implementation details.</p>
<p>For our chat app, we can suggest a user-centric storage management strategy if the interviewer is okay with that decision. Our strategy works as follows:</p>
<ol><li>Monitor available device storage through periodic checks.</li><li>Proactively alert users when space runs low, offering clear options to delete older messages, remove entire conversations, or clean up other app data.</li></ol>
<p>By giving users visibility and control over storage management, we help them make informed decisions about their chat history while preventing the app from consuming too much space.</p>

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<p>Now that we&#x27;ve established the client-backend communication protocol and outlined the client architecture, it&#x27;s time to dive deeper into some key areas. We&#x27;ll concentrate on the following five critical areas:</p>
<ul><li>Message ordering on the screen.</li><li>Sending messages.</li><li>DB implementation details.</li><li>Push notifications.</li></ul>
<div style="border-bottom:2px solid rgba(76,175,80,0.35);padding-bottom:8px;margin:28px auto 14px;max-width:52rem;width:96%;box-sizing:border-box"><span style="font-size:1.3em;font-weight:600;color:#4caf50">Message ordering on the screen</span></div>

<p>Message ordering is fundamental to chat apps. It&#x27;s what makes conversations feel natural and easy to follow. When messages appear in the wrong order, users can quickly become confused and frustrated. What seems like a simple requirement, showing messages in chronological order, actually involves several challenging edge cases on mobile devices:</p>
<ul><li>Network delays can cause messages to arrive out of sequence.</li><li>Offline messages need to be integrated once users reconnect.</li><li>Different devices may have mismatched system clocks.</li><li>Both past messages and new updates must maintain consistent ordering.</li></ul>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">Message lifecycle</span></div>

<p>Let&#x27;s explore how messages flow through our system to make a better decision on screen message ordering. Figure 10 illustrates this process:</p>
<ol><li>The Messages repository immediately creates and stores new messages, and the UI displays them on screen right away to provide instant feedback.</li><li>Behind the scenes, the repository forwards this message to the Messages remote data source, which packages it as a NewMessageRequest and sends it to the backend via HTTP.</li><li>The backend processes this request, generating a unique messageId for the message.</li><li>The backend then takes two parallel actions: (step 4.1) sends a 201 response back to the sender with a NewMessageResponse object, and (step 4.2) notifies the recipient, either through a WebSocket if they are active, or a push notification if they are inactive.</li><li>Once the sender&#x27;s client receives the HTTP response, it updates the message in local storage with the new messageId. Lastly, the UI refreshes to show successful delivery.</li></ol>
<p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-4-10-6VVXVSU6.svg" alt="Figure 10: Message lifecycle in the chat system" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;" /></p>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">Message ordering</span></div>

<p>In steps 1 and 5 of Figure 10, the client needs to display the message on the screen. To implement reliable message ordering in our client app, we need to make a key design decision: what field should we use to sort messages on the conversation screen?</p>
<p>The Message data model provides two potential fields for ordering: the messageId assigned by the server when the message is processed and the createdAt timestamp set by the client when the message is created. Table 2 explores trade-offs between different approaches.</p>
<table><thead><tr><th style="text-align: left;">Trade-offs for ordering messages in the client</th><th style="text-align: left;"></th><th style="text-align: left;"></th></tr></thead><tbody><tr><td style="text-align: left;"><strong>Order by</strong></td><td style="text-align: left;"><strong>Advantages</strong></td><td style="text-align: left;"><strong>Disadvantages</strong></td></tr><tr><td style="text-align: left;">messageId</td><td style="text-align: left;">A server-generated ID ensures consistent ordering across clients. <br/>Backend remains the source of truth.</td><td style="text-align: left;">The client doesn't have the messageId until the backend generates it. <br/>Doesn't support offline messaging effectively.</td></tr><tr><td style="text-align: left;">createdAt timestamp</td><td style="text-align: left;">Low latency. Immediate message display without waiting for the backend. <br/>Supports offline messaging.</td><td style="text-align: left;">Vulnerable to inconsistencies from device clock differences. <br/>Susceptible to client-side manipulation.</td></tr><tr><td style="text-align: left;">createdAt timestamp with client-backend clock synchronization</td><td style="text-align: left;">Combines timestamp benefits with improved reliability. <br/>Reduces ordering issues through synchronization, for example, via WebSocket events or Meta's Simple Precision Time Protocol [16].</td><td style="text-align: left;">Clock drift between devices can still cause problems. <br/>Complex to implement accurately across diverse networks.</td></tr></tbody></table>
<p>Table 2: Trade-offs for ordering messages in the client</p>
<p>While synchronizing client and backend clocks could work, it would add significant complexity and cost to our system. Instead, we can mitigate the downsides of messageId and createdAt approaches by <strong>combining</strong> them:</p>
<ul><li>Messages that have been successfully sent to and acknowledged by the server are ordered by messageId.</li><li>Messages that are still waiting to be sent or are in the process of being sent are ordered by createdAt timestamp and always appear at the end of the message list.</li></ul>
<p>Most popular chat apps use a server-assigned ID to sort messages on the client side. While some apps publish detailed documentation about it, such as Discord&#x27;s Snowflake [17] id approach [18], others, such as WhatsApp with its id field [19], Facebook Messenger with mid [20] or Slack with ts [21], are more opaque about their specifics. How to order pending messages is a product decision, and apps might decide to implement it in different ways.</p>
<p><strong>✅ Decisions made!</strong></p><p>The Conversation screen uses a <strong>hybrid approach</strong> to order messages on the screen:</p><ul><li>
Messages that have been successfully sent to and acknowledged by the server are ordered based on a server-generated messageId.
</li><li>
Messages that are still waiting to be sent or are in progress are ordered based on the client-generated createdAt timestamp and always appear at the end of the message list.
</li></ul>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">messageId responsibilities</span></div>

<p>In our system, the messageId plays a crucial role in maintaining the correct order of messages. They&#x27;re designed to be sortable by time, ensuring that newer messages always have higher IDs than older ones. The backend is the source of truth for generating these IDs to ensure data integrity, reliable delivery, and consistent message ordering.</p>
<p>While this might seem straightforward, creating these IDs correctly can be tricky, especially for a large-scale system. That&#x27;s why we typically delegate this responsibility to the backend. Servers are better suited to quickly and accurately generate unique, time-ordered IDs for millions of messages.</p>
<div style="border-bottom:2px solid rgba(76,175,80,0.35);padding-bottom:8px;margin:28px auto 14px;max-width:52rem;width:96%;box-sizing:border-box"><span style="font-size:1.3em;font-weight:600;color:#4caf50">Sending messages</span></div>

<p>Sending messages reliably in a mobile app isn&#x27;t always straightforward. Network issues can prevent messages from reaching their destination. To handle this challenge, we need a way of tracking different message states: messages not yet sent, messages in transit, and messages that failed to send.</p>
<p>The data layer can model those states with the following MessageRequest data model. Just before sending the message, the remote data source maps this into a NewMessageRequest object. This approach helps us handle various message scenarios and implement retry mechanisms if needed.</p>

<p>The status field tracks the message&#x27;s current state:</p>
<ul><li>DRAFT: The message is created but not yet sent by the client.</li><li>PENDING: The message is queued and ready to be sent to the backend.</li><li>SENT: The client has delivered the message to the backend via a send_messages event, but it hasn&#x27;t received a delivery acknowledgment as part of the messages event.</li><li>FAILED: Either the server reported a failure for this message or the request timed out.</li></ul>
<p><strong>📝 Note:</strong> You might notice there&#x27;s no SUCCEEDED status in our MessageRequest model. This is by design. When the server acknowledges a message&#x27;s delivery, we don&#x27;t simply update the status. Instead, we remove the MessageRequest entirely and create a new Message object with a RECEIVED status. The NewMessageResponse model contains the information we need for this conversion, including the backend-generated messageId and the fromMessageRequestId.</p><p>Similarly, you&#x27;ll never see a Message with a PENDING status in our local data source. The PENDING state is exclusive to message requests. When displaying messages in the UI, we combine both message requests and processed messages.</p>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">Sending messages data flow</span></div>

Removes the temporary message request from storage.

<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">Message delivery failures</span></div>

<p>While our previous example showed a successful message delivery, real-world scenarios aren&#x27;t always smooth. Sometimes, we might never receive a response for a particular NewMessageRequest. This can happen due to unstable network connections or other unforeseen issues.</p>
<p>When a request fails or times out, we don&#x27;t give up immediately. Instead, we update the MessageRequest status to FAILED and increment its failCount. This tracking allows us to decide whether to <strong>keep retrying with exponential backoff</strong> or stop retrying completely after a certain number of failed attempts.</p>
<p><strong>📌 Remember!</strong></p><p>Exponential backoff is a retry strategy whereby the delay between attempts is progressively increased. For instance, if we start with a one-second delay, the next retry might wait two seconds, then four seconds, and so on. This approach avoids overwhelming our system with rapid-fire retries and gives the network a chance to recover.</p>
<p>We don&#x27;t want to retry forever, so we set a <strong>maximum retry limit</strong>. If a message fails too many times, we automatically stop trying, mark it as permanently failed, and notify the user, offering them options to retry manually or delete the message.</p>
<p><strong>✅ Decision made!</strong> We use an <strong>exponential backoff algorithm</strong> to retry failed messages.</p>

<p>When it comes to implementing our database, SQLite stands out as an excellent choice for mobile devices. It&#x27;s a lightweight yet powerful relational database system that can be easily embedded into mobile operating systems. SQLite&#x27;s support for complex queries and transactions makes it well-suited for our chat application&#x27;s needs.</p>
<p><strong>🛠️ Platform implementation details</strong></p><p>Apple&#x27;s CoreData can use SQLite as its underlying storage mechanism. On Android, the most popular choices are Google&#x27;s Jetpack Room [26] and Square&#x27;s SQLDelight [27].</p>

This separation is important because unsynced requests don&#x27;t have a Message id yet.</p>

<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">Challenges</span></div>

<p>Storing messages locally on a device for a chat application sounds straightforward, but it comes with its own set of hurdles. Let&#x27;s explore these challenges and discuss practical solutions:</p>
<p>When working with multiple SQL statements, whether they&#x27;re included in the same or different SQL script files, the execution order matters. The Users table must be created before other tables since they depend on it. Keeping all these CREATE TABLE statements in a single SQL script file helps prevent execution order issues and makes the database setup more reliable.</p>
<p>For simplicity, consistency, and data integrity, we&#x27;ve chosen to consolidate all three tables (Users, Messages, and MessageRequests) within a single SQLite database file named chats.db.</p>
<p>While separating Users into its own database may promote modularity or support future cross-app reuse, this approach introduces several complications:</p>
<ul><li>Foreign key constraints across separate database files are not enforceable, so referential integrity must be maintained manually in the application layer.</li><li>Managing multiple databases in the same app requires careful orchestration, especially when setting up the database in different environments (development, testing, production).</li><li>Separate databases increase the likelihood of incorrect execution order, which can result in schema setup failures. Debugging these issues can be time-consuming and frustrating.</li></ul>
<p>By <strong>keeping all tables in a single chats.db database file</strong>, we&#x27;ve opted for simplicity and reliability in our current scenario. This choice allows us to focus on other aspects of the system design without introducing unnecessary complexity in our database setup.</p>
<p><strong>✅ Decisions made!</strong></p><ul><li>
The Users, Messages, and MessageRequests tables are kept in the same chats.db database.
</li><li>
CREATE TABLE statements are part of the same SQL script file.
</li></ul>
<p>As your chat app grows, the database can become large and slow down app performance. Both Android and iOS have limits on efficient data storage and access.</p>
<p>To address these performance challenges, we could:</p>
<ul><li>Split large tables into smaller ones. For example, divide the Messages and MessageRequests table by user IDs, where each logged-in user&#x27;s messages are stored in their own database file (e.g., chats-currentUserId.db). This helps manage database size and improve query performance for the current user&#x27;s data.</li><li>Add indexes to frequently searched columns, such as to_user_id. But be careful, while indexing speeds up searches, it can slow down adding new data.</li><li>Batch transactions by grouping multiple database writes into a single transaction instead of performing them individually. This reduces disk I/O operations and enhances throughput.</li><li>Set up proper error handling with retry mechanisms for temporary failures, especially during high load periods.</li><li>Use a connection pool to manage concurrent database access efficiently in case multiple threads access the database simultaneously. This prevents bottlenecks and ensures efficient resource usage.</li></ul>
<p><strong>📌 Remember!</strong> Mobile clients should perform all database operations on a background thread to prevent blocking the UI thread and ensure a responsive user experience.</p>

<p>Our design requirements include implementing push notifications to alert users about new messages when they are offline. We&#x27;ve already outlined the key components for push notifications in our high-level architecture diagram. Figure 14 illustrates the various elements involved in the push notification system. Let&#x27;s dive deeper into how these components interact and explore the platform-specific approaches for Android and iOS.</p>
<p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-4-14-NN2UYYQM.svg" alt="Figure 14: Push notifications–related components in the chat system design" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;" /></p>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">Sending push notifications from the backend</span></div>

<p>The backend triggers a push notification when a message is sent to an inactive user, that is, one without an active WebSocket connection to our servers. To deliver these notifications, the backend sends HTTPS requests to an external Push Notifications Provider. This provider is Firebase Cloud Messaging (FCM) for Android devices and Apple Push Notification Service (APNs) for iOS.</p>
<p>To correctly route notifications, the backend maintains a mapping between users and their devices. This mapping is created during the app installation process, where each device registers for push notifications in the background. The backend then stores a unique token for each user&#x27;s device, allowing for targeted notification delivery.</p>
<p><strong>📌 Remember!</strong></p><p>There are many types of push notifications you can send from the backend, for example, alert notifications that immediately alert users; silent notifications that are processed in the background without letting the user know; and rich media notifications that can include images, videos, or other media.</p><p>In our case, we send alert notifications to keep the user informed regarding how many unread messages they have and the message content.</p>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">Receiving push notifications</span></div>

<p>Android and iOS devices maintain a low-power, persistent TLS-encrypted TCP connection to FCM and APNS servers, respectively. This approach eliminates the need for individual apps to manage their own connections, which would be resource-intensive. When a notification arrives, the device either wakes up the relevant app or displays the notification if the app isn&#x27;t running.</p>
<p><strong>🛠️ Platform implementation details</strong></p><p>On Android and iOS, the system maintains a persistent connection to handle push notifications efficiently. Android uses Google Play Services for this purpose, while iOS relies on a background service called the push notification daemon.</p><p>For Android developers, implementing push notifications involves creating a service that extends FirebaseMessagingService. iOS developers, on the other hand, work with the UNUserNotificationCenterDelegate protocol, typically integrating it into the app&#x27;s UIApplicationDelegate.</p>
<p>The handling of notifications differs based on the app&#x27;s state:</p>
<ul><li>If the app is in the foreground, the Push Notifications Client processes the notification directly and calls the Messages Repository to sync messages.</li><li>If the app is in the background or terminated, the device shows the notification in the system tray. When the user taps on it, the system brings the app to the foreground, allowing the Push Notifications Client to access the notification payload.</li></ul>
<div style="margin:22px auto 12px;max-width:52rem;width:96%;box-sizing:border-box;padding-left:14px;border-left:3px solid rgba(76,175,80,0.55)"><span style="font-size:1.12em;font-weight:600;color:#e5edf7">Challenges with push notifications</span></div>

<p>While push notifications are a powerful feature, they come with several challenges that developers need to navigate:</p>
<ul><li>FCM, APNs, or even the device itself might throttle notifications, especially in power-saving modes where the system may delay delivery, limit network access, or batch requests to conserve battery.</li><li>Apps must obtain user consent to send notifications. If users decline or later opt out through device or app settings, your app loses this communication channel.</li><li>Android and iOS use distinct APIs and workflows for push notifications. This necessitates platform-specific code and can result in inconsistent user experiences due to varying feature support and behaviors.</li><li>Push notifications often need to direct users to specific screens within the app. This requires implementing URL schemes or universal links. Restoring the appropriate app state based on these deep links can be tricky, particularly if the app wasn&#x27;t already running.</li><li>Ensuring compatibility across various devices and OS versions is crucial but challenging. The asynchronous nature of push notifications can make debugging delivery issues particularly complex.</li></ul>
<p>Given these potential pitfalls in message delivery, order, and overall complexity, our chat app uses push notifications primarily as a lightweight alert mechanism to notify users of new messages.</p>
