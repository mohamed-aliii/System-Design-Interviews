<h2 class="bbg-h2" style="color: #e5edf7; border-left: 5px solid #39ff88; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Problem</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

Find the longest chain of consecutive numbers in an array. Two numbers are consecutive if they have a difference of 1.

</div>

<h3 class="bbg-h3" style="color: #38bdf8; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 22px 0 10px;">Example:</h3>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

**Input:** `nums = [1, 6, 2, 5, 8, 7, 10, 3]`

**Output:** `4`

**Explanation:**

- The longest chain of consecutive numbers is 5, 6, 7, 8.



</div>

<h2 class="bbg-h2" style="color: #e5edf7; border-left: 5px solid #39ff88; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Intuition</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

A naive approach to this problem is to sort the array. When all numbers are arranged in ascending order, consecutive numbers will be placed next to each other. This allows us to iterate through the array to identify the longest sequence of consecutive numbers.

</div>

This approach requires sorting, which takes $\mathcal{O}(n\log (n))$ time, where n denotes the length of the array. Let’s see how we could do better.

It’s important to understand that every number in the array can represent the start of some consecutive chain. One approach is to treat each number as the start of a chain and search through the array to identify the rest of its chain.

To do this, we can leverage the fact that for any number num, its next consecutive number will be num + 1. This means we’ll always know which number to look for when trying to find the next number in a sequence.

The code snippet for this approach is provided below:

</div>

<h4 class="bbg-h4" style="color: #f59e0b; font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; letter-spacing: 0; margin: 18px 0 8px;">Brute Force Python</h4>

```python
from typing import List

def longest_chain_of_consecutive_numbers_brute_force(nums: List[int]) -> int:
    if not nums:
        return 0
    longest_chain = 0
    # Look for chains of consecutive numbers that start from each number.
    for num in nums:
        current_num = num
        current_chain = 1
        # Continue to find the next consecutive numbers in the chain.
        while (current_num + 1) in nums:
            current_num += 1
            current_chain += 1
        longest_chain = max(longest_chain, current_chain)
    return longest_chain
```

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

This **Brute Force** approach takes $\mathcal{O}(n^3)$ time because of the nested operations involved:

- The outer for-loop iterates through each element, which takes $\mathcal{O}(n)$ time.

- For each element, the inner while-loop can potentially run up to n iterations if there is a long consecutive sequence starting from the current number.

- For each, while-loop iteration, an $\mathcal{O}(n)$ check is performed to see if the next consecutive number exists in the array.

This is slower than the sorting approach, but we can make a couple of optimizations to improve the **Time Complexity**. Let’s discuss these.

Optimization - hash set To find the next number in a sequence, we perform a linear search through the array. However, by storing all the numbers in a hash set, we can instead query this hash set in **Constant Time** to check if a number exists.

This reduces the **Time Complexity** from $\mathcal{O}(n^3)$ to $\mathcal{O}(n^2)$. Optimization - identifying the start of each chain In the **Brute Force** approach, we treat each number as the start of a chain.

This becomes quite expensive because we perform a linear search for every number to find the rest of its chain:

<h2 class="bbg-h2" style="color: #e5edf7; border-left: 5px solid #39ff88; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Implementation</h2>

<h4 class="bbg-h4" style="color: #f59e0b; font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; letter-spacing: 0; margin: 18px 0 8px;">Python Implementation</h4>

```python
from typing import List

def longest_chain_of_consecutive_numbers(nums: List[int]) -> int:
    if not nums:
        return 0
    num_set = set(nums)
    longest_chain = 0
    for num in num_set:
        # If the current number is the smallest number in its chain, search for
        # the length of its chain.
        if num - 1 not in num_set:
            current_num = num
            current_chain = 1
            # Continue to find the next consecutive numbers in the chain.
            while current_num + 1 in num_set:
                current_num += 1
                current_chain += 1
            longest_chain = max(longest_chain, current_chain)
    return longest_chain
```

<h2 class="bbg-h2" style="color: #e5edf7; border-left: 5px solid #39ff88; font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Complexity Analysis</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

**Time Complexity**: The **Time Complexity** of longest_chain_of_consecutive_numbers is $\mathcal{O}(n)$ because, although there are two loops, the inner loop is only executed when the current number is the start of a chain. This ensures that each chain is iterated through only once in the inner while-loop. Thus, the total number of iterations for both loops combined is $\mathcal{O}(n)$: the outer for-loop runs n times, and the inner while-loop runs a total of n times across all iterations, resulting in a combined **Time Complexity** of $\mathcal{O}(n+n)$=$\mathcal{O}(n)$. **Space Complexity**: The **Space Complexity** is $\mathcal{O}(n)$ since the hash set stores each unique number from the array.
