Given an integer array where each element is at most k positions away from its sorted position, sort the array in a non-decreasing order.

</div>

<h3 class="bbg-h3" style="color: var(--bbg-h3-color, var(--vscode-textLink-foreground, var(--jp-content-link-color, #2563eb))); font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 22px 0 10px;">Example:</h3>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

**Input:** `nums = [5, 1, 9, 4, 7, 10], k = 2`

**Output:** `[1, 4, 5, 7, 9, 10]`

A trivial solution to this problem is to sort the array using a standard sorting algorithm. However, since the input is partially sorted (k-sorted), we should assume there's a faster way to sort the array.

We can think about this problem backward. For any index i, the element that belongs at index i in the sorted array is located within the range [ i - k, i + k].

Below, we visualize how number 7, which is meant to be at index 3 when sorted, correctly falls within the range [3 - k, 3 + k] in the k-sorted array: This is a good start, but we can reduce this range even further. Consider index 0 from the above array.

We know the number which belongs at index 0 when sorted is somewhere in the range [0, 0 + k]:

If we continue this process for the rest of the array, we'll successfully sort the k-sorted array. The main inefficiency with this approach is finding the minimum number in the range [ i, i + k] at each index i.

Linearly searching for it will take $\mathcal{O}(k)$ time at each index. To improve this approach, we'd need a way to efficiently access the minimum value at each of these ranges.

A min-**Heap** would be perfect for this. Min-**Heap** For a min-**Heap** to determine the minimum value within each range [ i, i + k], it will always need to be populated with the values in these ranges as we iterate through the array.

<h4 class="bbg-h4" style="color: var(--bbg-h4-color, var(--vscode-terminal-ansiYellow, var(--jp-warn-color1, #b45309))); font-family: ui-monospace, SFMono-Regular, Menlo, Consolas, monospace; letter-spacing: 0; margin: 18px 0 8px;">Python Implementation</h4>

```python
from typing import List
import heapq

def sort_a_k_sorted_array(nums: List[int], k: int) -> List[int]:
    # Populate a min-heap with the first k + 1 values in 'nums'.
    min_heap = nums[:k+1]
    heapq.heapify(min_heap)
    # Replace elements in the array with the minimum from the heap at each
    # iteration.
    insert_index = 0
    for i in range(k + 1, len(nums)):
        nums[insert_index] = heapq.heappop(min_heap)
        insert_index += 1
        heapq.heappush(min_heap, nums[i])
    # Pop the remaining elements from the heap to finish sorting the array.
    while min_heap:
        nums[insert_index] = heapq.heappop(min_heap)
        insert_index += 1
    return nums
```

<h2 class="bbg-h2" style="color: var(--bbg-title-color, var(--vscode-foreground, var(--jp-content-font-color1, CanvasText))); border-left: 5px solid var(--bbg-accent-green, var(--vscode-charts-green, var(--jp-success-color1, #16a34a))); font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Complexity Analysis</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

**Time Complexity**: The **Time Complexity** of sort_a_k_sorted_array is $\mathcal{O}(n\log(k))$, where n denotes the length of the array. Here's why:

- We perform heapify on a min_heap of size k+1 which takes $\mathcal{O}(k)$ time. Note that k is upper-bounded by n in this operation since the **Heap** won't have more than n values.

- Then, we perform push and pop operations on approximately n-k values using the **Heap**. Since the **Heap** can grow up to a size of k+1, each push and pop operations takes $\mathcal{O}(\log(k))$ time. Therefore, this loop takes $\mathcal{O}(n\log(k))$ time in the **Worst Case**.

- The final while-loop runs in $\mathcal{O}(k\log(k))$ time since we pop k+1 values from the **Heap**. Note that k here is also upper-bounded by n.

Therefore, the overall **Time Complexity** is $\mathcal{O}(k)$+$\mathcal{O}(n\log(k))$+$\mathcal{O}(k\log(k))$=$\mathcal{O}(n\log(k))$. **Space Complexity**: The **Space Complexity** is $\mathcal{O}(k)$ since the **Heap** can grow up to k+1 in size.
