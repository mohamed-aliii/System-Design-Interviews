<li>Write operation per second: 100 million / 24 /3600 = 1160</li>
<li>Read operation: Assuming ratio of read operation to write operation is 10:1, read operation per second: 1160 * 10 = 11,600</li>
<li>Assuming the URL shortener service will run for 10 years, this means we must support 100 million <em> 365 </em> 10 = 365 billion records.</li>
<li>Assume average URL length is 100.</li>
<li>Storage requirement over 10 years: 365 billion * 100 bytes = 36.5 TB</li>
</ul>

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(0,230,118,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #00e676;">API Endpoints</span>
</div>

<p>API endpoints facilitate the communication between clients and servers. We will design the APIs REST-style. If you are unfamiliar with restful API, you can consult external materials, such as the one in the reference material [1]. A URL shortener primary needs two API endpoints.</p>
<ul>
<li>URL shortening. To create a new short URL, a client sends a POST request, which contains one parameter: the original long URL. The API looks like this:  POST api/v1/data/shorten</li>
<li>request parameter: {longUrl: longURLString}</li>
<li>return: shortURL</li>
<li>URL redirecting. To redirect a short URL to the corresponding long URL, a client sends a GET request. The API looks like this:  GET api/v1/shortUrl</li>
<li>return: longURL for HTTP redirection</li>
</ul>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(0,230,118,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #00e676;">URL redirecting</span>
</div>

<p>Figure 1 shows what happens when you enter a tinyurl onto the browser. Once the server receives a tinyurl request, it changes the short URL to the long URL with 301 redirect.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-8-1-GSSWK66W.svg" alt="Figure 1" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>The detailed communication between clients and servers is shown in Figure 2.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-8-3-YHOCRGSP.svg" alt="Figure 2" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>One thing worth discussing here is 301 redirect vs 302 redirect.</p>
<p><strong>301 redirect</strong>. A 301 redirect shows that the requested URL is “permanently” moved to the long URL. Since it is permanently redirected, the browser caches the response, and subsequent requests for the same URL will not be sent to the URL shortening service. Instead, requests are redirected to the long URL server directly.</p>
<p><strong>302 redirect</strong>. A 302 redirect means that the URL is “temporarily” moved to the long URL, meaning that subsequent requests for the same URL will be sent to the URL shortening service first. Then, they are redirected to the long URL server.</p>
<p>Each redirection method has its pros and cons. If the priority is to reduce the server load, using 301 redirect makes sense as only the first request of the same URL is sent to URL shortening servers. However, if analytics is important, 302 redirect is a better choice as it can track click rate and source of the click more easily.</p>
The most intuitive way to implement URL redirecting is to use hash tables. Assuming the hash table stores *<shortURL, longURL>* pairs, URL redirecting can be implemented by the following:
<ul>
<li>Get longURL: longURL = hashTable.get(shortURL)</li>
<li>Once you get the longURL, perform the URL redirect.</li>
</ul>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(0,230,118,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #00e676;">URL shortening</span>
</div>

<p>Let us assume the short URL looks like this: <code>www.tinyurl.com/{hashValue}</code>. To support the URL shortening use case, we must find a hash function <em>fx</em> that maps a long URL to the <em>hashValue</em>, as shown in Figure 3.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-8-4-3R6D55TD.svg" alt="Figure 3" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>The hash function must satisfy the following requirements:</p>
<ul>
<li>Each longURL must be hashed to one hashValue.</li>
<li>Each hashValue can be mapped back to the longURL.</li>
</ul>
<p>Detailed design for the hash function is discussed in deep dive.</p>
</section>

  <span style="font-size: 1.6em; font-weight: 700; color: #e5edf7; line-height: 1.25;">Step 3 - Design deep dive</span>

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<p>Up until now, we have discussed the high-level design of URL shortening and URL redirecting. In this section, we dive deep into the following: data model, hash function, URL shortening and URL redirecting.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(0,230,118,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #00e676;">Data model</span>
</div>

In the high-level design, everything is stored in a hash table. This is a good starting point; however, this approach is not feasible for real-world systems as memory resources are limited and expensive. A better option is to store *<shortURL, longURL><em> mapping in a relational database. Figure 4 shows a simple database table design. The simplified version of the table contains 3 columns: </em>id<em>, </em>shortURL, longURL*.
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(0,230,118,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #00e676;">Hash function</span>
</div>

<p>Hash function is used to hash a long URL to a short URL, also known as <em>hashValue</em>.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(0,230,118,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Hash value length</span>
</div>

<p>The <em>hashValue</em> consists of characters from [0-9, a-z, A-Z], containing 10 + 26 + 26 = 62 possible characters. To figure out the length of <em>hashValue</em>, find the smallest <em>n</em> such that <em>62^n ≥ 365 billion</em>. The system must support up to 365 billion URLs based on the back of the envelope estimation. Table 1 shows the length of <em>hashValue</em> and the corresponding maximal number of URLs it can support.</p>
<table><thead><tr><th><strong>n</strong></th><th><strong>Maximal number of URLs</strong></th></tr></thead><tbody><tr><td>1</td><td>62^1 = 62</td></tr><tr><td>2</td><td>62^2 = 3,844</td></tr><tr><td>3</td><td>62^3 = 238,328</td></tr><tr><td>4</td><td>62^ 4 = 14,776,336</td></tr><tr><td>5</td><td>62^5 = 916,132,832</td></tr><tr><td>6</td><td>62^6 = 56,800,235,584</td></tr><tr><td><strong>7</strong></td><td><strong>62^7 = 3,521,614,606,208 = ~3.5 trillion</strong></td></tr><tr><td>8</td><td>62^8 = 218,340,105,584,896</td></tr></tbody></table>
<p>Table 1</p>
<p>When <em>n = 7, 62 ^ n = ~3.5 trillion</em>, 3.5 trillion is more than enough to hold 365 billion URLs, so the length of <em>hashValue</em> is 7.</p>
<p>We will explore two types of hash functions for a URL shortener. The first one is “hash + collision resolution”, and the second one is “base 62 conversion.” Let us look at them one by one.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(0,230,118,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Hash + collision resolution</span>
</div>

<p>To shorten a long URL, we should implement a hash function that hashes a long URL to a 7-character string. A straightforward solution is to use well-known hash functions like CRC32, MD5, or SHA-1. The following table compares the hash results after applying different hash functions on this URL: <code>https://en.wikipedia.org/wiki/Systems_design</code></p>
<table><thead><tr><td><strong>Hash function</strong></td><td><strong>Hash value (Hexadecimal)</strong></td></tr></thead><tbody><tr><td>CRC32</td><td>5cb54054</td></tr><tr><td>MD5</td><td>5a62509a84df9ee03fe1230b9df8b84e</td></tr><tr><td>SHA-1</td><td>0eeae7916c06853901d9ccbefbfcaf4de57ed85b</td></tr></tbody></table>
<p>Table 2</p>
<p>As shown in Table 2, even the shortest hash value (from CRC32) is too long (more than 7 characters). How can we make it shorter?</p>
<p>The first approach is to collect the first 7 characters of a hash value; however, this method can lead to hash collisions. To resolve hash collisions, we can recursively append a new predefined string until no more collision is discovered. This process is explained in Figure 5.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-8-6-OHTOBFS7.svg" alt="Figure 5" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>This method can eliminate collision; however, it is expensive to query the database to check if a shortURL exists for every request. A technique called bloom filters [2] can improve performance. A bloom filter is a space-efficient probabilistic technique to test if an element is a member of a set. Refer to the reference material [2] for more details.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(0,230,118,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Base 62 conversion</span>
</div>

<p>Base conversion is another approach commonly used for URL shorteners. Base conversion helps to convert the same number between its different number representation systems. Base 62 conversion is used as there are 62 possible characters for <em>hashValue</em>. Let us use an example to explain how the conversion works: convert 1115710 to base 62 representation (1115710 represents 11157 in a base 10 system).</p>
<ul>
<li>From its name, base 62 is a way of using 62 characters for encoding. The mappings are: 0-0, ..., 9-9, 10-a, 11-b, ..., 35-z, 36-A, ..., 61-Z, where ‘a’ stands for 10, ‘Z’ stands for 61, etc.</li>
<li>1115710 = 2 x 622 + 55 x 621 + 59 x 620 = [2, 55, 59] -> [2, T, X] in base 62 representation. Figure 6 shows the conversation process.</li>
<li>Thus, the short URL is <code>https://tinyurl.com/</code> <strong>2TX</strong></li>
</ul>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(0,230,118,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Comparison of the two approaches</span>
</div>

<p>Table 3 shows the differences of the two approaches.</p>
<table><thead><tr><th><strong>Hash + collision resolution</strong></th><th><strong>Base 62 conversion</strong></th></tr></thead><tbody><tr><td>Fixed short URL length.</td><td>Short URL length is not fixed. It goes up with the ID.</td></tr><tr><td>Does not need a unique ID generator.</td><td>This option depends on a unique ID generator.</td></tr><tr><td>Collision is possible and needs to be resolved.</td><td>Collision is not possible because ID is unique.</td></tr><tr><td>It’s not possible to figure out the next available short URL because it doesn’t depend on ID.</td><td>It is easy to figure out what is the next available short URL if ID increments by 1 for a new entry. This can be a security concern.</td></tr></tbody></table>
<p>Table 3</p>
</section>

<div style="border-bottom: 2px solid rgba(0,230,118,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #00e676;">URL shortening deep dive</span>
</div>

<p>As one of the core pieces of the system, we want the URL shortening flow to be logically simple and functional. Base 62 conversion is used in our design. We build the following diagram (Figure 7) to demonstrate the flow.</p><ul>
<li>longURL is the input.</li>
<li>The system checks if the longURL is in the database.</li>
<li>If it is, it means the longURL was converted to shortURL before. In this case, fetch the shortURL from the database and return it to the client.</li>
<li>If not, the longURL is new. A new unique ID (primary key) Is generated by the unique ID generator.</li>
<li>Convert the ID to shortURL with base 62 conversion.</li>
<li>Create a new database row with the ID, shortURL, and longURL.</li>
</ul>

<li>Unique ID generator returns ID: 2009215674938.</li>
<li>Convert the ID to shortURL using the base 62 conversion. ID (2009215674938) is converted to “zn9edcu”.</li>
<li>Save ID, shortURL, and longURL to the database as shown in Table 4.</li>
</ul>

Figure 8 shows the detailed design of the URL redirecting. As there are more reads than writes, *<shortURL, longURL>* mapping is stored in a cache to improve performance.
<p>The flow of URL redirecting is summarized as follows:</p>
<ul>
<li>A user clicks a short URL link: https://tinyurl.com/zn9edcu</li>
<li>The load balancer forwards the request to web servers.</li>
<li>If a shortURL is already in the cache, return the longURL directly.</li>
<li>If a shortURL is not in the cache, fetch the longURL from the database. If it is not in the database, it is likely a user entered an invalid shortURL.</li>
<li>The longURL is returned to the user.</li>
</ul>
