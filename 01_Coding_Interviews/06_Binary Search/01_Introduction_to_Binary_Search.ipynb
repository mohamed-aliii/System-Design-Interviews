<h2 class="bbg-h2" style="color: var(--bbg-title-color, var(--vscode-foreground, var(--jp-content-font-color1, CanvasText))); border-left: 5px solid var(--bbg-accent-green, var(--vscode-charts-green, var(--jp-success-color1, #16a34a))); font-family: Inter, ui-sans-serif, system-ui, -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; letter-spacing: 0; margin: 28px 0 12px; padding-left: 12px;">Intuition</h2>

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

Let's say you have a standard, hard-copy English **Dictionary**, and you want to find the definition of the word "inheritance." How could we do this? One approach is to flick through it page by page, until reaching the page with "inheritance." But this is inefficient, and realistically, you're more likely to immediately open a page somewhere in the middle.

Let's say you do this and land on a page of words beginning with "M." Realizing that "inheritance" is somewhere to the left of "M" in the alphabet, you open a page somewhere between the letters "A" and "M." You continue this process of checking if "inheritance" is to the right or left of the current page, and narrowing the search down until you find the correct page. With this method, you've significantly reduced the number of pages to be searched, in contrast to linearly turning every page in order.

This is the fundamental concept of **Binary Search**.

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

**Binary Search** is an algorithm that searches for a value in a sorted data set. Even experienced developers can find it quite tricky to implement **Binary Search** correctly because "the devil's in the detail."

- How should the boundary variables left and right be initialized?

- Should we use left < right or left ≤ right as the exit condition in our while-loop?

- How should the boundary variables be updated? Should we choose left = mid, left = mid + 1, right = mid, or right = mid - 1?

This chapter offers clear, intuitive guidance on how to master these challenges, and confidently handle even the trickiest edge case. To begin any **Binary Search** implementation, do the following:

- Define the search space.

- Define the behavior inside the loop for narrowing the search space.

- Choose an exit condition for the while-loop.

- Return the correct value.

Let's break down these steps.

3. Choosing an exit condition

Choosing an exit condition for when the while-loop should terminate can be tricky. Our choices are primarily between left < right and left ≤ right. Both conditions are applicable in different situations.

- When the exit condition is left < right, the while-loop will break when left and right meet. 2

- On the other hand, left ≤ right ends once left has surpassed right.

<div class="bbg-card" style="background: linear-gradient(180deg, #111827, #172033); color: #e5edf7; border: 1px solid #263447; border-radius: 12px; box-shadow: 0 18px 45px rgba(0, 0, 0, 0.28); margin: 12px auto 18px; padding: 18px 40px; box-sizing: border-box; width: 96%;">

The **Time Complexity** of the **Binary Search** is $\mathcal{O}(\log(n))$, where n is the number of values in the search space. The algorithm is logarithmic because, in each iteration of the algorithm, it divides the search space in half until a single value is located, or no value is found. This reduction by half in each step is characteristic of logarithmic behavior.

</div>
