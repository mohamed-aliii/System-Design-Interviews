In this chapter, we will explore the design of a vending machine system that allows users to select and purchase products, dispense items, manage inventory, and process payments. Although real-world vending machines involve hardware components, we’ll focus on modeling the system’s states, data, and core functionality.

Here are the functional requirements based on the conversation:


* **Product selection**: Users should be able to select from a set of products. Each product has a unique product code, description, and price tag.

* **Inventory management**: Products are stored in specific racks within the vending machine. The system keeps track of the inventory level for each product in its respective rack.

* **Payment processing**: The system only accepts cash payments and can calculate change when needed.



Below are the non-functional requirements:


* The user interface must be intuitive, allowing users to complete a purchase with minimal instructions, and error messages should be clear and concise.

* The system must protect against unauthorized access, ensuring only admins can add, remove, or update products.

A use case diagram illustrates how actors (users or the system) interact with the vending machine system.


![Figure 2: Use Case Diagram of a Vending Machine](assets/image-7-2-GYGEWBUJ.svg)


<div style="margin: 22px 0 10px 0; border-left: 3px solid #d4a373; padding-left: 10px;">
  <span style="font-size: 1.15em; font-weight: 600; font-style: italic; color: #d4a373;">User use cases</span>
</div>


* **Insert Money**: The user inserts cash to initiate a purchase.

* **Select Product**: The user chooses a product by entering its unique product code.

* **Receive Product**: The user collects the dispensed product.

* **Receive Change**: The user receives any change if the inserted amount exceeds the product’s price.



<div style="margin: 22px 0 10px 0; border-left: 3px solid #d4a373; padding-left: 10px;">
  <span style="font-size: 1.15em; font-weight: 600; font-style: italic; color: #d4a373;">Admin use cases</span>
</div>


* **Add Product**: The admin adds new products to the inventory.

* **Remove Product**: The admin removes products from the inventory.

* **Update Inventory**: The admin updates the stock levels of existing products.



<div style="margin: 22px 0 10px 0; border-left: 3px solid #d4a373; padding-left: 10px;">
  <span style="font-size: 1.15em; font-weight: 600; font-style: italic; color: #d4a373;">System use cases</span>
</div>


* **Process Payment**: Validates the inserted cash and calculates change.

* **Dispense Product**: Releases the selected product from the appropriate rack.

* **Check Inventory**: Verifies the availability of a product before dispensing.

* **Display Message**: Shows messages or errors to guide the user.



<div style="border-left: 5px solid #d4a373; padding: 12px 0 12px 20px; margin: 35px 0 18px 0; background: linear-gradient(90deg, rgba(212,163,115,0.08) 0%, transparent 60%); border-radius: 4px;">
  <span style="font-size: 1.6em; font-weight: 700;">Identify Core Objects</span>
</div>


Before diving into the design, let’s enumerate the core objects:


* **VendingMachine**: The central entity that coordinates operations and serves as the main entry point (Facade pattern).

* **Product**: Represents items stored in the vending machine with attributes like identifier, price, and description.

* **Rack**: A designated slot holding a single product type with multiple units and a dispenser.

* **InventoryManager**: Keeps track of the inventory level within the vending machine.

* **PaymentProcessor**: Processes payment, tracks balance, and calculates change.



**Design choice**: Products are linked to racks because racks represent the physical storage locations. This aligns with the single responsibility principle.

The Product class represents a basic product with attributes such as product code, description, and price.


![Figure 3: Product class](assets/image-7-3-THTHWJUQ.svg)


**Design choice**: We have not modeled the inventory quantity within the product class. The stock level is constantly changing, so we introduce a separate InventoryManager class to manage stock levels.

The Rack class models a single rack space. Each rack is associated with a single product and can hold multiple units.


![Figure 4: Rack class](assets/image-7-4-CVZ25Q77.svg)


**Design choice**: We kept the Rack class focused on managing inventory count and product information, delegating actions like dispensing to InventoryManager.

The PaymentProcessor class manages payment acceptance, including tracking the current balance and returning change.


![Figure 6: PaymentProcessor class](assets/image-7-6-NV3ECVYC.svg)

The Transaction class acts as a data structure that tracks the current state of a purchase, encapsulating the selected product, the rack, and the total cost.


![Figure 7: Transaction class](assets/image-7-7-SMDS7IDA.svg)

This class serves as the core component using the Facade pattern to provide a single interface to the clients.


![Figure 8: VendingMachine class](assets/image-7-8-YWMJOC5D.svg)


**Design pattern**: The Vending Machine uses the Facade pattern. To prevent it from becoming a “god object,” facades should remain lightweight and delegate tasks to specialized classes.

<div style="border-left: 4px solid #d4a373; border-radius: 4px; overflow: hidden; margin: 15px 0;">

```java
class Product {
    final String productCode;
    final String description;
    final BigDecimal unitPrice;

    public Product(String productCode, String description, BigDecimal unitPrice) {
        this.productCode = productCode;
        this.description = description;
        this.unitPrice = unitPrice;
    }
}
```

</div>


**Implementation choice**: For monetary values, use `BigDecimal` for precision. Avoid `float` or `double` for currency, as they introduce rounding errors.


<div style="border-bottom: 2px solid rgba(212,163,115,0.3); padding-bottom: 8px; margin: 28px 0 14px 0;">
  <span style="font-size: 1.3em; font-weight: 600; color: #d4a373;">InventoryManager and Rack</span>
</div>


These classes work together to manage inventory. We use a `HashMap<String, Rack>` for efficient O(1) lookups by rack code.


<div style="border-left: 4px solid #d4a373; border-radius: 4px; overflow: hidden; margin: 15px 0;">

```java
public class InventoryManager {
    private Map<String, Rack> racks;

    public InventoryManager() {
        racks = new HashMap<>();
    }

    public Product getProductInRack(String rackCode) {
        return racks.get(rackCode).getProduct();
    }

    public void dispenseProductFromRack(Rack rack) {
        if (rack.getProductCount() > 0) {
            rack.setCount(rack.getProductCount() - 1);
        } else {
            throw new IllegalStateException("Cannot dispense product. Rack is empty.");
        }
    }

    public void updateRack(Map<String, Rack> racks) {
        this.racks = racks;
    }

    public Rack getRack(String name) {
        return racks.get(name);
    }
}
```

</div>


<div style="border-left: 4px solid #d4a373; border-radius: 4px; overflow: hidden; margin: 15px 0;">

```java
public class Rack {
    private final String rackCode;
    private final Product product;
    private int count;

    public Rack(final String rackCode, final Product product, final int count) {
        this.rackCode = rackCode;
        this.product = product;
        this.count = count;
    }

    public Product getProduct() { return product; }
    public int getProductCount() { return count; }
}
```

</div>


<div style="border-bottom: 2px solid rgba(212,163,115,0.3); padding-bottom: 8px; margin: 28px 0 14px 0;">
  <span style="font-size: 1.3em; font-weight: 600; color: #d4a373;">PaymentProcessor</span>
</div>


<div style="border-left: 4px solid #d4a373; border-radius: 4px; overflow: hidden; margin: 15px 0;">

```java
public class PaymentProcessor {
    private BigDecimal currentBalance = BigDecimal.ZERO;

    public void addBalance(BigDecimal amount) {
        currentBalance = currentBalance.add(amount);
    }

    public void charge(BigDecimal amount) {
        currentBalance = currentBalance.subtract(amount);
    }

    public BigDecimal returnChange() {
        BigDecimal change = currentBalance;
        currentBalance = BigDecimal.ZERO;
        return change;
    }

    public BigDecimal getCurrentBalance() {
        return currentBalance;
    }
}
```

</div>


<div style="border-bottom: 2px solid rgba(212,163,115,0.3); padding-bottom: 8px; margin: 28px 0 14px 0;">
  <span style="font-size: 1.3em; font-weight: 600; color: #d4a373;">VendingMachine</span>
</div>


The central component that models the vending machine’s behavior and handles user interactions.


<div style="border-left: 4px solid #d4a373; border-radius: 4px; overflow: hidden; margin: 15px 0;">

```java
class VendingMachine {
    private final List<Transaction> transactionHistory;
    private final InventoryManager inventoryManager;
    private final PaymentProcessor paymentProcessor;
    private Transaction currentTransaction;
    private VendingMachineState currentState;

    public VendingMachine() {
        transactionHistory = new ArrayList<>();
        currentTransaction = new Transaction();
        inventoryManager = new InventoryManager();
        paymentProcessor = new PaymentProcessor();
        this.currentState = new NoMoneyInsertedState();
    }

    void setRack(Map<String, Rack> rack) {
        inventoryManager.updateRack(rack);
    }

    void insertMoney(final BigDecimal amount) {
        paymentProcessor.addBalance(amount);
    }

    void chooseProduct(String rackId) {
        final Product product = inventoryManager.getProductInRack(rackId);
        currentTransaction.setRack(inventoryManager.getRack(rackId));
        currentTransaction.setProduct(product);
    }

    Transaction confirmTransaction() throws InvalidTransactionException {
        validateTransaction();
        paymentProcessor.charge(currentTransaction.getProduct().getUnitPrice());
        inventoryManager.dispenseProductFromRack(currentTransaction.getRack());
        currentTransaction.setTotalAmount(paymentProcessor.returnChange());
        transactionHistory.add(currentTransaction);
        Transaction completedTransaction = currentTransaction;
        currentTransaction = new Transaction();
        return completedTransaction;
    }

    private void validateTransaction() throws InvalidTransactionException {
        if (currentTransaction.getProduct() == null) {
            throw new InvalidTransactionException("Invalid product selection");
        } else if (currentTransaction.getRack().getProductCount() == 0) {
            throw new InvalidTransactionException("Insufficient inventory for product.");
        } else if (paymentProcessor.getCurrentBalance()
                .compareTo(currentTransaction.getProduct().getUnitPrice()) < 0) {
            throw new InvalidTransactionException("Insufficient fund");
        }
    }

    public void cancelTransaction() {
        paymentProcessor.returnChange();
        currentTransaction = new Transaction();
    }
}
```

</div>


The purchase flow: `insertMoney` → `chooseProduct` → `confirmTransaction` (validates, charges, dispenses, returns change).

  <span style="font-size: 1.3em; font-weight: 600; color: #d4a373;">Enforcing task sequences</span>
</div>


What if the interviewer asks: “How would you ensure that users insert money before selecting a product?”


To address this, we introduce the **State Pattern**. This pattern models the vending machine’s behavior as a set of well-defined states:


<div style="margin: 22px 0 10px 0; border-left: 3px solid #d4a373; padding-left: 10px;">
  <span style="font-size: 1.15em; font-weight: 600; font-style: italic; color: #d4a373;">NoMoneyInsertedState</span>
</div>


Represents the initial state where no money has been inserted. Users are only allowed to insert money. Transitions to MoneyInsertedState upon successful insertion.


<div style="margin: 22px 0 10px 0; border-left: 3px solid #d4a373; padding-left: 10px;">
  <span style="font-size: 1.15em; font-weight: 600; font-style: italic; color: #d4a373;">MoneyInsertedState</span>
</div>


Represents the state where money has been inserted. Enables product selection while preventing additional money insertion. Transitions to DispenseState upon successful product selection.


<div style="margin: 22px 0 10px 0; border-left: 3px solid #d4a373; padding-left: 10px;">
  <span style="font-size: 1.15em; font-weight: 600; font-style: italic; color: #d4a373;">DispenseState</span>
</div>


Represents the state where the machine is prepared to dispense. Handles dispensing and resets the machine to the initial state after completion.


This approach guarantees the sequence: **Insert Money → Select Product → Dispense Product**.


![Figure 10: VendingMachineState interface](assets/image-7-10-KINQ2NSU.svg)


<div style="border-bottom: 2px solid rgba(212,163,115,0.3); padding-bottom: 8px; margin: 28px 0 14px 0;">
  <span style="font-size: 1.3em; font-weight: 600; color: #d4a373;">Code changes</span>
</div>


<div style="border-left: 4px solid #d4a373; border-radius: 4px; overflow: hidden; margin: 15px 0;">

```java
public interface VendingMachineState {
    void insertMoney(VendingMachine VM, double amount);
    void selectProductByCode(VendingMachine VM, String productCode)
            throws InvalidStateException;
    void dispenseProduct(VendingMachine VM) throws InvalidStateException;
    String getStateDescription();
}
```

</div>


<div style="border-left: 4px solid #d4a373; border-radius: 4px; overflow: hidden; margin: 15px 0;">

```java
public class NoMoneyInsertedState implements VendingMachineState {
    @Override
    public void insertMoney(VendingMachine VM, double amount) {
        VM.addBalance(amount);
        VM.setState(new MoneyInsertedState());
    }

    @Override
    public void selectProductByCode(VendingMachine VM, String productCode)
            throws InvalidStateException {
        throw new InvalidStateException("Cannot select a product without inserting money.");
    }

    @Override
    public void dispenseProduct(VendingMachine VM) throws InvalidStateException {
        throw new InvalidStateException("Cannot dispense product without inserting money.");
    }

    @Override
    public String getStateDescription() {
        return "No Money Inserted State - Please insert money to proceed";
    }
}
```

</div>


The MoneyInsertedState and DispenseState classes follow the same structure, each enforcing their respective valid actions.

The State pattern is a behavioral pattern that allows an object to alter its behavior when its internal state changes.


In the vending machine design, we use the State pattern to manage states like NoMoneyInsertedState, MoneyInsertedState, and DispenseState.


<div style="margin: 22px 0 10px 0; border-left: 3px solid #d4a373; padding-left: 10px;">
  <span style="font-size: 1.15em; font-weight: 600; font-style: italic; color: #d4a373;">Problem</span>
</div>


Imagine we have a TrafficLight class with three states: Red, Yellow, or Green. Using conditionals becomes unscalable and hard to maintain as states increase.


<div style="margin: 22px 0 10px 0; border-left: 3px solid #d4a373; padding-left: 10px;">
  <span style="font-size: 1.15em; font-weight: 600; font-style: italic; color: #d4a373;">Solution</span>
</div>


Instead of relying on conditionals, we encapsulate the behavior associated with each state into separate classes. The context holds a reference to the current state object and delegates state-related tasks to it.
