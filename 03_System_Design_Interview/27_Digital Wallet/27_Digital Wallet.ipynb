The two updates need to be in a single atomic transaction.</div>
      </div>
    </div>
  </div>
</div>

<li>Support 1,000,000 TPS.</li>
<li>Reliability is at least 99.99%.</li>
<li>Support transactions.</li>

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Distributed transaction: two-phase commit</span>
</div>

<p>In a distributed system, a transaction may involve multiple processes on multiple nodes. To make a transaction atomic, the distributed transaction might be the answer. There are two ways to implement a distributed transaction: a low-level solution and a high-level solution. We will examine each of them.</p>
<p>The low-level solution relies on the database itself. The most commonly used algorithm is called two-phase commit (2PC). As the name implies, it has two phases, as in Figure 5.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-5-two-phase-commit-W3HB4IJO.svg" alt="Figure 5" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><ul>
<li>The coordinator, which in our case is the wallet service, performs read and write operations on multiple databases as normal. As shown in Figure 5, both databases A and C are locked.</li>
<li>When the application is about to commit the transaction, the coordinator asks all databases to prepare the transaction.</li>
<li>In the second phase, the coordinator collects replies from all databases and performs the following:  If all databases reply with a “yes”, the coordinator asks all databases to commit the transaction they have received.</li>
</ul>
<p>If any database replies with a “no”, the coordinator asks all databases to abort the transaction.</p>
<ul>
<li>If all databases reply with a “yes”, the coordinator asks all databases to commit the transaction they have received.</li>
<li>If any database replies with a “no”, the coordinator asks all databases to abort the transaction.</li>
</ul>
<p>It is a low-level solution because the prepare step requires a special modification to the database transaction. For example, there is an X/Open XA [6] standard that coordinates heterogeneous databases to achieve 2PC. The biggest problem with 2PC is that it’s not performant, as locks can be held for a very long time while waiting for a message from the other nodes. Another issue with 2PC is that the coordinator can be a single point of failure, as shown in Figure 6.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-6-coordinator-crashes-MTPFFNSD.svg" alt="Figure 6" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Distributed transaction: Try-Confirm/Cancel (TC/C)</span>
</div>

<p>TC/C is a type of compensating transaction [7] that has two steps:</p>
<ul>
<li>In the first phase, the coordinator asks all databases to reserve resources for the transaction.</li>
<li>In the second phase, the coordinator collects replies from all databases:  If all databases reply with “yes”, the coordinator asks all databases to confirm the operation, which is the Try-Confirm process.</li>
</ul>
<p>If any database replies with “no”, the coordinator asks all databases to cancel the operation, which is the Try-Cancel process.</p>
<ul>
<li>If all databases reply with “yes”, the coordinator asks all databases to confirm the operation, which is the Try-Confirm process.</li>
<li>If any database replies with “no”, the coordinator asks all databases to cancel the operation, which is the Try-Cancel process.</li>
</ul>
<p>It’s important to note that the two phases in 2PC are wrapped in the same transaction, but in TC/C each phase is a separate transaction.</p>
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">TC/C example</span>
</div>

<p>It would be much easier to explain how TC/C works with a real-world example. Suppose we want to transfer $1 from account A to account C. Table 2 gives a summary of how TC/C is executed in each phase.</p>
<table><thead><tr><th>Phase</th><th>Operation</th><th>A</th><th>C</th></tr></thead><tbody><tr><td>1</td><td>Try</td><td>Balance change: -$1</td><td>Do nothing</td></tr><tr><td>2</td><td>Confirm</td><td>Do nothing</td><td>Balance change: +$1</td></tr><tr><td></td><td>Cancel</td><td>Balance change: +$1</td><td>Do Nothing</td></tr></tbody></table>
<p>Table 2 TC/C example</p>
<p>Let’s assume the wallet service is the coordinator of the TC/C. At the beginning of the distributed transaction, account A has $1 in its balance, and account C has $0.</p>
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">First phase: Try</span>
</div>

<p>In the Try phase, the wallet service, which acts as the coordinator, sends two transaction commands to two databases:</p>
<ul>
<li>For the database that contains account A, the coordinator starts a local transaction that reduces the balance of A by $1.</li>
<li>For the database that contains account C, the coordinator gives it a NOP (no operation.) To make the example adaptable for other scenarios, let’s assume the coordinator sends to this database a NOP command. The database does nothing for NOP commands and always replies to the coordinator with a success message.</li>
</ul>
<p>The Try phase is shown in Figure 7. The thick line indicates that a lock is held by the transaction.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-7-try-phase-PCD3MQ6F.svg" alt="Figure 7" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Second phase: Confirm</span>
</div>

<p>If both databases reply “yes”, the wallet service starts the next Confirm phase.</p>
<p>Account A’s balance has already been updated in the first phase. The wallet service does not need to change its balance. However, account C has not yet received its $1 from account A in the first phase. In the Confirm phase, the wallet service has to add $1 to account C’s balance.</p>
<p>The Confirm process is shown in Figure 8.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-8-confirm-phase-LJRYUDAO.svg" alt="Figure 8" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Second phase: Cancel</span>
</div>

<p>What if the first Try phase fails? In the example above we have assumed the NOP operation on account C always succeeds, although in practice it may fail. For example, account C might be an illegal account, and the regulator has mandated that no money can flow into or out of this account. In this case, the distributed transaction must be canceled and we have to clean up.</p>
<p>Because the balance of account A has already been updated in the transaction in the Try phase, it is impossible for the wallet service to cancel a completed transaction. What it can do is to start another transaction that reverts the effect of the transaction in the Try phase, which is to add $1 back to account A.</p>
<p>Because account C was not updated in the Try phase, the wallet service just needs to send a NOP operation to account C’s database.</p>
<p>The Cancel process is shown in Figure 9.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-9-cancel-phase-4VJNJGDS.svg" alt="Figure 9" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Comparison between 2PC and TC/C</span>
</div>

<p>Table 3 shows that there are many similarities between 2PC and TC/C, but there are also differences. In 2PC, all local transactions are not done (still locked) when the second phase starts, while in TC/C, all local transactions are done (unlocked) when the second phase starts. In other words, the second phase of 2PC is about completing an unfinished transaction, such as an abort or commit, while in TC/C, the second phase is about using a reverse operation to offset the previous transaction result when an error occurs. The following table summarizes their differences.</p>
<table><thead><tr><td></td><td><strong>First Phase</strong></td><td><strong>Second Phase: success</strong></td><td><strong>Second Phase: fail</strong></td></tr></thead><tbody><tr><td>2PC</td><td>Local transactions are not done yet</td><td>Commit all local transactions Cancel all local transactions</td><td>Cancel all local transactions</td></tr><tr><td>TC/C</td><td><p>All local transactions are completed, either committed or canceled</p></td><td>Execute new local transactions if needed</td><td><p>Reverse the side effect of the already committed transaction, or
called “undo”</p></td></tr></tbody></table>
<p>All local transactions are completed, either committed or canceled</p>
<p>Reverse the side effect of the already committed transaction, or called “undo”</p>
<p>Table 3 2PC vs TC/C</p>
<p>TC/C is also called a distributed transaction by compensation. It is a high-level solution because the compensation, also called the “undo,” is implemented in the business logic. The advantage of this approach is that it is database-agnostic. As long as a database supports transactions, TC/C will work. The disadvantage is that we have to manage the details and handle the complexity of the distributed transactions in the business logic at the application layer.</p>
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Phase status table</span>
</div>

<p>We still have not yet answered the question asked earlier; what if the wallet service restarts in the middle of TC/C? When it restarts, all previous operation history might be lost, and the system may not know how to recover.</p>
<p>The solution is simple. We can store the progress of a TC/C as phase status in a transactional database. The phase status includes at least the following information.</p>
<ul>
<li>The ID and content of a distributed transaction.</li>
<li>The status of the Try phase for each database. The status could be “not sent yet”, “has been sent”, and “response received”.</li>
<li>The name of the second phase. It could be “Confirm” or “Cancel.” It could be calculated using the result of the Try phase.</li>
<li>The status of the second phase.</li>
<li>An out-of-order flag (explained soon in the section “Out-of-Order Execution”).</li>
</ul>
<p>Where should we put the phase status tables? Usually, we store the phase status in the database that contains the wallet account from which money is deducted. The updated architecture diagram is shown in Figure 10.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-10-phase-status-table-45KR4FXX.svg" alt="Figure 10" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Unbalanced state</span>
</div>

<p>Have you noticed that by the end of the Try phase, $1 is missing (Figure 11)?</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-11-unbalanced-state-ESMXLPTG.svg" alt="Figure 11" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>Assuming everything goes well, by the end of the Try phase, $1 is deducted from account A and account C remains unchanged. The sum of account balances in A and C will be $0, which is less than at the beginning of the TC/C. It violates a fundamental rule of accounting that the sum should remain the same after a transaction.</p>
<p>The good news is that the transactional guarantee is still maintained by TC/C. TC/C comprises several independent local transactions. Because TC/C is driven by application, the application itself is able to see the intermediate result between these local transactions. On the other hand, the database transaction or 2PC version of the distributed transaction was maintained by databases that are invisible to high-level applications.</p>
<p>There are always data discrepancies during the execution of distributed transactions. The discrepancies might be transparent to us because lower-level systems such as databases already fixed the discrepancies. If not, we have to handle it ourselves (for example, TC/C).</p>
<p>The unbalanced state is shown in Figure 11.</p>
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Valid operation orders</span>
</div>

<p>There are three choices for the Try phase:</p>
<table><thead><tr><th>Try phase choices</th><th>Account A</th><th>Account C</th></tr></thead><tbody><tr><td>Choice 1</td><td>-$1</td><td>NOP</td></tr><tr><td>Choice 2</td><td>NOP</td><td>+$1</td></tr><tr><td>Choice 3</td><td>-$1</td><td>+$1</td></tr></tbody></table>
<p>Table 4 Try phase choices</p>
<p>All three choices look plausible, but some are not valid.</p>
<p>For choice 2, if the Try phase on account C is successful, but has failed on account A (NOP), the wallet service needs to enter the Cancel phase. There is a chance that somebody else may jump in and move the $1 away from account C. Later when the wallet service tries to deduct $1 from account C, it finds nothing is left, which violates the transactional guarantee of a distributed transaction.</p>
<p>For choice 3, if $1 is deducted from account A and added to account C concurrently, it introduces lots of complications. For example, $1 is added to account C, but it fails to deduct the money from account A. What should we do in this case?</p>
<p>Therefore, choice 2 and choice 3 are flawed choices and only choice 1 is valid.</p>
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Out-of-order execution</span>
</div>

<p>One side effect of TC/C is the out-of-order execution. It will be much easier to explain using an example.</p>
<p>We reuse the above example which transfers $1 from account A to account C. As Figure 12 shows, in the Try phase, the operation against account A fails and it returns a failure to the wallet service, which then enters the Cancel phase and sends the cancel operation to both account A and account C.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-12-out-of-order-execution-OSSZ5U7M.svg" alt="Figure 12" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>Let’s assume that the database that handles account C has some network issues and it receives the Cancel instruction before the Try instruction. In this case, there is nothing to cancel.</p>
<p>The out-of-order execution is shown in Figure 12.</p>
<p>To handle out-of-order operations, each node is allowed to Cancel a TC/C without receiving a Try instruction, by enhancing the existing logic with the following updates:</p>
<ul>
<li>The out-of-order Cancel operation leaves a flag in the database indicating that it has seen a Cancel operation, but it has not seen a Try operation yet.</li>
<li>The Try operation is enhanced so it always checks whether there is an out-of-order flag, and it returns a failure if there is.</li>
</ul>
<p>This is why we added an out-of-order flag to the phase status table in the “Phase Status Table” section.</p>
</section>

<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Distributed transaction: Saga</span>
</div>

<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Linear order execution</span>
</div>

<p>There is another popular distributed transaction solution called Saga [8]. Saga is the de-facto standard in a microservice architecture. The idea of Saga is simple:</p>
<ul>
<li>All operations are ordered in a sequence. Each operation is an independent transaction on its own database.</li>
<li>Operations are executed from the first to the last. When one operation has finished, the next operation is triggered.</li>
<li>When an operation has failed, the entire process starts to roll back from the current operation to the first operation in reverse order, using compensating transactions. So if a distributed transaction has n operations, we need to prepare 2n operations: n for the normal case and another n for the compensating transaction during rollback.</li>
</ul>
<p>It is easier to understand this by using an example. Figure 13 shows the Saga workflow to transfer $1 from account A to account C. The top horizontal line shows the normal order of execution. The two vertical lines show what the system should do when there is an error. When it encounters an error, the transfer operations are rolled back and the client receives an error message. As we mentioned in the “Valid operation orders” section, we have to put the deduction operation before the addition operation.</p>

There are two ways to do it:</p>
<ul>
<li>Choreography. In a microservice architecture, all the services involved in the Saga distributed transaction do their jobs by subscribing to other services’ events. So it is fully decentralized coordination.</li>
<li>Orchestration. A single coordinator instructs all services to do their jobs in the correct order.</li>
</ul>
<p>The choice of which coordination model to use is determined by the business needs and goals. The challenge of the choreography solution is that services communicate in a fully asynchronous way, so each service has to maintain an internal state machine in order to understand what to do when other services emit an event. It can become hard to manage when there are many services. The orchestration solution handles complexity well, so it is usually the preferred solution in a digital wallet system.</p>

<p>Which one should we use in practice? The answer depends on the latency requirement. As Table 5 shows, operations in Saga have to be executed in linear order, but it is possible to execute them in parallel in TC/C. So the decision depends on a few factors:</p>
<ul>
<li>If there is no latency requirement, or there are very few services, such as our money transfer example, we can choose either of them. If we want to go with the trend in microservice architecture, choose Saga.</li>
<li>If the system is latency-sensitive and contains many services/operations, TC/C might be a better option.</li>

<div style="border-bottom: 2px solid rgba(38,166,154,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #26a69a;">Event sourcing</span>
</div>

<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Background</span>
</div>

<p>In real life, a digital wallet provider may be audited. These external auditors might ask some challenging questions, for example:</p>
<ul>
<li>Do we know the account balance at any given time?</li>
<li>How do we know the historical and current account balances are correct?</li>
<li>How do we prove that the system logic is correct after a code change?</li>
</ul>
<p>One design philosophy that systematically answers those questions is event sourcing, which is a technique developed in Domain-Driven Design (DDD) [9].</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Definition</span>
</div>

<p>There are four important terms in event sourcing.</p>
<ul>
<li>Command</li>
<li>Event</li>
<li>State</li>
<li>State machine</li>
</ul>
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Command</span>
</div>

<p>A command is the intended action from the outside world. For example, if we want to transfer $1 from client A to client C, this money transfer request is a command.</p>
<p>In event sourcing, it is very important that everything has an order. So commands are usually put into a FIFO (first in, first out) queue.</p>
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Event</span>
</div>

<p>Command is an intention and not a fact because some commands may be invalid and cannot be fulfilled. For example, the transfer operation will fail if the account balance becomes negative after the transfer.</p>
<p>A command must be validated before we do anything about it. Once the command passes the validation, it is valid and must be fulfilled. The result of the fulfillment is called an event.</p>
<p>There are two major differences between command and event.</p>
<ul>
<li>Events must be executed because they represent a validated fact. In practice, we usually use the past tense for an event. If the command is “transfer $1 from A to C”, the corresponding event would be “transferred $1 from A to C”.</li>
<li>Commands may contain randomness or I/O, but events must be deterministic. Events represent historical facts.</li>
</ul>
<p>There are two important properties of the event generation process.</p>
<ul>
<li>One command may generate any number of events. It could generate zero or more events.</li>
<li>Event generation may contain randomness, meaning it is not guaranteed that a command always generates the same event(s). The event generation may contain external I/O or random numbers. We will revisit this property in more detail near the end of the chapter.</li>
</ul>
<p>The order of events must follow the order of commands. So events are stored in a FIFO queue, as well.</p>
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">State</span>
</div>

<p>State is what will be changed when an event is applied. In the wallet system, state is the balances of all client accounts, which can be represented with a map data structure. The key is the account name or ID, and the value is the account balance. Key-value stores are usually used to store the map data structure. The relational database can also be viewed as a key-value store, where keys are primary keys and values are table rows.</p>
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">State machine</span>
</div>

<p>A state machine drives the event sourcing process. It has two major functions.</p>
<ul>
<li>Validate commands and generate events.</li>
<li>Apply event to update state.</li>
</ul>
<p>Event sourcing requires the behavior of the state machine to be deterministic. Therefore, the state machine itself should never contain any randomness. For example, it should never read anything random from the outside using I/O, or use any random numbers. When it applies an event to a state, it should always generate the same result.</p>
<p>Figure 14 shows the static view of event sourcing architecture. The state machine is responsible for converting the command to an event and for applying the event. Because state machine has two primary functions, we usually draw two state machines, one for validating commands and the other for applying events.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-14-static-view-of-event-sourcing-3BMIV7QG.svg" alt="Figure 14" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>If we add the time dimension, Figure 15 shows the dynamic view of event sourcing. The system keeps receiving commands and processing them, one by one.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-15-dynamic-view-of-event-sourcing-2C2OZVDV.svg" alt="Figure 15" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Wallet service example</span>
</div>

<p>For the wallet service, the commands are balance transfer requests. These commands are put into a FIFO queue. One popular choice for the command queue is Kafka [10]. The command queue is shown in Figure 16.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-16-command-queue-WXJN4Y22.svg" alt="Figure 16" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>Let us assume the state (the account balance) is stored in a relational database. The state machine examines each command one by one in FIFO order. For each command, it checks whether the account has a sufficient balance. If yes, the state machine generates an event for each account. For example, if the command is “A->$1->C”, the state machine generates two events: “A:-$1” and “C:+$1”.</p>
<p>Figure 17 shows how the state machine works in 5 steps.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-17-how-state-machine-works-67HWC4DZ.svg" alt="Figure 17" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><ul>
<li>Read commands from the command queue.</li>
<li>Read balance state from the database.</li>
<li>Validate the command. If it is valid, generate two events for each of the accounts.</li>
<li>Read the next Event.</li>
<li>Apply the Event by updating the balance in the database.</li>
</ul>
</section>

<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Reproducibility</span>
</div>

<p>The most important advantage that event sourcing has over other architectures is reproducibility.</p>
<p>In the distributed transaction solutions mentioned earlier, a wallet service saves the updated account balance (the state) into the database. It is difficult to know why the account balance was changed. Meanwhile, historical balance information is lost during the update operation. In the event sourcing design, all changes are saved first as immutable history. The database is only used as an updated view of what balance looks like at any given point in time.</p>
<p>We could always reconstruct historical balance states by replaying the events from the very beginning. Because the event list is immutable and the state machine logic is deterministic, it is guaranteed that the historical states generated from each replay are the same.</p>
<p>Figure 18 shows how to reproduce the states of the wallet service by replaying the events.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-18-reproduce-states-VHMK3NBY.svg" alt="Figure 18" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>Reproducibility helps us answer the difficult questions that the auditors ask at the beginning of the section. We repeat the questions here.</p>
<ul>
<li>Do we know the account balance at any given time?</li>
<li>How do we know the historical and current account balances are correct?</li>
<li>How do we prove the system logic is correct after a code change?</li>
</ul>
<p>For the first question, we could answer it by replaying events from the start, up to the point in time where we would like to know the account balance.</p>
<p>For the second question, we could verify the correctness of the account balance by recalculating it from the event list.</p>
<p>For the third question, we can run different versions of the code against the events and verify that their results are identical.</p>
<p>Because of the audit capability, event sourcing is often chosen as the de facto solution for the wallet service.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Command-query responsibility segregation (CQRS)</span>
</div>

<p>So far, we have designed the wallet service to move money from one account to another efficiently. However, the client still does not know what the account balance is. There needs to be a way to publish state (balance information) so the client, which is outside of the event sourcing framework, can know what the state is.</p>
<p>Intuitively, we can create a read-only copy of the database (historical state) and share it with the outside world. Event sourcing answers this question in a slightly different way.</p>
<p>Rather than publishing the state (balance information), event sourcing publishes all the events. The external world could rebuild any customized state itself. This design philosophy is called CQRS [11].</p>
<p>In CQRS, there is one state machine responsible for the write part of the state, but there can be many read-only state machines, which are responsible for building views of the states. Those views could be used for queries.</p>
<p>These read-only state machines can derive different state representations from the event queue. For example, clients may want to know their balances and a read-only state machine could save state in a database to serve the balance query. Another state machine could build state for a specific time period to help investigate issues like possible double charges. The state information is an audit trail that could help to reconcile the financial records.</p>
<p>The read-only state machines lag behind to some extent, but will always catch up. The architecture design is eventually consistent.</p>

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">

  <span style="font-size: 1.3em; font-weight: 600; color: #26a69a;">High-performance event sourcing</span>
</div>

<p>In the earlier example, we used Kafka as the command and event store, and the database as a state store. Let’s explore some optimizations.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">File-based command and event list</span>
</div>

<p>The first optimization is to save commands and events to a local disk, rather than to a remote store like Kafka. This avoids transit time across the network. The event list uses an append-only data structure. Appending is a sequential write operation, which is generally very fast. It works well even for magnetic hard drives because the operating system is heavily optimized for sequential reads and writes. According to this ACM Queue article [12], sequential disk access can be faster than random memory access in some cases.</p>
<p>The second optimization is to cache recent commands and events in memory. As we explained before, we process commands and events right after they are persisted. We may cache them in memory to save the time of loading them back from the local disk.</p>
<p>We are going to explore some implementation details. A technique called mmap [13] is great for implementing the optimizations mentioned previously. Mmap can write to a local disk and cache recent content in memory at the same time. It maps a disk file to memory as an array. The operating system caches certain sections of the file in memory to accelerate the read and write operations. For append-only file operations, it is almost guaranteed that all data are saved in memory, which is very fast.</p>

  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">File-based state</span>
</div>

<p>In the previous design, state (balance information) is stored in a relational database. In a production environment, a database usually runs in a stand-alone server that can only be accessed through networks. Similar to the optimizations we did for command and event, state information can be saved to the local disk, as well.</p>
<p>More specifically, we can use SQLite [14], which is a file-based local relational database or use RocksDB [15], which is a local file-based key-value store.</p>
<p>RocksDB is chosen because it uses a log-structured merge-tree (LSM), which is optimized for write operations. To improve read performance, the most recent data is cached.</p>

<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Snapshot</span>
</div>

<p>Once everything is file-based, let us consider how to accelerate the reproducibility process. When we first introduced reproducibility, the state machine had to process events from the very beginning, every time. What we could optimize is to periodically stop the state machine and save the current state into a file. This is called a snapshot.</p>
<p>A snapshot is an immutable view of a historical state. Once a snapshot is saved, the state machine does not have to restart from the very beginning anymore. It can read data from a snapshot, verify where it left off, and resume processing from there.</p>
<p>For financial applications such as wallet service, the finance team often requires a snapshot to be taken at 00:00 so they can verify all transactions that happened during that day. When we first introduced CQRS of event sourcing, the solution was to set up a read-only state machine that reads from the beginning until the specified time is met. With snapshots, a read-only state machine only needs to load one snapshot that contains the data.</p>
<p>A snapshot is a giant binary file and a common solution is to save it in an object storage solution, such as HDFS [16].</p>
<p>Figure 22 shows the file-based event sourcing architecture. When everything is file-based, the system can fully utilize the maximum I/O throughput of the computer hardware.</p>

<p>Before we explain the solution, let’s examine the parts of the system that need the reliability guarantee.</p>

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Reliability analysis</span>
</div>

<p>Conceptually, everything a node does is around two concepts; data and computation. As long as data is durable, it’s easy to recover the computational result by running the same code on another node. This means we only need to worry about the reliability of data because if data is lost, it is lost forever. The reliability of the system is mostly about the reliability of the data.</p>
<p>There are four types of data in our system.</p>
<ul>
<li>File-based command</li>
<li>File-based event</li>
<li>File-based state</li>
<li>State snapshot</li>
</ul>
<p>Let us take a close look at how to ensure the reliability of each type of data.</p>
<p>State and snapshot can always be regenerated by replaying the event list. To improve the reliability of state and snapshot, we just need to ensure the event list has strong reliability.</p>
<p>Now let us examine command. On the face of it, event is generated from command. We might think providing a strong reliability guarantee for command should be sufficient. This seems to be correct at first glance, but it misses something important. Event generation is not guaranteed to be deterministic, and also it may contain random factors such as random numbers, external I/O, etc. So command cannot guarantee reproducibility of events.</p>
<p>Now it’s time to take a close look at event. Event represents historical facts that introduce changes to the state (account balance.) Event is immutable and can be used to rebuild the state.</p>
<p>From this analysis, we conclude that event data is the only one that requires a high-reliability guarantee. We will explain how to achieve this in the next section.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Consensus</span>
</div>

<p>To provide high reliability, we need to replicate the event list across multiple nodes. During the replication process, we have to guarantee the following properties.</p>
<ul>
<li>No data loss.</li>
<li>The relative order of data within a log file remains the same across nodes.</li>
</ul>
<p>To achieve those guarantees, consensus-based replication is a good fit. The consensus algorithm makes sure that multiple nodes reach a consensus on what the event list is. Let’s use the Raft [17] consensus algorithm as an example.</p>
<p>The Raft algorithm guarantees that as long as more than half of the nodes are online, the append-only lists on them have the same data. For example, if we have 5 nodes and use the Raft algorithm to synchronize their data, as long as at least 3 (more than half) of the nodes are up as Figure 23 shows, the system can still work properly as a whole:</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-23-raft-YUKRLGJR.svg" alt="Figure 23" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>A node can have three different roles in the Raft algorithm.</p>
<ul>
<li>Leader</li>
<li>Candidate</li>
<li>Follower</li>
</ul>
<p>We can find the implementation of the Raft algorithm in the Raft paper. We will only cover the high level concepts here and not go into detail. In Raft, at most one node is the leader of the cluster and the remaining nodes are followers. The leader is responsible for receiving external commands and replicating data reliably across nodes in the cluster.</p>
<p>With the Raft algorithm, the system is reliable as long as the majority of the nodes are operational. For example, if there are 3 nodes in the cluster, it could tolerate the failure of 1 node, and if there are 5 nodes, it can tolerate the failure of 2 nodes.</p>
</section>

        <div style="color: #e5edf7; line-height: 1.7; font-size: 0.95em;">Yes, the system is more reliable and fault-tolerant. However, in order to handle 1 million TPS, one server is not enough. How can we make the system more scalable?</div>

<p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-24-raft-node-group-EFJORSSU.svg" alt="Figure 24" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>In Figure 24, we set up 3 event sourcing nodes. These nodes use the Raft algorithm to synchronize the event list reliably.</p>
<p>The leader takes incoming command requests from external users, converts them into events, and appends events into the local event list. The Raft algorithm replicates newly added events to the followers.</p>
<p>All nodes, including the followers, process the event list and update the state. The Raft algorithm ensures the leader and followers have the same event lists, while event sourcing guarantees all states are the same, as long as the event lists are the same.</p>
<p>A reliable system needs to handle failures gracefully, so let’s explore how node crashes are handled.</p>
<p>If the leader crashes, the Raft algorithm automatically selects a new leader from the remaining healthy nodes. This newly elected leader takes responsibility for accepting commands from external users. It is guaranteed that the cluster as a whole can provide continued service when a node goes down.</p>
<p>When the leader crashes, it is possible that the crash happens before the command list is converted to events. In this case, the client would notice the issue either by a timeout or by receiving an error response. The client needs to resend the same command to the newly elected leader.</p>
<p>In contrast, follower crashes are much easier to handle. If a follower crashes, requests sent to it will fail. Raft handles failures by retrying indefinitely until the crashed node is restarted or a new one replaces it.</p>

</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="border-bottom: 2px solid rgba(38,166,154,0.35); padding-bottom: 8px; margin: 28px auto 14px; max-width: 52rem; width: 96%; box-sizing: border-box;">
  <span style="font-size: 1.3em; font-weight: 600; color: #26a69a;">Distributed event sourcing</span>
</div>

<p>In the previous section, we explained how to implement a reliable high-performance event sourcing architecture. It solves the reliability issue, but it has two limitations.</p>
<ul>
<li>When a digital wallet is updated, we want to receive the updated result immediately. But in the CQRS design, the request/response flow can be slow. This is because a client doesn’t know exactly when a digital wallet is updated and the client may need to rely on periodic polling.</li>
<li>The capacity of a single Raft group is limited. At a certain scale, we need to shard the data and implement distributed transactions.</li>
</ul>
<p>Let’s take a look at how those two problems are solved.</p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Pull vs push</span>
</div>

<p>In the pull model, an external user periodically polls execution status from the read-only state machine. This model is not real-time and may overload the wallet service if the polling frequency is set too high. Figure 25 shows the pulling model.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-25-periodical-pulling-CFHWDXCB.svg" alt="Figure 25" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>The naive pull model can be improved by adding a reverse proxy [18] between the external user and the event sourcing node. In this design, the external user sends a command to the reverse proxy, which forwards the command to event sourcing nodes and periodically polls the execution status. This design simplifies the client logic, but the communication is still not real-time.</p>
<p>Figure 26 shows the pull model with a reverse proxy added.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-26-pull-model-with-reverse-proxy-S7JBWCC6.svg" alt="Figure 26" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>Once we have the reverse proxy, we could make the response faster by modifying the read-only state machine. As we mentioned earlier, the read-only state machine could have its own behavior. For example, one behavior could be that the read-only state machine pushes execution status back to the reverse proxy, as soon as it receives the event. This will give the user a feeling of real-time response.</p>
<p>Figure 27 shows the push-based model.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-27-push-model-C7LYOOFM.svg" alt="Figure 27" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p>
</section>


<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<div style="margin: 22px auto 12px; max-width: 52rem; width: 96%; box-sizing: border-box; padding-left: 14px; border-left: 3px solid rgba(38,166,154,0.55);">
  <span style="font-size: 1.12em; font-weight: 600; color: #e5edf7;">Distributed transaction</span>
</div>

<p>Once synchronous execution is adopted for every event sourcing node group, we can reuse the distributed transaction solution, TC/C or Saga. Assume we partition the data by dividing the hash value of keys by 2.</p>
<p>Figure 28 shows the updated design.</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-28-final-design-36NRH5KI.svg" alt="Figure 28" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><p>Let’s take a look at how the money transfer works in the final distributed event sourcing architecture. To make it easier to understand, we use the Saga distributed transaction model and only explain the happy path without any rollback.</p>
<p>The money transfer operation contains 2 distributed operations: A-$1 and C+$1. The Saga coordinator coordinates the execution as shown in Figure 29:</p><p class="bbg-figure" style="margin:1rem 0 1.25rem;padding:14px 16px;background:#ffffff;border-radius:10px;border:1px solid rgba(148,163,184,0.4);text-align:center;box-sizing:border-box;"><img src="assets/figure-29-final-design-in-a-numbered-sequence-RLSHN3GF.svg" alt="Figure 29" style="max-width:100%;height:auto;border-radius:8px;background:#ffffff;vertical-align:middle;" /></p><ul>
<li>User A sends a distributed transaction to the Saga coordinator. It contains two operations: A-$1 and C+$1.</li>
<li>Saga coordinator creates a record in the phase status table to trace the status of a transaction.</li>
<li>Saga coordinator examines the order of operations and determines that it needs to handle A-$1 first. The coordinator sends A-$1 as a command to Partition 1, which contains account A’s information.</li>
<li>Partition 1’s Raft leader receives the A-$1 command and stores it in the command list. It then validates the command. If it is valid, it is converted into an event. The Raft consensus algorithm is used to synchronize data across different nodes. The event (deducting $1 from A’s account balance) is executed after synchronization is complete.</li>
<li>After the event is synchronized, the event sourcing framework of Partition 1 synchronizes the data to the read path using CQRS. The read path reconstructs the state and the status of execution.</li>
<li>The read path of Partition 1 pushes the status back to the caller of the event sourcing framework, which is the Saga coordinator.</li>
<li>Saga coordinator receives the success status from Partition 1.</li>
<li>The Saga coordinator creates a record, indicating the operation in Partition 1 is successful, in the phase status table.</li>
<li>Because the first operation succeeds, the Saga coordinator executes the second operation, which is C+$1. The coordinator sends C+$1 as a command to Partition 2 which contains account C’s information.</li>
<li>Partition 2’s Raft leader receives the C+$1 command and saves it to the command list. If it is valid, it is converted into an event. The Raft consensus algorithm is used to synchronize data across different nodes. The event (add $1 to C’s account ) is executed after synchronization is complete.</li>
<li>After the event is synchronized, the event sourcing framework of Partition 2 synchronizes the data to the read path using CQRS. The read path reconstructs the state and the status of execution.</li>
<li>The read path of Partition 2 pushes the status back to the caller of the event sourcing framework, which is the Saga coordinator.</li>
<li>The Saga coordinator receives the success status from Partition 2.</li>
<li>The Saga coordinator creates a record, indicating the operation in Partition 2 is successful in the phase status table.</li>
<li>At this time, all operations succeed and the distributed transaction is completed. The Saga coordinator responds to its caller with the result.</li>
</ul>

<section class="bbg-section" role="region" style="box-sizing:border-box;max-width:52rem;width:96%;margin:0 auto 1.75rem;padding:1.35rem 1.75rem 1.55rem;background:linear-gradient(180deg,#111827,#172033);color:#e5edf7;border:1px solid #263447;border-radius:12px;box-shadow:0 18px 45px rgba(0,0,0,0.28);font-family:Inter,ui-sans-serif,system-ui,-apple-system,BlinkMacSystemFont,'Segoe UI',Roboto,sans-serif;font-size:1.0625rem;line-height:1.72;letter-spacing:0.011em">
<p>In this chapter, we designed a wallet service that is capable of processing over 1 million payment commands per second. After a back-of-the-envelope estimation, we concluded that a few thousand nodes are required to support such a load.</p>
<p>In the first design, a solution using in-memory key-value stores like Redis is proposed. The problem with this design is that data isn't durable.</p>
<p>In the second design, the in-memory cache is replaced by transactional databases. To support multiple nodes, different transactional protocols such as 2PC, TC/C, and Saga are proposed. The main issue with transaction-based solutions is that we cannot conduct a data audit easily.</p>
<p>Next, event sourcing is introduced. We first implemented event sourcing using an external database and queue, but it’s not performant. We improved performance by storing command, event, and state in a local node.</p>
<p>A single node means a single point of failure. To increase the system reliability, we use the Raft consensus algorithm to replicate the event list onto multiple nodes.</p>
<p>The last enhancement we made was to adopt the CQRS feature of event sourcing. We added a reverse proxy to change the asynchronous event sourcing framework to a synchronous one for external users. The TC/C or Saga protocol is used to coordinate Command executions across multiple node groups.</p>
